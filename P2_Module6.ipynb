## Module 6: Provisionnement et évaluation actuarielle

### 6.1 Survol

Le provisionnement assure qu'un régime de retraite dispose de fonds suffisants pour couvrir ses obligations. Les régimes à cotisations déterminées sont naturellement provisionnés par la valeur des cotisations. Pour les régimes à prestations déterminées, une évaluation actuarielle est nécessaire pour estimer les obligations, lesquelles ne correspondent pas toujours exactement aux montants dans la caisse de retraite, menant potentiellement à un excédent ou un déficit.

Les régimes à prestations déterminées doivent être provisionnés à l'avance selon les lois en vigueur, contrairement à une approche "pay-as-you-go" comme celle de la Pension de sécurité de la vieillesse gérée par le gouvernement fédéral.

Une évaluation actuarielle, réalisée par un Fellow de l'institut canadien des actuaires, permet de :
- Déterminer la cotisation nécessaire pour maintenir l'équilibre du régime.
- Identifier et mesurer tout surplus ou déficit actuariel.
- Évaluer si un déficit est dû à des circonstances aléatoires ou à des modifications du régime.
- Calculer la cotisation requise pour amortir un déficit actuariel potentiel.

Le provisionnement est évalué sous deux hypothèses :
- Celle d'une fermeture et liquidation de l'entreprise.
- Celle où l'entreprise reste en activité sur une longue période.

Les deux évaluations, effectuées simultanément, offrent des informations complémentaires.

Trois familles de méthodes sont utilisables pour les évaluations actuarielles :
- Méthodes de répartition des prestations, attribuant une partie de la valeur actuarielle des prestations projetées à une période donnée, en fonction des fluctuations des prestations accumulées ou projetées.
- Méthodes de répartition des cotisations, répartissant la valeur actuarielle des prestations projetées entre diverses périodes selon une approche nivelée.
- Méthodes de projection, affectant la valeur actuarielle des prestations projetées en fonction de la différence entre les valeurs projetées au début et à la fin de la période de projection, ajustées pour les prestations prévues pendant la période.


## 6.2 Raisons justifiant le provisionnement des régimes privés

- Le **provisionnement des régimes de retraite** est crucial pour **assurer le financement** et **respecter les exigences légales** et **comptables**.


- Il offre une **sécurité financière** aux participants, les protégeant contre les risques de non-paiement des prestations, contrairement aux approches **pay-as-you-go**.

- Les **avantages fiscaux** pour les régimes provisionnés incluent la **déductibilité des cotisations** et la **non-imposition des revenus de placement**.

- Les **régimes privés** nécessitent un provisionnement en raison de l'incertitude sur les revenus et la pérennité des entreprises, à l'inverse des régimes publics soutenus par les ressources gouvernementales.

- Les **objectifs** du provisionnement comprennent la **sécurité des prestations**, l'**équité intergénérationnelle**, et une **affectation ordonnée des cotisations** pour une **gestion prudente** du régime.

## 6.3 Mécanisme général d’une évaluation actuarielle

L’évaluation actuarielle est un processus qui se déroule en plusieurs étapes clés :

- **Évaluation du passif et du "coût normal"**, nécessitant :
  - La vérification des dispositions du régime telles que définies dans le règlement du régime, qui est le document légal constituant et régissant le régime.
  - L'examen des données relatives aux participants et bénéficiaires du régime.
  - La formulation d'hypothèses économiques, essentielles pour projeter les conditions de marché futures qui influencneront les rendements des investissements du régime.
  - L'établissement d'hypothèses démographiques, comme les taux de mortalité, de retraite, d'invalidité, etc., impactant le nombre et la valeur des prestations à verser.
  - Le choix d'une méthode d'évaluation appropriée, qui peut varier en fonction des spécificités du régime et des objectifs de l'évaluation.

- **Évaluation de l'actif** :
  - La vérification de la caisse du régime pour s'assurer de la précision des valeurs d'actif déclarées.
  - La sélection d'une méthode d'évaluation de l'actif, qui peut être la valeur marchande pure ou une valeur marchande ajustée pour lisser les gains et pertes sur plusieurs années, selon chaque classe d'actif (actions, obligations, etc.).

- **Estimation du surplus ou déficit actuariel** :
  - En confrontant les valeurs d'actif et de passif estimées, on détermine s'il existe un surplus (les actifs dépassent les passifs) ou un déficit (les passifs dépassent les actifs) actuariel, qui influencera les décisions de gestion du régime.

Le "coût normal" représente le coût du service courant, soit la valeur des prestations qui seront attribuées pour l'année de service suivant la date de l'évaluation. Cette valeur est essentielle pour déterminer les cotisations nécessaires pour maintenir le régime en équilibre.


## 6.4 Provisionnement sur base de liquidation

L’évaluation du provisionnement d’un régime sur base de terminaison évalue la suffisance des fonds par rapport aux obligations existantes à la date d’évaluation, en assumant la fin du régime :

- Le passif actuariel, déterminé par cette évaluation, représente la valeur de toutes les prestations déjà acquises à la date d'évaluation. Pour que le régime soit considéré comme solvable, les actifs doivent au minimum égaler ce passif.

- L'évaluation se base sur l'hypothèse que le régime prend fin à la date d'évaluation.

Cette approche, souvent appelée "base de solvabilité", est utilisée pour mesurer le niveau de provisionnement du régime et pour établir les cotisations nécessaires à couvrir un éventuel déficit actuariel.

Lors de ces calculs, l'actuaire doit :
- Concevoir un scénario de liquidation, comme l'achat de rentes auprès d'une compagnie d'assurance.
- Prendre en compte les prestations conditionnelles qui seraient dues en cas de liquidation, telles qu'une bonification des prestations ou une réduction du facteur d'ajustement pour la retraite anticipée.
- Utiliser la valeur marchande des actifs, supposant leur liquidation à la date de calcul.
- Inclure tous les frais associés à la fermeture du régime.

Les hypothèses actuarielles, définies par la loi, comprennent :
- Un rendement de la caisse basé sur les obligations à long terme du gouvernement du Canada pour les participants actifs, et sur le taux courant d'achat de rente immédiate pour les retraités.
- Aucune augmentation de salaire prévue.
- Généralement, aucune inflation n'est prévue (sauf dans les régimes indexés), du fait de la supposition de liquidation immédiate.
- Utilisation d'une table de mortalité prescrite.
- Choix de l'âge de départ à la retraite qui engendre le coût le plus élevé.
- Supposition qu'aucun employé ne quittera son emploi avant la liquidation.
- Des frais de liquidation réalistes, déduits de l'actif.
- Une méthode actuarielle basée sur la répartition des prestations.

## 6.5 Provisionnement sur base de permanence

L'évaluation du provisionnement d'un régime sur base de permanence évalue la capacité du régime à couvrir ses obligations, en prenant en compte non seulement les obligations actuelles mais aussi les cotisations et prestations futures :

- Le passif actuariel est la valeur actualisée de toutes les prestations attendues jusqu'à la retraite, diminuée de la valeur actualisée des cotisations futures. Les actifs du régime doivent au moins égaler ce passif pour que le régime soit considéré comme solvable.

- Cette évaluation suppose que le régime continue indéfiniment.

Cette approche, souvent appelée "base de continuité" ou "base de provisionnement", est principalement utilisée pour fixer les futurs taux de cotisation.

Pour réaliser cette évaluation, l'actuaire doit :
- Inclure des marges pour écarts défavorables, reconnaissant que les hypothèses ne se réaliseront pas toujours parfaitement.
- Prendre en compte toutes les prestations connues, y compris conditionnelles, attendues tant que le régime est actif.
- Utiliser un taux d'actualisation basé sur le rendement prévu des actifs du régime et la politique de placement, en particulier pour les titres à revenu fixe, supposant une détention jusqu'à maturité.

Les principales hypothèses incluent :
- Un rendement de la caisse basé sur les rendements futurs à très long terme des actifs actuels et futurs, conformément à la politique de placement du régime.
- Des augmentations de salaire prévues jusqu'à la retraite.
- L'inflation, directement pour les régimes indexés et indirectement dans l'estimation des rendements futurs.
- Une mortalité basée sur des tables réalistes.
- Un âge de départ à la retraite basé sur l'expérience du régime et l'impact potentiel de subventions pour la retraite anticipée, avec des taux de départ variant selon l'âge.
- Des taux de cessation d'emploi variant selon l'âge, ce qui peut réduire le coût du régime pour les régimes à salaire final, car une rente basée sur le salaire au moment du départ est généralement moins coûteuse qu'une rente basée sur un salaire projeté à la retraite.
- La proportion de participants ayant un conjoint et la différence d'âge entre les conjoints.
- L'incidence de l'invalidité.
- Les frais administratifs.
- Une marge pour écarts défavorables intégrée dans chaque hypothèse.

Selon les normes professionnelles de l'Institut canadien des actuaires, chaque hypothèse doit être intrinsèquement raisonnable, représentant la meilleure estimation de l'actuaire et incluant une marge pour écarts défavorables si la loi ou le mandat de l'actuaire le requiert.
