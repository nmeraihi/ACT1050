## 3.1 Survol

Les régimes de retraite publics canadiens sont les suivants:

- La Pension de sécurité de la vieillesse, une **allocation universelle** fournie par le Gouvernement du Canada
- Le Supplément de revenu garanti et l’allocation au conjoint, des **prestations d’aide sociale** associées à la Pension de sécurité de la vieillesse et également fournies par le Gouvernement du Canada
- Le Régime de rente du Québec et le Régime de pensions du Canada (quasi identique au Régime de rentes du Québec, mais dans les autres provinces), deux régimes **d’assurance sociale** organisés par les gouvernements provinciaux

## 3.2 Pension de sécurité de la vieillesse (PSV)

**Référence**: [https://www.canada.ca/fr/services/prestations/pensionspubliques/rpc/securite-vieillesse/paiements.html](https://www.canada.ca/fr/services/prestations/pensionspubliques/rpc/securite-vieillesse/paiements.html)


### 3.2.1 Objectif, admissibilité et montant

- L'objectif de la pension de sécurité de la vieillesse est de fournir un revenu de base aux résidents canadiens de 65 ans et plus, débutant le mois suivant leur 65'ème anniversaire.
- Pour être admissible, il faut :
  - Avoir 65 ans ou plus.
  - Avoir résidé au Canada pendant au moins 10 ans après 18 ans.
  - Résider actuellement au Canada ou y avoir résidé plus de 20 ans, ou bénéficier d'une entente de réciprocité internationale.

Pour une personne ayant résidé au Canada pendant **moins de 40 ans** après l’âge de 18 ans, on multiplie le montant normal par le facteur suivant:

$$
\%=\frac{\mathrm{Nb} \text { d'années de résidence après } 18 \text { ans }(\min .10 \text { ou } 20 ; \max 40)}{40}
$$



Une fois qu’une pension partielle est approuvée, elle ne peut pas être augmentée en raison d’années supplémentaires de résidence au Canada.

L’indexation se calcule avec un décalage de 2 mois, comme suit:

$$
P_{\text{Oct 2022 à Déc. 2022}}=P_{\text{Juil 2022 à Sept. 2022 }} \frac{(\mathrm{IPC}_{\text{Mai 2022}}+\mathrm{IPC}_{\text{Juin 2022}}+\mathrm{IPC}_{\text{Juillet 2022}})}{(\mathrm{IPC}_{\text{Février 2022}}+\mathrm{IPC}_{\text{Mars 2022}}+\mathrm{IPC}_{\text{Avril 2022}})}
$$

- La pension de sécurité de la vieillesse est imposable au niveau fédéral et provincial, car elle est considérée comme une allocation universelle.
- Un impôt de récupération spécial de 15% est appliqué sur le revenu total


- Le taux de remplacement de la pension varie selon le revenu du travailleur :
  - 26,7% pour un revenu de 30 000 \\$.
  - 16,0% pour un revenu de 50 000 \\$.
  - 11,4% pour un revenu de 70 000 \\$.

### 3.2.3 Financement
- Le financement de la pension de sécurité de la vieillesse provient des fonds généraux du gouvernement du Canada, selon une méthode "pay-as-you-go", sans pré-provisionnement, ce qui implique que la part des impôts dédiée à ce programme pourrait augmenter avec la hausse de la proportion de personnes de 65 ans et plus.

### 3.2.4 Réformes – Résumé

- **1985 :** Tentative de fusion de la PSV avec le Supplément de revenu garanti pour restreindre l’accès et contrôler les coûts (prévision du vieillissement démographique). Rejetée après forte opposition.
- **1989 :** Instauration d’un impôt de récupération ciblant les bénéficiaires à revenu élevé.
- **2012 :** Plan pour hausser l’âge d’admissibilité de 65 à 67 ans entre 2023 et 2029. Annulé en 2016.
- **Situation actuelle :** Le débat pourrait revenir à cause de la pression budgétaire accrue, notamment due à la pandémie. Sans réforme, le retour à l’équilibre est projeté vers 2070.

### 3.2.5 Exemples de calculs

a) Franck Einstein, maintenant âgé de 65 ans, est retourné vivre en Allemagne après avoir résidé au Canada pendant 25 ans après l’âge de 18 ans. Combien recevra-t-il à titre de Pension de sécurité de la vieillesse ?

Comme il a résidé plus de 20 ans au Canada, il a droit de recevoir sa pension en Allemagne. Il recevra donc (25 / 40) x 666,83\\$ = 416,77 \\$ / mois ou 5 001,23 \\$ / année.

b) Donald Trompé, qui a eu 65 ans en décembre 2015, déclare un revenu total, incluant la Pension de sécurité de la Vieillesse, de 100 000 \\$ en 2022. En supposant que le montant total de la pension se sécurité de la vieillesse reçue en 2022 sera de 8 002 \\$, quelle portion (en \\$) de sa pension doit-il remettre au titre de l’impôt de récupération?

Le seuil de récupération étant de 81 761 \\$ en 2022, il doit remettre le moindre entre 8 002 \\$ ou 15% x (100 000 \\$ - 81 761 \\$) = 2 736 \\$. Il remet donc 2 736 \\$.

c) Joëlle Métellus, qui a eu 65 ans en juillet 2017 et réside au Canada depuis l’âge de 25 ans, a choisi de repousser jusqu’à 70 ans le début de sa pension de sécurité de la vieillesse Combien recevra-t-elle à titre de Pension de sécurité de la vieillesse à compter d’août 2022 ?

Sa pension sera majorée de 60 mois x 0,6% / mois, soit 36%. Elle recevra donc 666,83 x 1,36 = 906,89 \\$ / mois ou 10 883 \\$ / année.

## 3.3 Supplément de revenu garanti (SRG)

**Référence**: [https://www.canada.ca/fr/services/prestations/pensionspubliques/rpc/securite-vieillesse/supplement-revenu-garanti.html](https://www.canada.ca/fr/services/prestations/pensionspubliques/rpc/securite-vieillesse/supplement-revenu-garanti.html)

### 3.3.1 Objectif et admissibilité
Vous pouvez obtenir cette prestation si :

* vous avez 65 ans ou plus;
* vous vivez au Canada;
* vous recevez la pension de la Sécurité de la vieillesse (SV);
* votre revenu est inférieur à 21 624 $ si vous êtes célibataire, veuf ou divorcé;


### 3.3.2 Montant du Supplément de revenu garanti

* Le Supplément de revenu garanti dépend du revenu combiné avec celui du conjoint ou conjoint de fait. 

* Les seuils d’admissibilité varient selon que votre conjoint reçoit ou non la pension de la Sécurité de la vieillesse ou l’Allocation.

* Ces montants sont révisés chaque année.


### 3.3.3 Stratégie fiscale visant le Supplément de revenu garanti

- Reporter la retraite et différer les retraits de régimes (RRQ, REER, rente d’employeur) permet de réduire temporairement le revenu entre 65 et 70 ans.
- Cette baisse de revenu peut donner accès au SRG, tout en bonifiant les rentes futures à partir de 70 ans.

### 3.3.4 Modification au montant du Supplément de revenu garanti

Le montant du Supplément est recalculé en cas de changement des revenus du bénéficiaire ou de modification de la situation familiale.

## 3.4 Allocation au conjoint et Allocation au survivant

**Référence**: [https://www.canada.ca/fr/services/prestations/pensionspubliques/rpc/securite-vieillesse/paiements.html#h2.3](https://www.canada.ca/fr/services/prestations/pensionspubliques/rpc/securite-vieillesse/paiements.html#h2.3)


### 3.4.1 Objectif et admissibilité

- L'Allocation au conjoint et l'Allocation au survivant ciblent les individus de 60 à 64 ans, partenaires de bénéficiaires de la PSV éligibles au Supplément de revenu garanti, ou veufs non remariés, dans le but de les protéger contre la pauvreté. Ces allocations ne sont pas sujettes à l'impôt.
- Conditions d'éligibilité :
    * votre époux ou conjoint de fait est décédé et que, depuis ce temps, vous ne vous êtes pas remarié ni ne vous êtes réengagé dans une union de fait;
    * vous avez entre 60 et 64 ans;
    * vous êtes un citoyen canadien ou un résident autorisé;
    * vous avez résidé au Canada pendant au moins 10 ans depuis l’âge de 18 ans;

### 3.4.2 Montant de l’Allocation au survivant – janvier à mars 2024

| Votre situation | Votre revenu annuel doit être | Montant maximal du paiement mensuel |
|-----------------|------------------------------|------------------------------------|
| Je suis un époux ou conjoint de fait survivant. | moins de 29 112 \\$ | jusqu'à 1 614,89 \\$ |



## 3.5 Régime de rentes du Québec et Régime de pensions du Canada (RRQ/RPC)
(référence : http://www.edsc.gc.ca/fr/rpc/sv/index.page)


### 3.5.1 Objectif du Régime de base

- Le RRQ et le RPC offrent un revenu de base à la retraite pour les travailleurs, incluant les autonomes.
- Ils visent à remplacer environ 25 % du revenu d’emploi moyen, avec un objectif combiné d’environ 40 % incluant la PSV.


#### 3.5.1 Objectif du Régime supplémentaire

- Un régime supplémentaire vise à hausser graduellement le taux de remplacement du revenu admissible de 25 % à 33,33 %.
- Un nouveau plafond (MSGA) et des cotisations additionnelles ont été introduits pour améliorer les rentes futures.


### 3.5.2 Admissibilité et participation au régime
- Tous les travailleurs de 18 ans et plus gagnant plus de 3 500 $ par année sont automatiquement inscrits au régime.
- Il existe une entente de réciprocité entre le RRQ et le RPC, permettant la reconnaissance mutuelle des revenus pour ces régimes.

### 3.5.3 Le Maximum des gains admissibles (MGA) et l’Exemption de base
- Le MGA est ajusté annuellement en fonction de la moyenne des salaires et est arrondi au multiple de 100 \\$ le plus proche. Pour 2025, le MGA est de 71 300 \\$.
- L'Exemption de base, le seuil de revenu jusqu'à 3 500 $ sur lequel on ne cotise pas, est gelé depuis plusieurs années. Cela affecte le calcul des cotisations et des bénéfices du régime.



### 3.5.4 Financement et cotisation

Le régime est financé à parts égales par l’employeur et les employés,
étant précisé que les travailleurs autonomes paient à la fois la part
des employés et celle des employeurs.

### Cotisation

- La cotisation au Régime de rentes du Québec (RRQ) pour un employé en 2024 est déterminée par la formule suivante : 6,15% x MAX [0; MIN (Revenu d’emploi, 68 500 $) – 3 500 $].
- Depuis 2018, un mécanisme automatique ajuste le taux de cotisation au RRQ pour maintenir l'équilibre entre les cotisations et les prestations sur le long terme.
- La période cotisable commence à l'âge de 18 ans et se termine à l'âge de début de la rente, avec un âge maximum de 70 ans, et se calcule en mois.
- Si un travailleur continue son activité en percevant la rente de retraite, il doit continuer de cotiser après la fin de sa période cotisable, ce qui entraîne une augmentation de la rente pour chaque année supplémentaire travaillée.

Exemples de calcul de la période cotisable :
- Pour Matthieu, qui choisit de prendre sa retraite à 60 ans, la période cotisable est de : 12 mois x (60 ans – 18 ans) = 504 mois ou 42 ans.
- Pour Fatima, qui choisit de prendre sa retraite à 65 ans, la période cotisable est de : 12 mois x (65 ans – 18 ans) = 564 mois ou 47 ans.


### 3.5.5 Les différentes prestations

Le RRQ et le RPC offrent principalement des rentes de retraite, mais incluent également des prestations pour invalidité et des prestations aux survivants d'un participant décédé.

##### Rente de retraite de base
- Le calcul de la rente de retraite de base suit la formule : 25 % x (Moyenne du MGA de l'année en cours et des 4 années précédentes) x Ratio moyen des gains / MGA.
- Lors du calcul du ratio moyen des gains par rapport au MGA, 15% des mois avec le plus faible ratio gains/MGA sont exclus de la période cotisable, et le ratio mensuel des gains par rapport au MGA est plafonné à 100% pour chaque mois.

Exemples de calcul de la rente de retraite :
- Pour Fatima, qui a 65 ans et une période cotisable de 564 mois, avec un ratio moyen des gains de 90,83% et un MGA moyen de 62 000 \\$ sur 5 ans, sa rente annuelle est de 14 078,65 \\$, soit 1 173,22 \\$ / mois.
- Matthieu, prenant sa retraite à 60 ans avec une période cotisable de 504 mois et un ratio moyen des gains de 93,59% sur un MGA moyen de 53 480 \\$, reçoit une rente annuelle de 12 513 \\$ ou 1 043 \\$ / mois, réduite à 668 \\$ / mois en raison de la prise de retraite anticipée.



##### Rente de retraite bonifiée
- Le calcul de la rente bonifiée est plus complexe. En 2022, la rente maximale de base est 25% x (MGA moyen sur 5 ans) = 14 925 \\$ / an, soit 1 243,75 \\$ / mois.
- La rente bonifiée pour 2022 est légèrement supérieure à la rente de base avec un montant de 1 253,59 \\$ / mois, représentant une bonification de 9,84 \\$ / mois ou 118,08 \\$ / an.

La rente est ajustée si elle est prise à un âge différent de 65 ans, pour refléter la durée plus longue ou plus courte de versement et le temps de capitalisation des cotisations.

### Prestations d’invalidité

- Le Régime de rentes du Québec (RRQ) propose deux types de prestations en cas d'invalidité :
  - Une rente pour le participant invalide, sous conditions d'admissibilité.
  - Une rente pour chaque enfant d'un participant invalide, si le parent est admissible à la rente d’invalidité.

### Rente d’invalidité du participant

Conditions d'admissibilité à la rente d’invalidité :
- Souffrir d'une invalidité grave et permanente reconnue par le RRQ.
- Être incapable de gagner plus de 19 656 $ en 2022 par le travail.
- Avoir moins de 65 ans.
- Avoir suffisamment cotisé récemment au RRQ.
- Pour les participants de 60 ans ou plus, des conditions d'admissibilité ajustées si incapables d'exercer leur dernier emploi.

Pour les personnes reconnues invalides, la rente de retraite diminue de 0,3 % à 0,4 % pour chaque mois où elles ont reçu :

* une rente de retraite avant 65 ans;
* une rente d'invalidité entre 60 et 65 ans avant janvier 2024.

Habituellement, la rente de retraite diminue plutôt de 0,5 % à 0,6 % pour chaque mois où une personne qui n'est pas reconnue invalide la reçoit avant 65 ans. Cette réduction est appliquée parce que l'âge normal de la retraite au Régime de rente du Québec est 65 ans.

### Exemple 1

Éric reçoit une rente d'invalidité depuis ses 58 ans. En janvier 2024, alors qu'il a 63 ans, une partie de sa rente d'invalidité est remplacée par sa rente de retraite. Cette rente est réduite pour chacun des 24 mois où il la reçoit avant 65 ans. Elle est aussi réduite pour les 36 mois où il a reçu une rente d'invalidité entre 60 et 65 ans avant janvier 2024. Sa rente de retraite est donc réduite à vie de 21 % au total, soit de 0,35 % pour chacun de ces 60 mois.

Cette rente est réduite pour chacun des 60 mois où elle l'aura reçue avant 65 ans. Sa rente de retraite est donc réduite à vie de 21 % au total, soit de 0,35 % pour chacun de ces 60 mois.

### Rente aux enfants d’un participant invalide

Admissibilité de la rente pour enfants :
- Pour les enfants mineurs vivant avec le participant invalide depuis au moins un an.
- Montant fixe de 83,99 \\$ / mois par enfant pour le RRQ en 2022.
- Pour le RPC, la rente est également disponible pour les enfants étudiants à temps plein âgés de 18 à 25 ans, avec un montant de 264,63 $ / mois par enfant en 2022.

### Aspects fiscaux et indexation :
- Les rentes d’invalidité sont imposables et ajustées annuellement en fonction de l’Indice des rentes chaque 1er janvier.


### Prestations en cas de décès du participant

- Le Régime de rentes du Québec offre trois prestations suite au décès d'un participant :
  - Une rente pour le conjoint survivant, si celui-ci est admissible.
  - Une rente pour chaque enfant du participant décédé.
  - Un montant forfaitaire.

Conditions d'attribution des prestations :
- Le décès du participant cotisant au régime.
- Une contribution suffisante au régime, définie comme au moins un tiers de la période cotisable ou un minimum de 10 ans.

### Rente au conjoint survivant

Critères pour la reconnaissance du conjoint survivant :
- Être marié, uni civilement ou conjoint de fait avec le défunt.
- Pour les conjoints de fait, cohabitation de trois ans minimum, ou un an en présence d'un enfant commun ou adopté.

Calcul de la rente pour le conjoint survivant en 2022 :
- Pour les conjoints de moins de 65 ans : **F $ / mois + 37,5% de la somme de la rente de retraite de base et de la rente de retraite supplémentaire**.
- Pour les conjoints de 65 ans et plus : **60% de la rente de retraite**.


### Prestations en cas de décès du participant

Référence :
- [Prestations aux survivants - Régime de rentes du Québec](http://www.rrq.gouv.qc.ca/fr/programmes/regime_rentes/prestations_survivants/Pages/prestations_survivants.aspx)

Le Régime de rentes du Québec offre trois prestations en cas de décès :
- Une rente payable au conjoint survivant, s'il est admissible.
- Une rente payable à chaque enfant d’un participant décédé.
- Un montant forfaitaire.

**F varie selon la situation du conjoint survivant :**

| Âge             | Situation                                 | Montant « F » | Montant maximal |
|-----------------|------------------------------------------|---------------|-----------------|
| moins de 45 ans | sans enfant à charge                     | 134,37 $      | 602,86 $        |
|                 | avec un ou des enfants à charge          | 487,12 $      | 955,61 $        |
|                 | invalide avec ou sans enfant à charge    | 524,61 $      | 993,10 $        |
| entre 45 et 64  | toutes les situations                    | 524,61$       | 993,10 $        |
| 65 ans et plus  | conjoint ne reçoit pas de rente de retraite | Sans objet | 746,65 $        |

Les montants maximaux mensuels varient selon l'âge et la situation.

**Les chiffres publiés par le RRQ sont basés sur une rente de retraite maximale de 1 249,31 $ / mois, ce qui crée une confusion avec la rente du régime de base et celle du régime bonifié.**

### La rente de survivant est imposable.

Dans le cas du RPC, les différences sont :
- Aucune rente n’est payable à un conjoint de moins de 35 ans sans enfant à charge et qui n’est pas invalide.
- Le montant « F » est de 204,69 $ / mois entre 45 ans et 65 ans, sans variation selon la situation du conjoint.
- Entre 35 et 45 ans, la rente est calculée au prorata en fonction de l'âge au décès.

#### Rente d’orphelin

Un enfant d'un participant décédé a droit à une rente fixe si :
- L'enfant est mineur et résidait avec le participant décédé depuis au moins un an.
- En 2022, la rente est de 264,63 $ / mois par enfant pour le RRQ et identique pour le RPC.
- Cette rente est imposable au nom de l'enfant.
- Pour le RPC, la rente est également payable à un enfant entre 18 et 25 ans s'il étudie à temps plein.

#### Prestation forfaitaire au décès

La succession d'un participant décédé peut recevoir :
- RRQ : 2 500 \\$ ou moins selon les frais funéraires.
- RPC : Le minimum entre 6 fois la rente de retraite de base et 2 500 \\$.

Cette prestation est imposable au bénéficiaire.
