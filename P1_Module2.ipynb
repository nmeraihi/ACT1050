# Module 2 : L’encadrement de la profession actuarielle?


**Objectifs du Module 2**


- Connaître le statut juridique de l’actuaire

- Connaître le rôle de l’institut canadien des actuaires

- Connaître les règles de déontologie applicables aux actuaires canadiens

- Comprendre la formation de l’actuaire

- Connaître les associations actuarielles dont l’étudiant fera éventuellement partie

## 2.1 Statut juridique de l’actuaire

### 2.1.1 L’actuariat est-une profession un peu spéciale

Les actuaires ne font pas partie d’un ordre professionnel et ne sont pas régis par le Code des professions du Québec (contrairement aux avocats, comptables, médecins, traducteurs, sexologues, etc.).

L’Office des professions du Québec reconnaît 54 professions et 46 ordres professionnels, dont on peut consulter la liste au moyen du lien ci-dessous : [Office des professions du Québec](https://www.opq.gouv.qc.ca/fileadmin/documents/Systeme_professionnel/54_professions_et_ordres_22_juillet_2015.pdf)

Par conséquent, n’importe qui peut se prétendre actuaire.

Cependant, les lois et règlements suivants font référence à un «actuaire», qui est habituellement défini comme un membre « Fellow » de l’Institut canadien des actuaires:

- La *Loi de l'impôt sur le revenu* du Canada;
- La *Loi sur les impôts* du Québec;
- La *Loi sur le régime de rentes du Québec*;
- Les *Règlements sur les normes de prestation de pension* en vertu de la *Loi canadienne sur les normes de prestation de pension*;
- Les *Règlements sur les prestations de pension* en vertu des lois sur les prestations de pension des provinces de la Colombie-Britannique, de l'Alberta, de la Saskatchewan, du Manitoba, de l'Ontario, de la Nouvelle-Écosse, du Nouveau-Brunswick et de Terre-Neuve;
- L'*International Financial Business Act* de la Colombie-Britannique;

- La *Loi sur le régime de retraite des enseignants* de l'Ontario et la *Loi sur les assurances*;
- La *Loi sur les assurances* et la *Loi sur les accidents du travail et les maladies professionnelles* du Québec;
- La *Loi sur les régimes complémentaires de retraite* du Québec;
- La *Loi sur les sociétés d'assurances*;
- Le *Code criminel du Canada*.
- La *Loi sur les sociétés d'assurances* du Canada


De plus, en 1991, la loi fédérale a créé le rôle d’**« actuaire désigné »** d’une société d’assurance. L’actuaire désigné a les caractéristiques et les responsabilités suivantes :


- Chaque société d’assurance a l’obligation d’avoir un actuaire désigné. Ce dernier peut être un employé de la société ou un consultant externe;
- Il est nommé officiellement par le conseil d'administration de la société d’assurance;
- Il a accès à l'information de gestion,
- Il doit faire rapport au conseil d’administration et aux autorités concernées sur toute transaction ou situation qui, de son avis, a des répercussions défavorables importantes sur la situation financière de la société,
- Il doit faire un rapport annuel au conseil d'administration
- Il doit produire le « Rapport de l'actuaire désigné » qui accompagne les états financiers publiés de la société.


Malgré la facilité avec laquelle on peut se prétendre actuaire, les lois mentionnées ci-dessus ont pour effet qu’on peut considérer un « Fellow » de l’Institut canadien des actuaires comme beaucoup plus actuaire que n’importe qui d’autre.


## 2.2 L’Institut canadien des actuaires

### 2.2.1 Rôle de l’Institut


L'Institut canadien des actuaires (ICA) est l’association nationale représentant les professionnels de l'actuariat au Canada et l’organisme national de la profession actuarielle au Canada. Il a été fondé le 18 mars 1965 en vertu d’une Loi fédérale.

Cette loi stipule que l'Institut a pour objet de:


- Faire avancer et progresser la science actuarielle;
- Favoriser l'application de la science actuarielle à l'activité humaine;
- Établir, de favoriser et de maintenir un niveau élevé de **compétence et d'éthique** dans la profession actuarielle.


On peut trouver beaucoup d’information sur l’institut sur son site web : [ICA Website](https://www.cia-ica.ca/fr/), notamment ici : [À propos de nous - ICA](https://www.cia-ica.ca/fr/les-actuaires/).

Essentiellement, l’ICA :


- Favorise l'avancement de la science actuarielle par la recherche;

- Assure la prestation de programmes de formation et de qualification des membres actuels et éventuels;

- S'assure que les services actuariels fournis par ses membres respectent des normes professionnelles de la plus haute qualité;

- S'auto-réglemente et veille à l'application d'un code de déontologie;

- Défend les intérêts de la profession auprès des gouvernements et du public en ce qui a trait à l'élaboration de ses politiques publiques.

Le [plan stratégique 2023-2026 de l'ICA](https://www.cia-ica.ca/app/themes/wicket/custom/dl_file.php?p=36285&fid=17361) est centré sur l'engagement envers la sécurité financière des Canadiens dans un contexte post-pandémique marqué par des perturbations et défis sans précédent, notamment dans les domaines du changement climatique, de la santé et de la technologie.

Le plan stratégique comprend cinq priorités stratégiques :

1. **Marque de l'ICA** : Renforcer la confiance dans l'ICA en tant qu'autorité sur la profession actuarielle.

2. **Engagement des membres** : Faire de l'ICA un partenaire de carrière à long terme pour les membres.

3. **Intérêt public** : Prioriser le devoir envers le public au-dessus des besoins de la profession.

4. **Technologie** : Utiliser la technologie de manière optimale pour répondre aux besoins des membres.

5. **Services basés sur la connaissance** : Améliorer les compétences des membres grâce à l'éducation et à la recherche.

### 2.2.2 Membres de l’Institut canadien des actuaires


L’ICA regroupe deux types de membres, les Associés et les « Fellows », et compte également des Affiliés et des correspondants.


- Les « Fellows » sont des membres de plein droit, qui seuls ont le droit de signer les documents pour lesquels la Loi exige la signature d’un actuaire. Les « Fellows » sont autorisés à faire suivre leur nom des lettres FICA (Fellow, Institut canadien des actuaires) ou FCIA (Fellow, Canadian Institute of Actuaries). Dès la confirmation de leur inscription à ce titre, tous les « Fellows » sont admissibles au vote sur les sujets relevant de l’Institut.

- Les Associés sont des membres intermédiaires, qui peuvent accomplir toutes les tâches actuarielles sauf celles réservées aux « Fellows ». Ils sont autorisés à faire suivre leur nom des lettres AICA (associé, Institut canadien des actuaires) ou ACIA (Associate, Canadian Institute of Actuaries). Cinq années complètes après avoir accédé au statut d’associé au sein de l’Institut, les membres se voient également attribuer le droit de vote sur les sujets relevant de l’Institut.

- Un Affilié doit être un résident du Canada au moment de présenter sa demande d’inscription et avoir l’intention de devenir « Fellow » de l’ICA après avoir rempli les conditions d’admissibilité au titre de « Fellow ». Les demandes d’adhésion à titre d’affilié entraînent une évaluation initiale de la formation et des qualifications du candidat afin de déterminer l’équivalence de sa formation aux exigences d’admissibilité à l’ICA.


L’un des principes directeurs de l’Institut canadien des actuaires est de placer l’intérêt public au-dessus de l’intérêt de ses membres.


### 2.2.3 Adhésion à l’Institut canadien des actuaires

Avant la fondation de l’Institut et pendant les premières décennies qui ont suivi, il y avait relativement peu d’actuaires au Canada. Les actuaires étaient alors membres d’associations américaines et internationales et l’Institut canadien des actuaires n’acceptait comme membres que des membres à part entière (« Fellows ») de ces associations.

Les deux associations qui ont fourni le plus de membres à l’Institut sont la Society of Actuaries (SOA) et la Casualty Actuarial Society (CAS). D’autres actuaires pratiquant au Canada ont été formés par l’Institute of Actuaries d’Angleterre et par la Faculty of Actuaries d’Écosse, maintenant regroupées sous le nom de IFAO (Institute & Faculty of Actuaries).

Vers la fin du XIX<sup>ème</sup> siècle, les sociétés d’actuaires ont décidé d’uniformiser leurs critères d’admission en imposant des examens d’entrée aux candidats. À l’époque, l’actuariat n’était pas vraiment enseigné à l’université et les actuaires étaient souvent des gens qui avaient appris l’actuariat en travaillant dans diverses sociétés d’assurance.

Le fait que l’ICA admettait principalement des membres de la SOA et de la CAS a eu pour effet que la formation actuarielle a longtemps été basées sur les examens de la SOA et de la CAS. C’est pourquoi, même aujourd’hui, on fait très souvent référence aux examens des sociétés d’actuaires dans la formation des actuaires canadiens.

Au cours des dernières années, l’ICA a adopté une vision selon laquelle la formation des actuaires canadiens relève de l’Institut, qui a développé son propre syllabus d’enseignement. La SOA et la CAS sont maintenant considérés comme des sous-traitants à qui on peut déléguer la responsabilité de valider la formation des nouveaux actuaires. En pratique, même si on parle couramment d’examens de la SOA ou de la CAS, il s’agit essentiellement d’examens reconnus par l’ICA.

Depuis 2015, on peut remplacer certains examens de la CAS ou de la SOA par la réussite (avec une note d’au moins B ou B+ selon le cas) des cours correspondant à ces examens si on étudie dans une université dont le programme d’actuariat est agréé par l’ICA.

En 2021, l’ICA a modifié radicalement son programme d’agrément universitaire. L’agrément sera désormais basé sur l’ensemble du programme d’actuariat d’une université agréée. Les étudiants qui commencent leur programme à compter de l’automne 2021 n’auront plus à faire les examens préliminaires de la SOA et de la CAS pour devenir membres de l’ICA. Il suffira de réussir le baccalauréat en actuariat, plus certains modules et examens administrés par l’ICA. Des règles de transition s’appliqueront aux étudiants qui ont commencé leur programme d’actuariat depuis l’automne 2015.

Ce changement intervient pour plusieurs raisons, dont celles-ci :

- La reconnaissance de la qualité des programmes d’actuariat des universités canadiennes;

- La nécessité de pouvoir réagir rapidement à des changements dans le domaine de l’actuariat en réduisant le délai pour inclure ces changements dans le programme d’études;

- La cohérence avec une formation axée sur le perfectionnement professionnel continu;

- La réduction du temps de cheminement entre l’entrée dans un programme d’actuariat et l’obtention des titres d’Associé, puis de Fellow de l’ICA;

- La réduction du coût total de la formation extra-universitaire;

- Le désir d’attirer dans la profession une population plus diversifiée;

- Une volonté de modernité;

- La nécessité de mettre fin à la dépendance de l’ICA envers la SOA et la CAS.

- Si on prévoit travailler aux États-Unis dans un rôle où l’actuaire doit signer des documents réservés, sans avoir à passer par l’entente de reconnaissance mutuelle entre l’ICA et l’AAA (American Academy of Actuaries).

### 2.2.3.2 Adhésion à l’aide des examens administrés par la SOA


La SOA travaille à l’élaboration d’un programme d’agrément universitaire pour l’obtention du titre d’Associé, qui ressemble à l’ancien système de l’ICA. Cependant, elle ne reconnaît pas le programme d’agrément universitaire de l’ICA. Plus de détails [ici](https://www.soa.org/education/general-info/asa-micro-credentials/).


> La SOA accorde le statut d’Associé aux candidats ayant réussi les examens suivants :

> - Examen P (probabilités)

> - Examen FM (mathématiques financières)

> - Examen FAM (matématiques actuarielles fondamentales)

> - Examen ALTAM (mathématiques actuarielles avancées, à long terme) ou Examen SSTAM (mathématiques actuarielles avancées, à court terme)

> - Examen SRM (statistiques pour la modélisation des risques)

> - Examen PA (analyse prédictive)

> - Examen ATPA (analyse prédictive – sujets avancés)


Les candidats doivent également démontrer leur réussite dans les domaines suivants :

- Statistiques mathématiques

- Économie

- Finance et comptabilité

Et compléter le Module fondamental pré-actuariel, le cours d’éléments fondamentaux de pratique professionnelle, et le cours de professionnalisme à l’intention des futurs Associés de la SOA.

Un Associé de la SOA (ASA) peut devenir Associé de l’ICA en réussissant l’Atelier sur le professionnalisme de l’ICA. Les exigences d’admission à la SOA sont détaillées [ici](https://www.soa.org/education/exam-req/edu-asa-req.aspx).

Pour devenir Fellow de la SOA, il faut d’abord être ASA, puis choisir un domaine de spécialisation et passer trois examens spécialisés, réussir 4 modules spécialisés, un module commun sur la prise de décision et la communication, et le cours d’admission au Fellowship.

Un Fellow de la SOA (FSA) peut devenir Fellow de l’ICA s’il a passé les examens à contenu canadien de la SOA et s’il complète le Cours orienté vers la pratique de l’ICA.

### 2.2.3.2 Adhésion à l’aide des examens administrés par la CAS


La CAS accorde le statut d’Associé aux candidats ayant réussi les examens suivants :

- Examen 1 (probabilités)

- Examen 2 (mathématiques financières)

- Examen 3F (Économie-Finance)

- Examen MAS-I et MAS-II (statistiques actuarielles modernes)

- Examen 5 (techniques de base en tarification et provisionnement)

- Examen 6 (Règlementation et rapports financiers)

Les candidats doivent également démontrer leur réussite dans les domaines de l’économie et la finance et comptabilité, et suivre deux cours en ligne sur la gestion de risque et opérations d’assurance, ainsi que sur la comptabilité d’assurance, analyse des couvertures d’assurance, loi et règlementation de l’assurance.

Un Associé de la CAS (ACAS) peut devenir Associé de l’ICA en réussissant l’Atelier sur le professionnalisme de l’ICA.

Pour devenir « Fellow » de la CAS, il faut d’abord être ACAS, puis passer trois examens spécialisés.

Un « Fellow » de la CAS (FCAS) peut devenir « Fellow » de l’ICA s’il a passé les examens à contenu canadien de la CAS et s’il complète le Cours orienté vers la pratique de l’ICA.

### 2.2.3.4 Autres exigences


Pour devenir « Fellow » de l’Institut canadien des actuaires, il faut aussi avoir suivi le cours orienté vers la pratique, satisfaire à un critère général d’expérience de 3 ans dans le domaine de l’actuariat, et à un critère d’expérience canadienne de 12 mois-personnes de travail actuariel canadien.


## 2.2.4 Règles de déontologie de l’Institut canadien des actuaires

Les actuaires, membres de l’ICA, sont soumis aux règles de déontologie de l’Institut, visant à placer l’intérêt du public avant celui des membres de la profession. Les principales obligations des actuaires incluent l’action avec honnêteté, intégrité, compétence et diligence, la qualification adéquate pour les services rendus, la conformité aux normes de pratique applicables, la divulgation des rémunérations, l’évitement des conflits d’intérêts, la non-utilisation des services pour induire en erreur, la confidentialité des informations, le respect professionnel, la publicité non trompeuse, l’utilisation appropriée des titres et désignations, la divulgation des condamnations criminelles, la collaboration avec les organismes de l’ICA, et la dénonciation de non-conformité apparente aux règles ou aux normes de pratique

## 2.3 Les autres associations d’actuaires

### 2.3.1 La Society of Actuaries

La Society of Actuaries ([SOA](https://www.soa.org)) est un organisme international ayant ses racines aux États-Unis et dédié à la formation des actuaires et à la recherche en actuariat. La SOA compte plus de 30 000 actuaires parmi ses membres.

Ses origines remontent à 1889, alors qu’il y avait moins de 100 actuaires en Amérique du Nord. Son système d’examens a été institué en 1896, d’abord pour tester les qualifications professionnelles acquises en travaillant. Puis, entre 1910 et 1920, l’aspect éducatif des examens a été mis de l’avant, alors qu’on a commencé à y intégrer du matériel pédagogique.

Jusqu’en 2012, la Society of Actuaries concentrait ses activités dans le domaine des risques reliés aux personnes (mortalité morbidité, intérêt, placements, etc.). En 2012, la SOA a décidé d’étendre ses activités à l’assurance IARD, dans le cadre de sa vision stratégique d’être le principal fournisseur d’éducation actuarielle. En fait, la SOA a d’abord offert à la CAS de fusionner avec elle, ce qui aurait permis de centraliser la formation des actuaires issus du système Nord-Américain mais la CAS a refusé, jugeant vraisemblablement qu’il était plus avantageux pour ses membres de demeurer dans un organisme dédié à l’assurance IARD et à la gestion des risques plutôt que de devenir une minorité à l’intérieur d’une organisation beaucoup plus nombreuse.

Présentement, outre une formation de base en actuariat, la SOA offre aussi une formation spécialisée dans les domaines suivants:

- Finance corporative et gestion de risque d’entreprise

- Finance quantitative et investissement

- Assurance vie individuelle et rentes individuelles

- Régimes de retraite

- Assurance maladie et assurance collective

- Assurance IARD


On peut avoir plus de détails ici : [SOA Specialized Training](https://www.soa.org/Education/Exam-Req/edu-fsa-req.aspx)

D’autres informations plus générales sont disponibles ici : [SOA General Information](https://www.soa.org/education/general-info/default.aspx)

Le taux de réussite aux examens avancés varie entre 40% et 60% des candidats. Le taux de réussite aux examens préliminaires devrait être de l’ordre de 30% à 40% des candidats.


### 2.3.2 La Casualty Actuarial Society

La Casualty Actuarial Society ([CAS](https://www.casact.org/)) est un organisme international ayant ses racines aux États-Unis et dédié à la formation des actuaires et à la recherche en actuariat exclusivement en assurance IARD et gestion de risques. La CAS compte plus de 8 000 actuaires parmi ses membres.

Ses origines remontent à 1914. Son système d’examens a été institué en 1915, et incluait alors un examen portant sur les régimes de retraite et les programmes sociaux reliés à la retraite. Aujourd’hui, la Cas se concentre sur l’assurance IARD et la gestion des risques.

Tout comme pour la SOA, le taux de réussite aux examens avancés varie entre 40% et 60% des candidats. Le taux de réussite aux examens préliminaires devrait être de l’ordre de 30% à 40% des candidats.

### 2.3.3 L’American Academy of Actuaries

L’American Academy of Actuaries ([AAA](http://www.actuary.org/)) regroupe plus de 19 500 membres des différentes associations actuarielles en Amérique du Nord oeuvrant dans tous les domaines. 

L’Académie est responsable des règles et normes de pratique et de déontologie aux États-Unis. En particulier, elle est responsable des règles applicables aux actuaires qui signent des opinions actuarielles aux États-Unis.

En pratique, il peut arriver qu’un actuaire soit devenu « Fellow » de l’Institut canadien des actuaires dans le domaine d’expertise de la SOA en passant par le programme d’agrément universitaire. Dans ce cas, l’actuaire ne sera pas membre de la SOA mais pourra quand même signer une opinion actuarielle s’il devient membre de l’Académie en vertu de l’entente de reconnaissance mutuelle de l’Académie et de l’ICA.

L’Académie représente également la profession actuarielle auprès du public américain et des autorités de réglementation des différents États et du gouvernement fédéral des États-Unis.

### 2.3.4 L’Institute and Faculty of Actuaries

L’Institute and Faculty of Actuaries ([IFoA](https://www.actuaries.org.uk/)) est l’organisme professionnel agréé unique consacré à l’éducation, au développement et à l’encadrement des actuaires au Royaume-Uni. L’IFOA résulte de la fusion, en 2010, de la Faculty of Actuaries d’Écosse, fondée en 1856 et de l’Institute of Actuaries d’Angleterre, fondé en 1848.

L’IFoA a pour mission de faire progresser, dans l’intérêt du public, toute chose ayant trait à la science actuarielle et à ses applications. Il a aussi pour mission de régir et de promouvoir la profession actuarielle. Plus d'informations sont disponibles ici : [IFoA Strategy and Objectives](https://www.actuaries.org.uk/about-us/our-strategy-and-objectives)

L’IFoA offre une série d’examens permettant d’obtenir le titre de Fellow, ainsi qu’un programme d’agrément universitaire. Il existe une entente de réciprocité entre l’ICA et l’IFoA permettant aux actuaires d’un pays de pratiquer dans l’autre pays une fois que les exigences d’expérience professionnelle dans le nouveau pays sont satisfaites.

### 2.3.5 L’Association actuarielle internationale (AAI)

Fondée en 1895, l’Association Actuarielle Internationale ([AAI](https://www.actuaries.org)) est le regroupement international des associations actuarielles professionnelles locales; elle compte un certain nombre de sections à vocation spéciale représentant des actuaires individuels. 

L’AAI a pour but de favoriser l’épanouissement d’une profession mondiale reconnue comme possédant les compétences techniques et la fiabilité professionnelle nécessaire pour servir l’intérêt public. Plus d'informations sont disponibles ici : [AAI Sections](https://www.actuaries.org/ABOUT/Brochures/Sections_FR.pdf)

L’AAI, à titre d’organisation internationale regroupant les diverses associations actuarielles, a pour mission :

- de représenter la profession actuarielle et d’en faire valoir le rôle, la réputation et la reconnaissance à l’échelle internationale;

- de promouvoir le professionnalisme, d’élaborer des normes de formation et de stimuler la recherche, en collaboration étroite avec les associations membres et les diverses sections, de façon à répondre à l’évolution des besoins.


Dans ce cadre, L’AAI travaille présentement à l’élaboration d’un syllabus comprenant les sujets sur lesquels elle considère que les actuaires doivent recevoir une formation. Le contenu de ce syllabus sera graduellement intégré au syllabus des sociétés locales. Dans le cas des sociétés du monde anglo-saxon (ICA, SOA, CAS, IFoA), nos examens et exigences de formation devraient demeurer plus exigeants que la base mise sur pied par l’AAI.


### 2.3.6 L’Association nationale des étudiants en actuariat

L’Association nationale des étudiants en actuariat ([ANEA](https://www.anea-asna.ca)) est une association canadienne fondée en 1989 qui a pour but principal de fournir un soutien professionnel à tous les étudiants de la communauté actuarielle canadienne.

L’ANEA tient un congrès annuel au début du mois de janvier, réunissant des étudiants en actuariat et des commanditaires corporatifs de partout au pays pendant trois jours de réseautage. Cette occasion donne aux étudiants des occasions pour apprendre et grandir. Ce congrès a habituellement lieu en janvier. L’ANEA offre aussi des webinaires et de l’information sur des offres d’emploi.

L’ANEA regroupe des étudiants de 19 universités, dont plusieurs ne participent pas au programme d’agrément universitaire de l’ICA.

### 2.3.7 L’International Association of Black Actuaries (IABA)

- **Fondation et Mission** : Fondée en 1992, l'IABA a pour mission de soutenir et encourager la présence et le succès des actuaires noirs, visant à augmenter la diversité dans la profession actuarielle.


- **Programmes Éducatifs** : Offre des bourses d'études, des programmes de mentorat, et des opportunités de réseautage pour aider les étudiants noirs et les professionnels à poursuivre une carrière en actuariat.


- **Promotion de la Diversité** : Travaille activement à la promotion de la diversité et de l'inclusion dans le domaine de l'actuariat, sensibilisant à l'importance de la diversité dans ce secteur.

- **Soutien Professionnel** : Fournit des ressources pour le développement professionnel de ses membres, y compris des séminaires, des ateliers, et des conférences.

- **Réseau Global** : Bien que concentrée sur la communauté noire, l'IABA accueille des membres de toutes les origines et travaille à l'échelle internationale.

- **Engagement Communautaire** : S'engage dans la communauté avec des programmes de sensibilisation pour encourager les jeunes issus de milieux sous-représentés à envisager des carrières en actuariat.

## Résumé des "Règles de déontologie" pour le cours ACT1050 - Janvier 2024

1. **Intégrité Professionnelle (Règle 1)**: Les membres doivent agir avec honnêteté, intégrité et compétence, en remplissant leurs responsabilités envers le public et en maintenant la réputation de la profession actuarielle.

2. **Normes de Qualification (Règle 2)**: Les membres ne doivent fournir des services professionnels que s'ils sont qualifiés pour le faire et répondent aux normes de qualification professionnelle en vigueur.

3. **Normes de Pratique (Règle 3)**: Les services fournis par les membres doivent adhérer aux normes de pratique pertinentes.

4. **Divulgation (Règle 4)**: Les membres doivent divulguer de manière complète et rapide aux clients ou employeurs la source de toute rémunération directe ou indirecte reçue en relation avec une mission professionnelle.

5. **Conflit d'Intérêts (Règle 5)**: Les membres ne doivent pas fournir de services créant un conflit d'intérêts réel ou potentiel, à moins que certaines conditions ne soient remplies, telles que la divulgation complète à toutes les parties concernées.

6. **Contrôle des Produits (Règle 6)**: Les membres doivent prendre des mesures raisonnables pour s'assurer que leurs services ne sont pas utilisés pour induire en erreur les autres ou pour violer ou contourner la loi.

7. **Confidentialité (Règle 7)**: Les informations confidentielles obtenues lors de missions professionnelles ne doivent pas être divulguées, sauf autorisation ou obligation légale.

8. **Courtoisie et Collaboration (Règle 8)**: Les membres doivent réaliser leurs services professionnels avec courtoisie et respect, évitant les critiques injustifiées d'autres membres, et collaborer avec autrui dans l'intérêt des clients ou employeurs.

9. **Publicité (Règle 9)**: Les membres ne doivent pas se livrer à une publicité ou à une sollicitation commerciale pour des services professionnels qui est fausse, trompeuse ou donne une impression négative de la profession.

10. **Titres et Désignations (Règle 10)**: Les membres doivent utiliser uniquement les titres et désignations d'un organisme actuariel reconnu conformément aux pratiques autorisées par cet organisme.

11. **Obligations Parallèles (Règle 11)**: Les membres doivent divulguer toute condamnation pénale conformément à des articles spécifiques des statuts administratifs.

Ces règles soulignent l'importance de la conduite éthique, du professionnalisme, de la transparence et de la responsabilité dans le domaine actuariel. Comprendre et adhérer à ces principes est crucial pour les étudiants aspirant à devenir professionnels dans ce domaine.

### Préambule

Les présentes Règles de déontologie précisent les normes professionnelles et d’éthique auxquelles les membres2 doivent se conformer et ainsi servir l’intérêt public. Les annotations fournissent des explications supplémentaires, des renseignements ou des conseils aux membres de la profession actuarielle sur la façon d’interpréter et d’appliquer les Règles. Les membres ont la responsabilité professionnelle de connaître les Règles et annotations et de se tenir au courant des révisions. En plus de ces Règles, les membres sont assujettis à la loi applicable et aux règles de déontologie ou aux normes d’éthique promulguées par un organisme actuariel reconnu dans les juridictions où ils rendent des services professionnels. Les services professionnels sont réputés être rendus dans les juridictions où les membres ont l’intention qu’ils soient utilisés, à moins d’une entente à l’effet contraire entre l’organisme actuariel reconnu pour de telles juridictions et l’Institut. Les membres sont responsables de se procurer au besoin les traductions de la loi ou des règles de déontologie.


### Définitions

Les termes suivants, qui apparaissent en italiques dans les présentes Règles, ont le sens indiqué ci-après :

- ***Renseignements confidentiels*** : Les renseignements qui ne sont pas du domaine public et dont le membre a pris connaissance conjointement avec les services professionnels exécutés pour le compte d’un client ou d’un employeur. Cela peut englober les renseignements exclusifs ou de diffusion restreinte en vertu de la loi ou que le membre a des raisons de croire que le client ou l’employeur ne souhaiterait pas voir divulgués.

- ***Utilisateur direct*** : Le client ou l’employeur ou toute autre personne qui retient les services du membre, ayant eu la possibilité de choisir le membre et étant à même de communiquer directement avec lui au sujet de ses qualifications, de son travail et de ses recommandations

- ***Rémunération indirecte*** : Toute contrepartie matérielle reçue de quelque source que ce soit relativement à une mission pour laquelle le membre offre ses services professionnels (des exemples pouvant inclure des bonis de volume, des honoraires de démarcheur et des commissions), à l’exception de la rémunération directe pour ces services.

### Services professionnels

La prestation de conseils, de recommandations ou d’opinions qui reposent sur des analyses actuarielles, incluant d’autres services fournis de temps à autre par le membre à un client ou un employeur.


### Organisme actuariel reconnu

Un organisme ayant été accepté en tant que membre titulaire de l’Association Actuarielle Internationale, ou encore un organe établissant des normes de pratique, un organe conseil ou un organe de discipline auquel cet organisme a délégué son autorité.


### Intégrité professionnelle
#### Règle 1

Le membre agit avec honnêteté, intégrité et compétence, et de manière à remplir les responsabilités de la profession envers le public et à maintenir la réputation de la profession actuarielle.

#### Annotations

- ***Annotation 1-1***: Le membre rend ses services professionnels avec habilité et diligence.

- ***Annotation 1-2***: Le membre a la responsabilité professionnelle de ne pas s’associer avec quoi que ce soit qu’il sait, ou devrait savoir, être faux ou trompeur.

- ***Annotation 1-3***: Un membre ne s’engage dans aucune affaire professionnelle impliquant malhonnêteté, fraude, tromperie ou fausse représentation et ne commet aucun acte qui puisse donner une image défavorable de la profession actuarielle.

### Normes de qualification
#### Règle 2

Le membre ne rend des services professionnels que s’il est qualifié pour le faire et qu’il satisfait aux normes de qualification professionnelle continue applicables.

#### Annotations

- ***Annotation 2-1***: Le membre a la responsabilité professionnelle de respecter les normes de qualification professionnelle continue promulguées par un organisme actuariel reconnu dans la juridiction où le membre rend ses services professionnels et de se tenir au courant des révisions apportées à ces normes.

- ***Annotation 2-2***: Les « normes de qualification professionnelle continue » promulguées par l’Institut canadien des actuaires sont définies à l’article 1.01(16.01.2) des statuts administratifs.

### Normes de pratique
#### Règle 3

Le membre veille à ce que les services professionnels rendus par lui ou sous sa direction répondent aux normes de pratique pertinentes.

#### Annotations

- ***Annotation 3-1***: Le membre a la responsabilité professionnelle de respecter les normes de pratique promulguées par un organisme actuariel reconnu dans la juridiction où le membre rend ses services professionnels et de se tenir au courant des principes et normes de pratique reconnus dans la juridiction où il rend ses services professionnels.

### Divulgation
#### Règle 4

Le membre divulgue entièrement et sans retard au client ou à l’employeur la source de toute rémunération indirecte ou directe que lui ou sa compagnie ont reçue ou peuvent recevoir relativement à une mission dans le cadre de laquelle il rend des services professionnels à ce client ou à cet employeur.

#### Annotations

- ***Annotation 4-1***: « Divulgation entière et sans retard » désigne une divulgation de tous les faits importants relativement à la rémunération indirecte ou directe qui peuvent être pertinents aux fins de la décision du client ou de l’employeur et une divulgation suffisamment rapide pour permettre au client ou à l’employeur de prendre une décision éclairée et indépendante. Cette divulgation devrait se faire par écrit.

- ***Annotation 4-2***: S’il n’est pas indépendant sur les plans financier et organisationnel relativement à toute question liée aux services professionnels rendus, le membre devrait divulguer entièrement et sans retard au client ou à l’employeur toute relation pertinente non apparente.

- ***Annotation 4-3***: Le membre qui travaille pour une compagnie établie à plusieurs endroits est soumis aux exigences de divulgation entière et sans retard des sources de la rémunération que la compagnie pour laquelle il travaille a reçue ou peut recevoir en rapport avec les services professionnels qui touchent la mission spécifique exécutée pour ce client, sans égard à l’endroit où cette rémunération est reçue.

### Conflits d’intérêts
#### Règle 5

Le membre ne rend pas de services professionnels qui puissent susciter un conflit d’intérêt réel ou potentiel, sauf :  
(a) si la capacité du membre d’agir équitablement n’est pas affaiblie;  
(b) si le conflit d’intérêt a été divulgué entièrement et sans retard à tous les utilisateurs directs actuels et éventuels connus; et  
(c) si tous les utilisateurs directs actuels et éventuels connus ont expressément convenu de l’exécution des services par le membre.

#### Annotations

- ***Annotation 5-1***: « Divulgation entière et sans retard » désigne une divulgation de tous les faits importants relativement au conflit, incluant la nature de l’influence ou de la relation, de même que la nature et l’ampleur de l’intérêt, qui peuvent être pertinents aux fins de la décision de l’utilisateur direct et une divulgation suffisamment rapide pour permettre à l’utilisateur direct de prendre une décision éclairée et indépendante. Cette divulgation devrait se faire par écrit.

### Contrôle du produit


#### Règle 6

Le membre qui rend des services professionnels doit prendre des mesures raisonnables pour s’assurer que ces services ne soient pas utilisés pour induire en erreur d’autres parties ou pour enfreindre ou contourner la loi.

##### Annotation 6-1

Les documents préparés par le membre peuvent être utilisés par une autre partie d’une manière susceptible d’influer sur les agissements d’un tiers. Le membre devrait reconnaître le risque que ces documents puissent être mal cités, mal interprétés ou autrement mal utilisés et prendre des mesures raisonnables pour veiller à ce que l’information soit précise et présentée de façon équitable.

##### Annotation 6-2

Les documents préparés par le membre devraient identifier le client ou l’employeur pour lequel ils sont préparés. Ces documents devraient également identifier celui-ci comme source des renseignements, en quelle qualité le membre a agi à cet égard, tout aspect des services professionnels dont le membre n’assume pas la responsabilité et la mesure dans laquelle le membre ou d’autres personnes seront disponibles pour fournir au client ou à l’employeur des renseignements et des explications supplémentaires le cas échéant.


### Confidentialité


#### Règle 7

Nul membre ne peut divulguer à une autre partie les renseignements confidentiels obtenus dans le cadre d’une mission professionnelle exécutée pour le compte d’un client ou d’un employeur, sauf s’il y est autorisé explicitement ou implicitement par le client ou l’employeur, qu’il y est tenu en vertu de la Règle 13, ou si le Conseil de déontologie, une équipe d’enquête, un tribunal disciplinaire ou un tribunal d’appel lui en a fait la demande relativement à toute question disciplinaire prévue à la section 20 des statuts administratifs, ou qu’il y est tenu par la loi.


### Courtoisie et collaboration


#### Règle 8

Le membre rend ses services professionnels avec courtoisie et respect professionnel, évite les critiques injustifiables ou déplacées à l’égard d’autres membres et accorde sa collaboration aux autres dans l’intérêt du client ou de l’employeur.

##### Annotation 8-1

Il peut surgir des différences d’opinions entres membres, particulièrement dans le choix des hypothèses et des méthodes. Les discussions sur ces différences, que ce soit directement entre membres ou dans le cadre d’observations présentées à un client par un membre sur le travail d’un autre, devraient se dérouler en toute objectivité ainsi qu’avec courtoisie et respect.

##### Annotation 8-2

Dans le cours d’une mission ou de son emploi, le membre peut se trouver dans une situation telle qu’il serait dans les meilleurs intérêts du client ou de l’employeur que le membre formule une opinion différente de celle exprimée par un autre membre, ainsi qu’une explication des facteurs qui appuient cette deuxième opinion. Aucun article des Règles ne peut être réputé empêcher le membre d’exprimer au client ou à l’employeur cette deuxième opinion.

##### Annotation 8-3

Si le membre est invité à conseiller un client ou un employeur et qu’il sait ou a des motifs raisonnables de croire qu’un autre membre agit déjà à titre professionnel relativement à la même question ou a agi en cette qualité récemment, il serait normalement prudent de consulter l’autre membre, afin de se préparer adéquatement à la mission et de formuler un jugement informé quant à savoir s’il existe des circonstances relatives à une infraction potentielle aux Règles, qui pourraient influer sur l’acceptation ou le refus de la mission.


### Publicité


#### Règle 9

Le membre ne s’adonne à quelque activité de publicité ou de sollicitation d’affaires à l’égard de services professionnels dont il sait ou devrait savoir qu’elle est fausse ou trompeuse ou qui donne une image défavorable de la profession ou de la compétence et de l’intégrité de l’un de ses membres.

##### Annotation 9-1

« Publicité et sollicitation d’affaires » englobe toutes les communications, quel que soit le moyen, communications verbales comprises, qui peuvent directement ou indirectement exercer une influence sur une personne ou une organisation dans la décision de savoir si elle a besoin de services professionnels ou pour choisir telle personne ou telle entreprise pour l’exécution de tels services professionnels.


### Titres et désignations


#### Règle 10

Le membre n’utilise les titres d’affiliation et désignations d’un organisme actuariel reconnu que si cet usage est conforme à la pratique autorisée par cet organisme.

##### Annotation 10-1

«Titre» s’entend de tout titre conféré par un organisme actuariel reconnu à l’égard d’un poste particulier au sein de cet organisme. «Désignation» s’entend de la mention expresse du statut de membre de cet organisme.


### Obligations parallèles


#### Règle 11

Le membre divulgue toute condamnation au criminel conformément à l’article 3.1.12(1), 3.1.12(2) ou 3.1.12(3) des statuts administratifs.

##### Annotation 11-1

« Condamnation au criminel » fait un renvoi à la définition de « condamnation au criminel » précisée dans les statuts administratifs.

##### Annotation 11-2

Une condamnation au criminel divulguée par un membre conformément à l’article 3.1.12(1), 3.1.12(2) ou 3.1.12(3) des statuts administratifs est assujettie à un examen et à une évaluation par un conseil, une direction ou une commission désignée de l’Institut qui peut déterminer que la question devrait être référée au Conseil de déontologie et soumise aux procédures disciplinaires de l’Institut.

#### Règle 12

Le membre répond promptement, sincèrement et complètement à toute demande de renseignements reçue du Conseil de déontologie, d’une équipe d’enquête, d’un tribunal disciplinaire, d’un tribunal d’appel ou de tout membre de ces groupes, et offre son entière collaboration à ces derniers relativement à toute question disciplinaire prévue à la section 20 des statuts administratifs.

#### Règle 13

Un membre qui prend connaissance d’un cas important de non-conformité apparente aux Règles ou aux normes de pratique de la part d’un autre membre doit tenter de discuter de la situation avec l’autre membre et résoudre la nonconformité apparente. En l’absence de discussion et de résolution, le membre doit signaler la non-conformité apparente au Conseil de déontologie, sauf si un tel signalement est contraire à la loi ou lorsque le membre est appelé à intervenir dans un contexte conflictuel, et ce pour la durée du contexte conflictuel.

Dans le but de favoriser l’éducation parmi les membres et ainsi acquitter les responsabilités de la profession envers le public, un membre qui s’interroge sur l’esprit ou l’intention des normes de pratique, ou de la pratique actuarielle généralement reconnue si aucune norme n’existe, peut consulter en toute confidentialité le président (ou vice-président) d’une Direction désignée constituée par le Conseil d’administration conformément à l’article 9.06 des statuts administratifs ou d’une commission de pratique appropriée. Lorsqu’il est consulté en sa qualité de président (ou vice-président) et prend connaissance d’un cas important de non-conformité apparente aux normes de pratique de la part d’un autre membre, il ne doit pas signaler la non-conformité apparente au Conseil de déontologie.

Lorsqu’un membre, en sa qualité de titulaire d’un poste au sein de l’Institut désigné par le Conseil d’administration de temps à autre par résolution, ou de toute autre entité désignée ainsi, prend connaissance d’un cas important de non-conformité apparente aux normes de pratique de la part d’un autre membre, il ne doit pas signaler la non-conformité apparente au Conseil de déontologie.

***Un membre ne s’engage dans aucune affaire professionnelle impliquant
malhonnêteté, fraude, tromperie ou fausse représentation et ne commet
aucun acte qui puisse donner une image défavorable de la profession
actuarielle.

**Normes de qualification**

**Règle 2**

Le membre ne rend des services professionnels que s’il est qualifié pour
le faire et qu’il satisfait aux normes de qualification professionnelle
continue applicables.

***Annotation 2-1***

Le membre a la responsabilité professionnelle de respecter les normes de
qualification professionnelle continue promulguées par un organisme
actuariel reconnu dans la juridiction où le membre rend ses services
professionnels et de se tenir au courant des révisions apportées à ces
normes.

**Annotation 2-2**

Les « normes de qualification professionnelle continue » promulguées par
l’Institut canadien des actuaires sont définies à l’article
1.01(16.01.2) des statuts administratifs.

**  
**

**Normes de pratique**

**Règle 3**

Le membre veille à ce que les services professionnels rendus par lui ou
sous sa direction répondent aux normes de pratique pertinentes

***Annotation 3-1  
***Le membre a la responsabilité professionnelle de respecter les normes
de pratique promulguées par un organisme actuariel reconnu dans la
juridiction où le membre rend ses services professionnels et de se tenir
au courant des principes et normes de pratique reconnus dans la
juridiction où il rend ses services professionnels.

**Divulgation**

**Règle 4**

Le membre divulgue entièrement et sans retard au client ou à l’employeur
la source de toute rémunération indirecte ou directe que lui ou sa
compagnie ont reçue ou peuvent recevoir relativement à une mission dans
le cadre de laquelle il rend des services professionnels à ce client ou
à cet employeur.

***Annotation 4-1***  
« Divulgation entière et sans retard » désigne une divulgation de tous
les faits importants relativement à la rémunération indirecte ou directe
qui peuvent être pertinents aux fins de la décision du client ou de
l’employeur et une divulgation suffisamment rapide pour permettre au
client ou à l’employeur de prendre une décision éclairée et
indépendante. Cette divulgation devrait se faire par écrit.

***Annotation 4-2  
***S’il n’est pas indépendant sur les plans financier et organisationnel
relativement à toute question liée aux services professionnels rendus,
le membre devrait divulguer entièrement et sans retard au client ou à
l’employeur toute relation pertinente non apparente

***Annotation 4-3  
***Le membre qui travaille pour une compagnie établie à plusieurs
endroits est soumis aux exigences de divulgation entière et sans retard
des sources de la rémunération que la compagnie pour laquelle il
travaille a reçue ou peut recevoir en rapport avec les services
professionnels qui touchent la mission spécifique exécutée pour ce
client, sans égard à l’endroit où cette rémunération est reçue.

**  
**

**Conflits d’intérêts**

**Règle 5**

Le membre ne rend pas de services professionnels qui puissent susciter
un conflit d’intérêt réel ou potentiel, sauf :  
(a) si la capacité du membre d’agir équitablement n’est pas affaiblie;  
(b) si le conflit d’intérêt a été divulgué entièrement et sans retard à
tous les utilisateurs directs actuels et éventuels connus; et  
(c) si tous les utilisateurs directs actuels et éventuels connus ont
expressément convenu de l’exécution des services par le membre.

***Annotation 5-1  
***« Divulgation entière et sans retard » désigne une divulgation de
tous les faits importants relativement au conflit, incluant la nature de
l’influence ou de la relation, de même que la nature et l’ampleur de
l’intérêt, qui peuvent être pertinents aux fins de la décision de
l’utilisateur direct et une divulgation suffisamment rapide pour
permettre à l’utilisateur direct de prendre une décision éclairée et
indépendante. Cette divulgation devrait se faire par écrit

**Contrôle du produit**

**Règle 6**

Le membre qui rend des services professionnels doit prendre des mesures
raisonnables pour s’assurer que ces services ne soient pas utilisés pour
induire en erreur d’autres parties ou pour enfreindre ou contourner la
loi.

***Annotation 6-1***  
Les documents préparés par le membre peuvent être utilisés par une autre
partie d’une manière susceptible d’influer sur les agissements d’un
tiers. Le membre devrait reconnaître le risque que ces documents
puissent être mal cités, mal interprétés ou autrement mal utilisés et
prendre des mesures raisonnables pour veiller à ce que l’information
soit précise et présentée de façon équitable.

***Annotation 6-2  
***Les documents préparés par le membre devraient identifier le client
ou l’employeur pour lequel ils sont préparés. Ces documents devraient
également identifier celui-ci comme source des renseignements, en quelle
qualité le membre a agi à cet égard, tout aspect des services
professionnels dont le membre n’assume pas la responsabilité et la
mesure dans laquelle le membre ou d’autres personnes seront disponibles
pour fournir au client ou à l’employeur des renseignements et des
explications supplémentaires le cas échéant.

**  
**

**Confidentialité**

**Règle 7**

Nul membre ne peut divulguer à une autre partie les renseignements
confidentiels obtenus dans le cadre d’une mission professionnelle
exécutée pour le compte d’un client ou d’un employeur, sauf s’il y est
autorisé explicitement ou implicitement par le client ou l’employeur,
qu’il y est tenu en vertu de la Règle 13, ou si le Conseil de
déontologie, une équipe d’enquête, un tribunal disciplinaire ou un
tribunal d’appel lui en a fait la demande relativement à toute question
disciplinaire prévue à la section 20 des statuts administratifs, ou
qu’il y est tenu par la loi.

**Courtoisie et collaboration**

**Règle 8**

Le membre rend ses services professionnels avec courtoisie et respect
professionnel, évite les critiques injustifiables ou déplacées à l’égard
d’autres membres et accorde sa collaboration aux autres dans l’intérêt
du client ou de l’employeur.

***Annotation 8-1  
***Il peut surgir des différences d’opinions entres membres,
particulièrement dans le choix des hypothèses et des méthodes. Les
discussions sur ces différences, que ce soit directement entre membres
ou dans le cadre d’observations présentées à un client par un membre sur
le travail d’un autre, devraient se dérouler en toute objectivité ainsi
qu’avec courtoisie et respect.

***Annotation 8-2***  
Dans le cours d’une mission ou de son emploi, le membre peut se trouver
dans une situation telle qu’il serait dans les meilleurs intérêts du
client ou de l’employeur que le membre formule une opinion différente de
celle exprimée par un autre membre, ainsi qu’une explication des
facteurs qui appuient cette deuxième opinion. Aucun article des Règles
ne peut être réputé empêcher le membre d’exprimer au client ou à
l’employeur cette deuxième opinion.

**Annotation 8-3  
**Si le membre est invité à conseiller un client ou un employeur et
qu’il sait ou a des motifs raisonnables de croire qu’un autre membre
agit déjà à titre professionnel relativement à la même question ou a agi
en cette qualité récemment, il serait normalement prudent de consulter
l’autre membre, afin de se préparer adéquatement à la mission et de
formuler un jugement informé quant à savoir s’il existe des
circonstances relatives à une infraction potentielle aux Règles, qui
pourraient influer sur l’acceptation ou le refus de la mission.

Le membre qui agit comme conseiller prospectif supplémentaire ou nouveau
devrait demander au client ou à l’employeur de consentir à cette
consultation. Lorsque le client ou l’employeur a donné son autorisation,
le membre initial collabore en fournissant les renseignements voulus
tels que les données pertinentes, documents de travail et autres, et
peut exiger une rémunération raisonnable à l’égard du travail nécessaire
pour réunir et transmettre les renseignements voulus. Le membre initial
ne peut refuser de consulter ou de collaborer avec le membre à cause de
questions de rémunération non résolues avec le client ou l’employeur, à
moins que ce refus soit conforme à une entente préalable avec le client
ou l’employeur. Il n’est pas tenu d’inclure quoi que ce soit d’exclusif,
par exemple des communications internes ou des logiciels.

**Publicité**

**Règle 9**

Le membre ne s’adonne à quelque activité de publicité ou de
sollicitation d’affaires à l’égard de services professionnels dont il
sait ou devrait savoir qu’elle est fausse ou trompeuse ou qui donne une
image défavorable de la profession ou de la compétence et de l’intégrité
de l’un de ses membres

***Annotation 9-1  
***« Publicité et sollicitation d’affaires » englobe toutes les
communications, quel que soit le moyen, communications verbales
comprises, qui peuvent directement ou indirectement exercer une
influence sur une personne ou une organisation dans la décision de
savoir si elle a besoin de services professionnels ou pour choisir telle
personne ou telle entreprise pour l’exécution de tels services
professionnels.

**Titres et désignations**

**Règle 10**

Le membre n’utilise les titres d’affiliation et désignations d’un
organisme actuariel reconnu que si cet usage est conforme à la pratique
autorisée par cet organisme.

***Annotation 10-1  
***«Titre» s’entend de tout titre conféré par un organisme actuariel
reconnu à l’égard d’un poste particulier au sein de cet organisme.  
«Désignation» s’entend de la mention expresse du statut de membre de cet
organisme.

**Obligations parallèles**

**Règle 11**

Le membre divulgue toute condamnation au criminel conformément à
l’article 3.1.12(1), 3.1.12(2) ou 3.1.12(3) des statuts administratifs.

***Annotation 11-1***  
« Condamnation au criminel » fait un renvoi à la définition de «
condamnation au criminel » précisée dans les statuts administratifs. \[

***  
***

***Annotation 11-2  
***Une condamnation au criminel divulguée par un membre conformément à
l’article 3.1.12(1), 3.1.12(2) ou 3.1.12(3) des statuts administratifs
est assujettie à un examen et à une évaluation par un conseil, une
direction ou une commission désignée de l’Institut qui peut déterminer
que la question devrait être référée au Conseil de déontologie et
soumise aux procédures disciplinaires de l’Institut.

**Règle 12**

Le membre répond promptement, sincèrement et complètement à toute
demande de renseignements reçue du Conseil de déontologie, d’une équipe
d’enquête, d’un tribunal disciplinaire, d’un tribunal d’appel ou de tout
membre de ces groupes, et offre son entière collaboration à ces derniers
relativement à toute question disciplinaire prévue à la section 20 des
statuts administratifs.

**Règle 13**

Un membre qui prend connaissance d’un cas important de non-conformité
apparente aux Règles ou aux normes de pratique de la part d’un autre
membre doit tenter de discuter de la situation avec l’autre membre et
résoudre la nonconformité apparente. En l’absence de discussion et de
résolution, le membre doit signaler la non-conformité apparente au
Conseil de déontologie, sauf si un tel signalement est contraire à la
loi ou lorsque le membre est appelé à intervenir dans un contexte
conflictuel, et ce pour la durée du contexte conflictuel.

Dans le but de favoriser l’éducation parmi les membres et ainsi
acquitter les responsabilités de la profession envers le public, un
membre qui s’interroge sur l’esprit ou l’intention des normes de
pratique, ou de la pratique actuarielle généralement reconnue si aucune
norme n’existe, peut consulter en toute confidentialité le président (ou
vice-président) d’une Direction désignée constituée par le Conseil
d’administration conformément à l’article 9.06 des statuts
administratifs ou d’une commission de pratique appropriée. Lorsqu’il est
consulté en sa qualité de président (ou vice-président) et prend
connaissance d’un cas important de non-conformité apparente aux normes
de pratique de la part d’un autre membre, il ne doit pas signaler la
non-conformité apparente au Conseil de déontologie.

Lorsqu’un membre, en sa qualité de titulaire d’un poste au sein de
l’Institut désigné par le Conseil d’administration de temps à autre par
résolution, ou de toute autre entité désignée ainsi, prend connaissance
d’un cas important de non-conformité apparente aux normes de pratique de
la part d’un autre membre, il ne doit pas signaler la non-conformité
apparente au Conseil de déontologie.
