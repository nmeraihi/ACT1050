## Module 4 : Aspect fiscal

### Objectifs du Module #4

- Comprendre le rôle du gouvernement dans l’établissement des paramètres d'aide fiscale à l'épargne-retraite.
- Comprendre le rôle de la Loi de l'impôt sur le revenu dans la détermination des limites de cotisations à un régime de retraite et à un REER.
- Connaître le calcul du facteur d'équivalence (FE).
- Comprendre les limites applicables aux régimes à prestations déterminées.

### 4.1 Survol

Les gouvernements encouragent l'épargne pour la retraite pour plusieurs raisons :
- Les revenus de retraite peuvent réduire ou éliminer la nécessité du Supplément de revenu garanti, allégeant ainsi les paiements fédéraux.
- L'accumulation d'argent dans les régimes de retraite se fait à l'abri de l'impôt, générant un revenu imposable lorsque retiré, créant ainsi un actif d'impôt différé pour les gouvernements.

Pour encourager l'épargne pour la retraite, les gouvernements appliquent une fiscalité favorable, notamment :
- La déduction fiscale des cotisations aux régimes enregistrés, dans certaines limites.
- L'exonération fiscale des revenus de placement dans ces régimes jusqu'à leur retrait.

### 4.2 Loi de l’impôt sur le revenu

#### 4.2.1 Contexte

* La Loi de l'impôt sur le revenu est une **loi fédérale**. Pour la plupart des provinces, à l'exception du Québec, l'impôt provincial est un pourcentage de l'impôt fédéral.
* Le gouvernement fédéral définit donc les règles fiscales principales et les provinces s'alignent généralement sur celles-ci.
* Le Québec a son propre système d'impôt sur le revenu mais tend à suivre les règles fédérales, tout en y ajoutant ses propres spécificités.

#### Cadre fiscal de l’épargne-retraite

- La loi C-52 (1990) a uniformisé les règles de cotisation entre REER et régimes collectifs, renforçant l’équité fiscale.
- Elle a aussi prévu des hausses et une indexation des plafonds de cotisation pour encourager l’épargne à long terme.


### 4.2.2 Agrément des régimes

- Pour profiter d’un avantage fiscal, un régime de retraite doit être enregistré auprès de l’ARC (agrément).
- Les cotisations sont déductibles, les revenus croissent à l’abri de l’impôt, mais les prestations sont imposables à la sortie.


#### Obtention de l’agrément

- Un régime est agréé s’il respecte les règles fiscales : versements de retraite admissibles, limites de cotisation, début des prestations avant 71 ans, etc.
- Il doit aussi démontrer un financement adéquat, interdire la cession des droits et se conformer aux normes actuarielles reconnues.


#### Régime individuel : Le REER

Le REER est un régime individuel enregistré d’épargne retraite. On peut y cotiser jusqu’aux limites prévues par la Loi de l’impôt sur le revenu. Les fonds s’y accumulent à l’abri de l’impôt mais tout retrait est pleinement imposable. À la retraite, on ne peut utiliser un REER que pour acheter une rente ou un FERR (Fonds enregistré de revenu de retraite).

#### Régimes non agréés (non enregistrés)

- Ces régimes ne donnent aucun avantage fiscal : les cotisations sont imposables pour les employés et les revenus de placement aussi.
- Ils sont surtout utilisés pour offrir des prestations au-delà des plafonds fiscaux, notamment à des hauts salariés dans certaines grandes entreprises.


#### Compte d’épargne libre d’impôt (CELI)

- Le CELI permet une croissance non imposable des placements, avantageuse sur le long terme.
- Il peut être utilisé stratégiquement pour maximiser le Supplément de revenu garanti en réduisant temporairement le revenu entre 65 et 70 ans.


### 4.2.3 Limites d’épargne retraite

- La loi vise à offrir le même niveau d'aide fiscale à l'épargne retraite pour tous les contribuables ayant le même revenu, indépendamment du mode d'épargne.
- L’aide fiscale est limitée à 18 % du revenu gagné, selon le type de régime (RPA ou REER), avec un plafond annuel variable.
- Les prestations dans un régime à prestations déterminées sont converties en équivalents de cotisation, et aucun avantage fiscal n’est accordé au-delà des limites permises.

### 4.2.4 Facteur d’équivalence (FE)

La loi C-52 de 1990 a créé la notion de facteur d’équivalence, afin de ramener sur une même base comparative les régimes à prestations déterminées, ceux à cotisations déterminées et les REER.

Les employeurs qui offrent un régime à prestations déterminées ou à cotisations déterminées sont tenus de déclarer, dans les 60 jours qui suivent la fin de chaque année civile, le FE des participants via le feuillet d’impôt T-4.

Dans le cas d'un régime à prestations déterminées, le facteur d'équivalence représente donc la valeur présumée (et non pas la valeur réelle) du crédit de rente gagné au cours de l'année précédente dans le cadre du régime de retraite. On le calcule comme suit :

**FE = 9 x Crédit de rente – 600 \\$**


Le 600\\$ soustrait au FE vise à permettre aux participants à un régime à prestations déterminées dont le crédit de rente est élevé de quand même pouvoir cotiser à un REER.

Dans le cas d'un régime à cotisations déterminées, le facteur d'équivalence est la valeur totale de la cotisation de l'employeur et de celle de l'employé participant, soit :

$$
\text { FE }=\text { Cotisation}^{\text{ Employeur }}+ \text { Cotisation}^{\text{ Employé }}
$$

**Exemple 1** :
Un employé gagne 60 000 \\$. Son régime de retraite à prestations déterminées (formule à salaire de carrière), lui crédite cette année une rente égale à 1,5% de son salaire. Son FE sera alors : 9 x (1,5% x 60 000 \\$) – 600 \\$ = 7 500 \\$


**Exemple 2** :
Un employé gagne 60 000 \\$. Son régime de retraite à prestations déterminées (formule à salaire final), lui crédite cette année une rente égale à 1,5% de son salaire. Son FE sera alors : 9 x (1,5% x 60 000 \\$) – 600 \\$ = 7 500 \\$, soit exactement le même que dans l’exemple précédent, parce que la mécanique du calcul du FE ne tient pas compte du type de formule d'un régime à prestations déterminées.

**Exemple 3** :
Un employé gagne 60 000 \\$. Son régime de retraite est à cotisations déterminées. Il y cotise 4% de son salaire et l'employeur cotise lui aussi 4% du salaire.

Son FE sera alors : (4% + 4%) x 60 000 \\$ = 4 800 \\$. Par conséquent, on peut considérer que le gouvernement considère que ce régime est 36% moins généreux que les régimes à prestations déterminées auxquels participent les employés des exemples 1 et 2.

Certaines caractéristiques du FE sont dignes de mention :

- Le FE ne varie pas selon l'âge du participant (autrement dit, pour un même crédit de rente, un participant de 25 ans et un participant de 60 ans auront le même FE.).
- Le FE ne varie pas selon le sexe.
- Le FE ne varie pas selon l'environnement des taux d'intérêt.
- Par conséquent, le FE surestime la valeur du crédit de rente à certains âges et sous-estime cette valeur à d'autres âges.

On peut tirer quelques conclusions intéressantes du tableau précédent :

- Les régimes à prestations déterminées sont nettement plus avantageux avec des taux d'intérêt bas, leur valeur pouvant être jusqu'à 7 fois supérieure pour un jeune de 25 ans avec un taux de 2% par rapport à un taux de 6%.
- Le facteur d'équivalence (FE) initial, conçu dans un contexte de taux d'intérêt élevés, tendait à surévaluer les crédits de rente des régimes à prestations déterminées.
- En considérant les frais administratifs et les rendements réduits pour les particuliers, le facteur de 9 devient moins irréaliste, en particulier avec les taux d'intérêt actuellement bas.
- Avec une rente indexée à l'inflation et un rendement net de 2% à 3%, le facteur de 9 est réaliste et correspond à la valeur actualisée d'une rente différée vers 40 ans, débutant à 65 ans, ce qui indique que le législateur aurait envisagé des régimes très généreux lors de l'établissement de ce facteur.



Au-delà de la mathématique, si on suppose que le facteur 9 est raisonnable, et qu’on applique la formule du FE à un employé qui gagne 85 000 \\$ et participe à un régime qui crédite 1,3% du salaire jusqu’à concurrence du MGA de 2022 (64 900 \\$) et 2% du salaire en excédent, (régime à prestations déterminées typique), on obtient un FE égal à  

9 x [ 1,3 % x 64 900 \$ + 2% x (85 000 \$ - 64 900 \$) ] – 600 \$  

= 9 x 1 245,70 \$ - 600 \$  

= 10 611 \$, soit 12,48% de son salaire.  

Ceci correspond à une cotisation relativement élevée, illustrant la générosité du régime à prestations déterminées, dans la mesure où le facteur de 9 est raisonnable (taux de rendement bas, peu d’inflation).

Il ne faut pas oublier non plus que le FE est uniquement un artifice fiscal. Le vrai coût du régime varie selon le tableau précédent. Par exemple, avec un rendement net de 2%, le coût d’un crédit de rente de 1 245,70 \$ accordé à un homme de 50 ans devient  

13,98 x 1 245,70 \$ = 17 414,89 \$, soit 20,5% d’un salaire de 85 000 \$.

Avec un rendement net de 4%, on obtient  
8,43 x 1 245,70 \$ = 10 501,25 \$, soit 12,4% d’un salaire de 85 000 \$.


### Espace créé dans un REER

Les participants à un régime de retraite ne peuvent pas cotiser à leur REER le plein montant de leur limite REER car ils bénéficient déjà d’une valeur gagnée dans leur régime de retraite. L’espace REER ainsi créé, et qui représente le maximum cotisable au REER relativement à une année est donc égal à

MIN (18% x Revenu de l’année précédente; plafond REER) – FE de l’année précédente



### 4.2.5 Limite maximale sur le crédit de rente annuel offert par un régime à prestations déterminées

En plus de limiter l'aide fiscale à l'épargne retraite en termes de cotisations ou d'équivalent de cotisation, la loi de l'impôt limite aussi le crédit de rente annuel qui peut être gagné dans un régime à prestations déterminées. À noter que la limite est la même pour un régime à salaire de carrière que pour un régime à salaire final.

| ... | ... | ... | ... | ... |

| 2022 | 3 420,00 \$ | 171 000 \$ | 30 180 \$ | 29 210 \$ |


On peut tirer certaines observations du tableau précédent :

- Avant la réforme des pensions, les régimes à prestations déterminées étaient nettement avantagés car ils permettaient d’accumuler annuellement une valeur pouvant atteindre près de 15 000 \\$ alors que les régimes à cotisations déterminées et les REER étaient limités à une valeur annuelle de 5 500 \\$.
- Depuis 2004, le FE correspondant au crédit de rente maximum permis est très voisin de la limite de cotisation applicable aux REER et aux régimes à cotisations déterminées. C’est logique si on veut traiter les REER de façon cohérente avec les régimes de retraite.



### Particularités

Le plafond présenté au tableau précédent s’applique à une rente débutant à la première des dates suivantes :

- L’âge de 60 ans
- L’atteinte de 30 années de service
- Le moment où Âge + Service = 80

Si la retraite survient avant la première de ces dates, on doit réduire le plafond de ¼ % par mois d’anticipation de la retraite, sur base permanente.

#### Exemple :
Le régime de la compagnie ABC prévoit une rente de retraite de 2% du salaire final (moyenne des 3 dernières années) par année de service, payable à 65 ans. Le régime prévoit une réduction de 4% par année d’anticipation de la rente avant l’âge normal de retraite de 65 ans.

Un employé dont le salaire final moyen (3 dernières années) est de 160 000 \$ prend sa retraite en 2022 après 20 ans de service. Quelle sera la rente viagère maximale qu’il pourra recevoir s’il prend sa retraite à :
a) 65 ans
b) 57 ans

##### a) Rente à 65 ans :
Comme l’âge est supérieur à 60 ans, le plafond n’est pas réduit.  On utilise alors un plafond de 3 420,00 \\$ / année.
La rente normale est de 20 ans x 2% x 160 000 \\$ = 64 000 \\$
Le plafond est de 3 420 \\$ x 20 ans = 68 400 \\$
Donc sa rente sera égale à 64 000 \\$ / année car ce montant est en-dessous du plafond.


##### b) Rente à 57 ans :
Comme l’âge est inférieur à 60 ans, on doit calculer l’âge minimal pour lequel le plafond ne serait pas réduit :  
Min [ A, B, C ] où  
A = 60 ans  
B = années de services manquantes pour avoir 30 années + âge = (30 – 20) + 57 = 67  
C = 57 (= âge actuel) +[ 80 – ( 57+20) ] / 2 = 58,5  
on anticipe de 1,5 ans  
réduction du plafond = 1,5 x 12 mois x ¼% = 4,5%
Plafond réduit = 68 400 \\$ x (1-4,5%) = 65 322 \\$
Rente prévue par le régime : 64 000 \\$ x [ 1 – (65-57) x 4% ]  
= 64 000 \\$ x 68% = 43 520 \\$
Rente versée = Min (65 322 \\$; 43 520 \\$) = 43 520 \\$

### 4.2.6 Droits de cotisation inutilisés

Depuis la réforme des pensions, les plafonds de cotisation au REER sont tellement élevés que de nombreux canadiens ne versent pas dans leur REER la pleine cotisation qu’ils ont le droit de verser.



- Les contribuables ont la possibilité de **reporter indéfiniment** toute cotisation inutilisée à leur REER pour l'utiliser plus tard.
- Les **droits de cotisation non utilisés** à un REER sont indiqués sur l'**avis de cotisation** de l'Agence du revenu du Canada suite à la déclaration d'impôt.


### 4.2.7 Facteur d’équivalence rectifié (FER)

- Le FER redonne de l’espace REER à un participant quittant un régime à prestations déterminées si le facteur d’équivalence a été trop élevé.
- Il vise à corriger l’effet d’un calcul standardisé (facteur 9) souvent défavorable lors du transfert des droits à un nouveau régime ou REER.


**FER = Somme cumulative des FE imputés au participant – Valeur de transfert**

Évidemment, si le participant qui quitte le régime reçoit une rente différée, il n’a pas droit au FER car il n’y a pas d’équivalent monétaire (valeur de transfert) à comparer au FE.


### 4.2.8 Facteur d’équivalence pour services passés (FESP)

- Le FESP réduit l’espace REER lorsqu’un régime bonifie rétroactivement les droits à pension pour des services après 1989.
- Il reflète les cotisations fictives qui auraient été comptabilisées si ces améliorations avaient existé à l’époque.


Le FESP se calcule de la façon suivante :

**FESP = Somme des FE après amélioration – Somme des FE avant amélioration**

- Les **FESP (Facteurs d'Équivalence pour Service Passé)** se divisent en deux catégories :
  - Ceux nécessitant **une déclaration simple à l'ARC**, typiquement pour des bonifications globales ne dépassant pas **3 x la limite de cotisation** d'un régime à cotisations déterminées, avec moins de 25% de hauts salariés.
  - Ceux requérant **une certification préalable de l'ARC**, généralement pour des bonifications ciblées ou dépassant **3 x la limite de cotisation** d'un régime à cotisations déterminées. Si le FESP excède **(8 000 \$ + marge REER inutilisée)**, un retrait imposable du REER peut être nécessaire pour accommoder le FESP.


### 4.2.9 Traitement des prestations de raccordement

- Pour les rentes de **régimes à prestations déterminées débutant avant 65 ans**, une **prestation de raccordement** peut compenser l'absence de la pension de sécurité de la vieillesse et de la pleine rente du RRQ.
- La **prestation de raccordement maximale** est fixée par la Loi de l'impôt sur le revenu à la somme potentielle de la pension de sécurité de la vieillesse et de la rente du RRQ à 65 ans.
- Ce maximum est **réduit de 3%** pour chaque année de retraite anticipée avant 60 ans.
- En cas de retraite avant 65 ans, une **limite globale** s'applique à la rente de retraite et à la prestation de raccordement, en se basant sur un prorata du MGA et les années de service.



### 4.2.10 Valeurs de transfert

- Lors d'un départ d'un **régime à prestations déterminées**, le transfert de la valeur de la rente vers un **CRI** est limité par la Loi à un **multiple de la rente accumulée**, variant de **9,0 avant 50 ans** à **11,3 à 71 ans**.
- Le **calcul des valeurs de transfert** est une tâche importante pour les **jeunes actuaires** dans le domaine des régimes de retraite.





### 4.2.11 Les FERR (Fonds enregistré de revenu de retraite) et les FRV (Fonds de revenu viager)
- À la retraite, un **REER doit être converti en rente ou en FERR** avant la fin de l'année des **71 ans** du titulaire.
- Bien qu'un retrait complet soit possible à tout moment, cela augmente le **revenu imposable**; le **FERR et la rente** aident à répartir l'impact fiscal sur plusieurs années.
- La loi définit un **pourcentage minimum de retrait annuel du FERR**, sans maximum, sachant que des retraits plus élevés entraînent plus d'impôts.
- Ce pourcentage **augmente avec l'âge** jusqu'à environ 90 ans. Un **Fonds de revenu viager** est un type de FERR pour les transferts de régimes de retraite.
