## 3.1 Notion de risque

### 3.1.1 Définition générale d’un risque

Le Dictionnaire Larousse ([Larousse](https://www.larousse.fr/dictionnaires/francais/risque/69557)) donne 4 définitions d’un risque :

- Possibilité, probabilité d'un fait, d'un événement considéré comme un mal ou un dommage.

- Danger, inconvénient plus ou moins probable auquel on est exposé.

- Fait de s'engager dans une action qui pourrait apporter un avantage, mais qui comporte l'éventualité d'un danger.

- Préjudice, sinistre éventuel que les compagnies d'assurance garantissent moyennant le paiement d'une prime.


Ces quatre définitions comprennent des éléments communs :

- Un aspect défavorable si le risque se réalise.

- Un élément d’incertitude.

### 3.1.2 Autres définitions d’un risque

On retrouve partout des définitions relativement semblables :

#### Selon Wikipédia

Le **risque** est la possibilité de survenue d'un événement indésirable, la probabilité d’occurrence d'un péril probable ou d'un [aléa (la tournure imprévisible que peut prendre un événement)](https://fr.wikipedia.org/wiki/Al%C3%A9a). 

En statistiques descriptives, il est défini comme la probabilité d'exposition à un danger, à un événement (maladie, décès, accident) pendant un intervalle de temps défini. 

En [gestion des risques](https://fr.wikipedia.org/wiki/Gestion_des_risques), il est l’association de quatre facteurs : un [danger](https://fr.wikipedia.org/wiki/Danger), une [probabilité](https://fr.wikipedia.org/wiki/Probabilit%C3%A9) d'occurrence, sa gravité et de son acceptabilité.

#### Selon le jargon en Assurance

Le risque correspond à une exposition à un danger, un préjudice ou autre événement dommageable, inhérent à une situation ou une activité. Il est caractérisé par :

- Une probabilité de survenue de cet événement (différente de 0% et de 100%). Exemple : avoir un 1 lors du jet d’un dé est de 1/6.

- L’ampleur de ses conséquences (perte).


Il peut être appliqué à une personne, une population, des biens, l'environnement ou le milieu naturel.

Il se distingue d’une incertitude à laquelle ne peut être assignée une probabilité, tel qu’exprimé par [Frank Knight](http://fr.wikipedia.org/wiki/Frank_Knight) en 1921, comme par exemple l’intensité du réchauffement de la planète ou les effets des organismes génétiquement modifiés.

Il s’agit de la possibilité que quelque chose de néfaste se produise quelque part dans l’avenir, une situation qui pourrait être dangereuse ou avoir des conséquences défavorables. [Oxford Learner's Dictionaries](https://www.oxfordlearnersdictionaries.com/definition/american_english/risk_1)

#### Selon l’Institut national de santé publique du Québec

Un risque est la combinaison de conséquences négatives sur la santé de la population et de la probabilité d’observer ces conséquences à la suite d’une exposition à un agent dangereux. [Institut national de santé publique du Québec](https://www.inspq.qc.ca/evaluation-et-gestion-des-risques/la-gestion-des-risques-en-sante-publique-au-quebec-cadre-de-reference/les-concepts-cles-de-la-gestion-des-risques-la-sante/les-caracteristiques-du-risque).

Les mêmes éléments précités reviennent :

- Un aspect défavorable si le risque se réalise.

- Un élément d’incert

### 3.1.3 Définition d’un risque en contexte d’assurance et d’actuariat


En actuariat, un risque est essentiellement un événement dont la
survenance est aléatoire (due au hasard), dont la probabilité est
estimable et qui cause une perte qu’on peut chiffrer monétairement.

Par exemple :


- Le risque de tornade :

- Une tornade est un événement dévastateur et imprévisible, sauf
  quand on l’aperçoit. L’apparition d’une tornade à tel ou tel
  endroit est essentiellement due au hasard, tout comme sa
  puissance et les dégâts qu’elle provoque. Mais on a des
  statistiques générales sur la fréquence annuelle des tornades au
  Québec ainsi que sur les dégâts qu’elles provoquent.


- Le risque qu’une personne soit incapable de travailler en raison d’un accident ou d’une maladie :

- Ce risque comporte plusieurs sous-risques : le risque d’accident
  invalidant, le risque de maladie invalidante, le risque que
  l’invalidité dure 1 mois, 3 mois, 2 ans, 5 ans, etc. Parfois,
  lorsque la finalité est unique (par exemple tarifer le risque
  d’invalidité), on va regrouper certains sous-risques ensemble.
  On a des statistiques sur la fréquence et la durée des
  invalidités, et on peut en évaluer le coût en tenant compte des
  salaires perdus pendant l’invalidité.

Les caractéristiques suivantes, incluses dans la définition d’un risque
en contexte d’assurance, sont essentielles :

- **Survenance aléatoire :**

- La survenance de l’événement défavorable est due au seul hasard. 
  C’est-à-dire que même si un événement est certain (comme le décès : 
  on va tous mourir un jour), le moment où cet événement survient 
  comporte un fort élément de hasard. Par exemple, même si un vieillard 
  de 90 ans a beaucoup plus de chances de décéder qu’une femme âgée 
  de 20 ans, il n’est pas certain que le vieillard va mourir cette année 
  ou même qu’il va mourir avant la jeune femme (un accident est vite arrivé…).
- Évidemment, dans le cas d’un événement incertain à la fois quant
  à sa survenance et au moment de cette survenance (accident mortel, 
  par exemple), le caractère aléatoire est encore plus évident.
- Un événement qui n’a pas de caractère aléatoire ne constitue pas
  un risque assurable. Par exemple, si on met le feu à sa propre maison, 
  l’incendie n’est pas assurable car pas dû au hasard ou à la malchance. 
  Par conséquent, la réalisation du risque ne doit pas découler d’un acte 
  volontaire de l’assuré. Même chose dans le cas d’un suicide planifié, 
  qui n’est pas la conséquence d’une maladie mentale.
- Évidemment, le risque ne doit pouvoir se matérialiser que dans l’avenir.


- **Probabilité estimable :**

- Une probabilité de survenance du risque est essentielle à l’estimation 
  de son coût.
- On peut définir le coût théorique d’un risque comme le risque d’incendie :
  - **Coût** = Probabilité x Perte potentielle (en \$),
    - *Pour un risque de décès*:
      Coût = Probabilité X Montant d’assurance vie (en \$)
    - *Pour un risque récurrent, comme la maladie*:
      Coût = Fréquence X Sévérité (en \$)
- La fréquence de réalisation d’un risque récurrent est directement
  reliée à la probabilité de réalisation de ce risque.
- Si on n’a aucune idée de la probabilité de réalisation d’un risque,
  on ne peut appliquer aucune des formules ci-dessus et le risque
  n’est pas tarifable objectivement.
- Évidemment, la probabilité de réalisation du risque doit être
  substantiellement moindre que 1,00 sinon on n’aurait plus de risque
  mais une certitude ou une quasi-certitude.

- **Perte pouvant être chiffrée monétairement :**

- Le deuxième terme de chacune des formules ci-dessus est le
  montant d’argent associé à la perte.
- Si on n’a aucune idée de la perte résultant de la réalisation
  d’un risque, on ne peut appliquer aucune des formules ci-dessus
  et le risque n’est pas tarifable objectivement.
- De façon pratique il est rare qu’on ait la situation simple
  d’une assurance vie d’une durée d’un an, dont le coût
  actuariel = Prob (décès de la personne) X montant d’assurance  
  On voit plus souvent une situation comme en assurance
  automobile, où le coût ressemble plutôt à  
  **Coût $=\sum$ [ Prob (Perte = z \\$) x Z \\$ ]**  
  car la perte découlant d’un accident est plus ou moins élevée
  (pare-brise à remplacer vs ailes, pare-chocs, pare-brise et
  capot à remplacer vs perte totale du véhicule).  
  Souvent, on va simplifier cette formule comme ceci :  
  **Coût = Prob (subir une perte) X perte moyenne (\$)**  
  lorsque la formule de la sommation des produits des
  probabilités par les montants est impossible à calculer, car
  souvent les statistiques dont on dispose incluent seulement une
  fréquence globale et une distribution des montants de perte.

- **Importance réelle :**

- La valeur monétaire de la perte qu’on voudra assurer doit être
  assez importante pour justifier les efforts et les coûts
  administratifs reliés à l’obtention de l’assurance.
- Le bien assuré ou la personne assurée doit exister réellement
- Le préjudice monétaire causé par la réalisation du risque doit
  avoir une certaine importance (on n’assurera pas un crayon, par exemple)


- **Non catastrophique pour l’assureur :**

- La réalisation du risque ne doit pas mettre en péril la capacité
  de l’assureur à honorer tous ses engagements envers tous ses
  assurés. L’assureur doit d’ailleurs toujours garder le capital
  minimum requis par la Loi.


- **Aspect licite :**

- L’élément assuré ne doit pas être contraire à la Loi. Par
  exemple, un producteur de cannabis agréé peut assurer sa récolte
  contre l’incendie ou le vol, mais un producteur artisanal ne
  peut pas assurer ses 100 plants de cannabis si la loi ne lui
  permet pas d’en posséder plus de 3.


## 3.2 Aversion pour le risque

### 3.2.1 Définition générale


L’aversion pour les risques est une attitude qui s’exprime par le fait qu’on est prêt à payer plus que le coût « actuariel » d’un risque pour s’en débarrasser.



Par « ***coût actuariel*** », on entend le coût (espérance) mathématique, soit la probabilité de survenance du risque multipliée par la perte moyenne, donc :

***Coût «actuariel» = Probabilité x Perte potentielle (en \\$),***

Le coût actuariel suppose qu’il n’y a aucun frais externe associé au risque.

Par exemple, une personne peut se débarrasser du risque d’incendie et de perte de sa maison qui vaut 500 000\\$ en achetant une assurance contre l’incendie. Si l’espérance mathématique du risque d’incendie est de 250\\$ (soit 1 chance sur 2 000 de subir une perte de 500 000\\$) le coût actuariel est de 250\\$, comme on l’a vu au Module 1.

Si la personne ne fait rien, deux situations peuvent se produire :

- La maison ne brûle pas et elle ne perd rien;

- La maison brûle et elle vient de s’appauvrir de 500 000\\$


Si un assureur se présente et lui offre d’assurer la maison pour 500 000\\$, moyennant une contrepartie (prime) de 400\\$, elle a deux choix :


- Perdre 400\\$ de façon certaine

- Courir le risque de perdre 500 000\\$

Sachant que le coût actuariel de cette assurance est de 250\\$, et que le seul assureur disposé à assurer la maison demande 400\\$ la personne va fort probablement choisir l’assurance car la perte certaine de 400\\$ est minime par rapport à la perte potentielle de 500 000\\$.

Si on est prêt à payer plus que le coût actuariel d’un risque afin de le transmettre à quelqu’un d’autre, on a de ***l’aversion pour le risque***.

L’aversion pour le risque est renforcée par l’asymétrie entre la perte maximale et le coût pour se débarrasser d’un risque. Ainsi, on est prêt à payer 400\\$ pour se débarrasser d’un risque qui peut causer une perte de 500 000\\$ avec une probabilité de 1/2000 mais on est moins disposé à payer 12 000\\$ pour se défaire d’un risque qui peut causer une perte de 20 000\\$ ou même une perte de 24 000\\$ (équivalent actuariel) avec une probabilité de $1\2$.

La distribution Normale est une distribution symétrique :

![Distribution Normale](media/image2.png)

Le problème de la distribution Normale, c’est qu’elle est bornée à gauche à $-\infty$, ce qui est incohérent avec une perte assurable, qui ne peut logiquement pas être négative.

Avec un risque distribué normalement, la moyenne (correspondant à la prime pure, donc non augmentée des frais) est au milieu de la courbe. Comme la Normale est, en réalité, bornée à $-\infty$, on ne peut considérer la courbe comme étant bornée à zéro à gauche que si $( P(x < 0) \rightarrow 0 )$, ce qui implique que $( P (X > 2\mu) \rightarrow 0 )$ aussi, et donc $( \mu )$ égale la moitié de la plus haute perte plausible. 

Dans un tel cas, la perte maximale qu’on peut protéger en s’assurant ne vaut pas la peine de payer la prime, surtout si on considère le risque une année à la fois car au bout de 2 ans on aurait payé une prime cumulative égale à la perte maximale plausible.

Si le risque est distribué de façon très asymétrique et borné à zéro à gauche, il devient intéressant de payer une petite prime pour se débarrasser d’un risque peu probable mais coûteux. Dans l’exemple ci-dessous, la moyenne (prime pure) est de 1,0 mais la probabilité que la perte dépasse 10, et même 20 ou encore plus (en extrapolant) est non négligeable et l’assurance devient intéressante.

### 3.2.2 Conséquences de l’aversion pour le risque

L’aversion pour le risque a pour *conséquence première* le fait qu’on cherchera très fortement à se débarrasser d’un risque qu’on n’est pas capable de supporter confortablement, comme le risque d’incendie d’une maison ou le risque de décès prématuré du gagne-pain de la famille.

La *deuxième conséquence* de l’aversion pour le risque est de rendre possible l’assurance. En effet, dans la vraie vie, un assureur ne peut pas accepter un risque à sa valeur actuarielle pour les raisons suivantes :

- S’il prend tous ses risques à leur valeur actuarielle et qu’une fluctuation statistique défavorable survient, les risques coûteront alors plus cher que leur valeur actuarielle;

- L’assureur doit inclure ses frais administratifs dans la prime qu’il demande pour assurer un risque;

- L’assureur doit inclure ses coûts de distribution (commissions à l’intermédiaire qui vend l’assurance au client, etc.)

- L’assureur doit finalement payer des taxes sur les primes qu’il perçoit.

## 3.3 Neutralité face au risque


Comme on l’a vu plus haut, l’aversion pour le risque accompagne une asymétrie entre la perte potentielle et le coût pour se défaire du risque. Quand on n’a pas cette asymétrie, on sera neutre face au risque (comme dans l’exemple d’une perte potentielle de 20 000 \\$ ou 24 000 \\$ en contrepartie d’un coût de 12 000 \\$ avec probabilité de 50%).


## 3.4 La tolérance face au risque

Lorsque la perte potentielle est relativement petite, on peut vouloir conserver un risque plutôt que le transférer à quelqu’un d’autre. On a alors une tolérance face à ce risque.

Par exemple, si la calculatrice d’une personne vaut 50\\$ et que la probabilité de se la faire voler est égale à 1/50, le coût actuariel du risque est de 1,00 \\$. Si un assureur n’avait aucun frais ni taxe à payer, il accepterait peut-être d’assurer cette calculatrice moyennant une prime de 1\\$. Est-ce que le propriétaire de la calculatrice accepterait ? Probablement pas car, en cas de perte de la calculatrice, c’est plus simple d’en racheter une que de faire une réclamation à l’assureur (et courir un autre risque, soit celui que l’assureur demande désormais 2,00 \\$ pour assurer la calculatrice.

Évidemment, avec ses frais administratifs, il est improbable qu’un assureur demande moins que 25\\$ ou 40\\$ pour ’assurer la calculatrice. Dans ce cas, la question ne se pose quasiment plus : cela ne vaut pas la peine de payer 25\\$ pour assurer une calculatrice qui vaut 50\\$ et a une probabilité de 1/50 de se faire voler.

Dans ces deux exemples, le propriétaire de la calculatrice aura de la tolérance pour le risque de se faire voler sa calculatrice.

## 3.5 L’appétit pour le risque

Pour qu’on puisse transférer un risque à un assureur, moyennant une prime d’assurance, il faut que l’assureur y trouve son compte. Même si la prime est le double du coût actuariel, la perte potentielle reste démesurée par rapport à la prime. Alors comment l’assureur peut-il accepter ce risque ?

Grâce à la mise en commun des risques, expliquée un peu plus loin, un assureur peut combiner des risques en nombre suffisant pour que sa perte totale ne dépasse pas son coût actuariel total de plus qu’un certain pourcentage, avec un certain niveau de probabilité. Par exemple, avec 160 risques assurés chacun pour 100 000 \\$ et ayant chacun une probabilité de survenance de 5%, on s’attend à 8 sinistres. Or, la combinaison des risques a pour effet que la probabilité d’avoir 8 sinistres ou moins est de 59% et que la probabilité d’avoir 15 sinistres ou moins est de 99,33%.

Si l’assureur demande une prime égale au double du coût actuariel, plus ses frais, il est virtuellement assuré de réaliser un bon profit. Les calculs sont expliqués en détails à la section 3.7. C’est principalement la mise en commun des risques qui permet à l’assureur d’avoir de l’appétit pour le risque.

D’autres éléments, tels la capacité financière de l’assureur et la possibilité pour l’assureur de transférer des risques à un réassureur aident aussi l’assureur à avoir de l’appétit pour le risque.

## 3.6 Les différentes attitudes face au risque

### 3.6.1 Évitement du risque

On peut rendre certains risques irréalisables, comme par exemple éviter le risque d’inondation de la maison à la suite d’un débordement de rivière en construisant la maison sur une hauteur. On peut aussi éviter le risque d’incendie en construisant une maison en béton et en n’ayant rien d’inflammable à l’intérieur.

Évidemment, cette stratégie est plus théorique que pratique et elle est totalement inapplicable face à certains risques qui comportent un élément de certitude comme le décès.

### 3.6.2 Prévention


On peut essayer d’empêcher le risque de survenir. Par exemple, une personne qui voyage dans des pays où l’hépatite est endémique a avantage à se faire vacciner contre cette maladie, qui peut avoir des effets très graves. De même en voyant un médecin de famille régulièrement et en faisant attention à notre santé, on réduit la probabilité de décès.


### 3.6.3 Mitigation


On peut réduire la gravité d’un risque ou le montant de la perte. Par exemple, en installant des gicleurs dans un immeuble, on limite considérablement les dégâts en cas d’incendie. En assurance de personnes, il est plus difficile de trouver un moyen de réduire le coût de la réalisation d’un risque, mais toute mesure permettant de réduire la durée d’une éventuelle invalidité et la perte de salaire qui en résulte constitue une forme de mitigation.


### 3.6.4 Rétention

On peut assumer complètement ou partiellement un risque, soit parce qu’on ne peut pas éviter le risque ni le mitiger ni le transférer à quelqu’un d’autre ou si on juge, après analyse, que le coût du transfert de risque serait disproportionné.

Par exemple, un homme de 50 ans considéré difficilement assurable par un assureur ne peut obtenir qu’une assurance vie de 50 000\\$ qui va se terminer à 65 ans, au coût de 500\\$ / année. Il considère qu’il n’a pas plus de 5% de chances de mourir avant 65 ans. Il prend plutôt 50 000\\$ qu’il investit au nom de sa conjointe dans des actions d’une grande banque canadienne (BMO) qui prennent 5% de valeur par année, en plus de verser un dividende de 4%. S’il survit jusqu’à 65 ans, sa conjointe aura reçu 30 000\\$ en dividendes et ses actions vaudront alors 103 946\\$, soit un profit de 33 946\\$ (soit 133 946\\$ - 50 000\\$ de prestations de décès – 50 000\\$ investis initialement). Si on ajoute les 7 500\\$ de primes économisées, on se retrouve avec un profit net de 41 446\\$.

Si cet homme décède à 60 ans, sa conjointe aura reçu 20 000\\$ de dividendes + 81 445\\$ de capital + 5 000\\$ de primes non utilisées moins 50 000\\$ de prestation de décès non reçue car pas d’assurance, moins les 50 000\\$ investis initialement, soit un profit de 6 445\\$.

### 3.6.4 Diversification des risques

On peut diversifier les risques auxquels on fait face, principalement en matière de placements, (au lieu d’investir tout ce qu’une personne possède en actions d’une seule grande banque), on peut diversifier les placements en achetant des actions de compagnies très différentes les unes des autres, ce qui produit un portefeuille diversifié. C’est moins évident avec des risques assurables, toutefois…

Les assureurs, en acceptant beaucoup de risques différents, qui sont peu ou pas corrélés, donc des événements indépendants, se trouvent à effectuer une diversification des risques.

### 3.6.5 Transfert du risque


On peut transférer partiellement ou totalement une partie du risque à une tierce partie (gouvernement ou compagnie d’assurance)

L’aversion pour le risque fait en sorte que la personne préfère payer une petite prime afin de ne pas avoir à subir les conséquences monétaires d’assumer un gros risque.

La prime est alors le coût que la personne est prête à payer pour transférer le risque à une compagnie d’assurance.

La personne préfère une petite perte certaine (la prime) pour ne pas assumer la perte financière due à l’évènement.

### 3.7 La mise en commun des risques

Quand on combine plusieurs risques indépendants, la probabilité que tous ces risques se réalisent la même année diminue au fur et à mesure qu’on augmente le nombre de risques. En fait, il est extrêmement probable que seuls certains risques se réaliseront.

Le nombre de combinaisons possibles de $N$ parmi $Z$ se calcule facilement avec la formule d’analyse combinatoire $\frac{\mathrm{Z} !}{\mathrm{N} ! \mathrm{x}(\mathrm{Z}-\mathrm{N}) !}$

$$\frac{\mathrm{Z} !}{\mathrm{N} ! \mathrm{x}(\mathrm{Z}-\mathrm{N}) !}$$

```python
import pandas as pd
import numpy as np
from scipy.stats import binom
from scipy.special import comb

# Define the parameters
probability_of_success = 0.05  # Probability of a single success
number_of_trials = 1000        # Number of trials (number of insured in your case)

df = pd.DataFrame({'Nb de sinistres': range(number_of_trials + 1)})

df['Probabilité dune combinaison'] = binom.pmf(df['Nb de sinistres'], number_of_trials, probability_of_success)

df['Nb de combinaisons possibles'] = comb(number_of_trials, df['Nb de sinistres'], exact=False)

df['Probabilité totale'] = df['Probabilité dune combinaison'] * df['Nb de combinaisons possibles']

df['Probabilité cumulative'] = binom.cdf(df['Nb de sinistres'], number_of_trials, probability_of_success)

df['Probabilité dune combinaison'] = df['Probabilité dune combinaison'].round(15)
df['Probabilité totale'] = df['Probabilité totale'].round(15)
df['Probabilité cumulative'] = df['Probabilité cumulative'].round(15)

# Save to CSV
df.to_csv('binomial_distribution.csv', index=False)
```

Ceci nous amène à un aspect fondamental de l’actuariat, de plus en plus exigé par les autorités de réglementation et par les penseurs de la profession : mesurer la suffisance des ressources des assureurs en termes d’écarts et de dispersion. Autrement dit, ne plus calculer une réserve actuarielle en se servant de probabilités moyennes (comme des taux de mortalité) mais plutôt en utilisant une série de calculs stochastiques, afin d’établir la réserve non pas comme étant le montant moyen à détenir mais plutôt le montant qui permet à l’assureur d’être certain à x% de pouvoir honorer ses engagements.

Jusqu’à maintenant, nous avons supposé que l’assureur débutait son année en ne possédant rien d’autres que les primes qu’il doit recevoir. Dans la réalité, l’assureur doit posséder un capital minimal pour avoir le droit d’être en affaires

La présence d’un tel capital permet à l’assureur de réduire ses primes pour les situer entre le niveau « sécuritaire » et le niveau du coût actuariel.

Ainsi, supposons maintenant que l’assureur débute son année avec un capital de 20 000\\$. Avec 40 assurés, il a besoin de 60 000\\$ au total pour être certain à 99% de tenir ses engagements. Comme il possède déjà 20 000\\$, il lui suffit de collecter 40 000\\$, soit 1 000\\$ / assuré, pour satisfaire ce critère de 99%. La présence d’un capital de départ permet ainsi de ramener la prime à un niveau plus réaliste et acceptable pour l’assuré.

De la même manière, avec 160 assurés et un capital de départ de 20 000\\$, l’assureur, qui a besoin de 150 000\\$ au total, pourrait se contenter de collecter 130 000\\$ en primes, soit 130 000 / 160 = 812,50\\$ au lieu de 937,50\\$. Dans cette situation, il a encore une probabilité de 94,11% de subir 12 sinistres ou moins, faisant du même coup un profit d’au moins 10 000\\$.

### 3.7 Les différentes classifications d’assurance


Il y a plusieurs façons de classer l’assurance :


- Par rapport au preneur de risque

    - on a l’auto-assurance, l’assurance privée et l’assurance sociale.


- Par rapport à l’objet de l’assurance (personne ou chose)

    - on a l’assurance de personnes et l’assurance IARD.


- Par rapport à l’approche de la couverture

    - on a l’assurance individuelle et l’assurance collective.


- Par rapport à l’assuré

    - on a l’assurance personnelle et l’assurance commerciale.


- Par rapport à la dette d’un assuré

    - on a l’assurance-crédit.


Évidemment, ces classifications ne sont pas mutuellement exclusives l’une par rapport à l’autre. Ainsi, on peut avoir une assurance de personne privée et individuelle ou une assurance de personne collective et sociale.


### 3.7.1 L’auto-assurance


Il s’agit essentiellement de la rétention d’un risque qu’on ne veut pas ou qu’on ne peut pas éviter ni transférer. On conserve donc le risque en totalité ou en partie.

Quelques exemples :

- La franchise en assurance médicaments

- Le fait de ne pas s’assurer contre les dommages à son propre véhicule résultant d’une collision dont on est responsable.

### 3.7.2 L’assurance privée


Il s’agit de l’assurance offerte par les assureurs privés.

- Elle est basée sur l’équité actuarielle **stricte**, c’est-à-dire que chacun doit payer une prime proportionnelle à son risque.

- Elle est parfois obligatoire

- Au Québec, l’assurance médicaments est obligatoire quand on a accès à un régime collectif (Loi sur l’Assurance médicaments du Québec).

- Au Québec, l’assurance automobile couvrant la responsabilité civile du propriétaire du véhicule est obligatoire.

### 3.7.3 L’assurance sociale<sup>e</sup>


Il s’agit de l’assurance offerte par le gouvernement d’un pays ou d’une région (État ou province).

- Elle est normalement obligatoire

- Elle est fondée sur des principes d’assurance, mais élargis de façon à inclure des personnes dont le risque est trop élevé pour qu’elles aient accès à une couverture abordable auprès d’un assureur

- Par exemple, au Canada, tous les résidents permanents ont droit à l’assurance maladie même si leur état de santé ne leur permettrait pas d’avoir accès à l’assurance dans un libre marché comme celui des États-Unis

- Elle est basée sur l’équité actuarielle **élargie**, c’est-à-dire que chacun doit payer une prime proportionnelle au risque moyen de l’ensemble des assurés

- Elle est habituellement financée par des primes ou des cotisations établies par la Loi

- Elle vise à protéger la population

- Voici quelques exemples :

  - Rentes de retraite, de survivant et d’invalide du régime de rentes du Québec
  - Assurance maladie du Québec
  - Assurance emploi
  - Assurance parentale du Québec
  - Assurance en cas d’accident du travail ou maladie professionnelle gérée par la CNESST.


### 3.7.4 L’assurance de personnes


Il s’agit de l’assurance qui couvre les risques de décès et de maladie ou invalidité d’une personne.

- L’assurance vie couvre le risque de décès et ses conséquences financières néfastes.

- L’assurance invalidité verse une rente si la personne n’est plus capable de travailler (risque de morbidité).

- L’assurance maladie rembourse les frais médicaux ou paramédicaux causés par l’état de santé d’une personne (risque de morbidité).

- L’assurance contre les maladies graves verse un montant d’argent si la personne est atteinte d’une maladie énumérée au contrat (risque de morbidité).

* Les rentes protègent une personne contre le risque de survivre à ses épargnes (risque de longévité).

### 3.7.5 L’assurance IARD


Il s'agit de l'assurance qui couvre les pertes reliées à la propriété et à la responsabilité civile.

#### Assurance des biens

- Couvre incendie, vol, destruction, vandalisme, etc.


#### Assurance de la responsabilité civile

- Couvre les conséquences de notre négligence.


#### Assurance automobile

- Couvre blessures corporelles, responsabilité civile, dommages résultant d’une collision, feu, vol, vandalisme, bris de vitre.


#### Assurances diverses

- Comprend l'assurance maritime, récoltes, etc.

### 3.7.6 L’assurance individuelle


Il s'agit de l'assurance qui couvre les pertes subies par un individu.


### 3.7.7 L’assurance collective

Il s'agit de l'assurance qui couvre les pertes subies par un membre d'un groupe légitime ou une de ses personnes à charge.

#### Par « groupe légitime »

- On entend un groupe formé pour une raison autre que l'obtention d'assurance.

- Au Québec, il s'agit habituellement d’employées d'un même employeur ou de personnes ayant une occupation rémunératrice commune. C'est la condition pour que le régime puisse offrir une assurance médicaments collective. On rencontre aussi des membres de clubs sociaux ou de sociétés d'entraide, mais il est illégal de leur fournir une assurance médicaments.


#### Les protections les plus courantes sont

- Assurance vie

- Assurance en cas de décès ou infirmité à la suite d'un accident

- Assurance invalidité (courte et longue durée)

- Assurance médicaments et soins de santé

- Assurance soins dentaires

### 3.7.8 L’assurance personnelle


Il s'agit de l'assurance qui couvre les pertes subies par un individu dans le cadre de sa vie privée (décès, vol, etc.).


### 3.7.9 L’assurance commerciale

Il s'agit de l'assurance qui couvre les pertes subies par une entreprise (incendie, responsabilité, etc.).

#### Principalement de l’a|ssurance IARD

- Inclut aussi l'assurance sur la vie des actionnaires pour financer le rachat des parts au décès d'un actionnaire important, dans les petites entreprises.

### 3.7.10 L’assurance crédit


Il s'agit d'une assurance vendue par les institutions financières et reliées aux emprunts et dettes de crédit des particuliers.

- Elle paye habituellement le solde de la dette en cas de décès et le versement mensuel minimum exigé par le prêteur si l’assuré devient invalide.
