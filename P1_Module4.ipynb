### 4.1.2 Le Conseil d’administration

Le Conseil d’administration est composé d’une majorité d’administrateurs externes, donc non employés de la société. Il comprend aussi le Président et le Directeur général, si ces deux postes sont distincts, de même que le Secrétaire corporatif.

### 4.1.3 Les grands responsables

Il y a habituellement 4 postes de la plus haute responsabilité :

- Chef de la Direction (souvent le Président du Conseil d’administration)

  « Chiel Executive Officer »  

- Chef des Opérations (souvent le Président)  

  « Chief Operating Officer »

- Chef des Finances  

  « Chief Financial Officer »

- Chef de la gestion des risques  

  « Chief Risk Officer »
Et de plus en plus souvent un 5ème :

- Chef des technologies de l’information  

  « Chief Information Technology Officer »


Ces dirigeants chapeautent toute la société d’assurance.

### 4.1.4 Les principales divisions

Quoique le modèle puisse varier d’une société à l’autre, les principales divisons d’une société d’assurance de personnes sont les suivantes :


#### 4.1.4.1 Le département de la comptabilité

Ce département est responsable de :

- La perception des primes.

- L’émission des chèques de prestations ou du dépôt direct des prestations.

- La vérification interne.

- Les relations avec les auditeurs / vérificateurs externes.

- La préparation des rapports financiers.

- La préparation des rapports d’impôt et le paiement des impôts et taxes.

#### 4.1.4.2 Le département de la finance et des investissements

Ce département est responsable de :

- L’application de la politique de placement.

- Le choix, la répartition et la gestion des placements.

- La gestion du parc immobilier de la Société (immeubles possédés et baux).

#### 4.1.4.3 Le département l’actuariat corporatif

Ce département est responsable de :

- L’évaluation des provisions actuarielles (sommes que doit détenir l’assureur et qui, additionnées aux primes futures, permettront à l’assureur de se décharger de toutes ses obligations d’assurance).

- La préparation du rapport de l’Actuaire désigné (portant sur l’évaluation des provisions actuarielles).

- L’examen de la santé financière ([L’examen de la santé financière (ESF)](http://www.cia-ica.ca/docs/default-source/2013/213077f.pdf?sfvrsn=2)) (L’examen de la santé financière (ESF) est un processus de simulation de crise qui s’inscrit dans la gestion globale des risques de l’assureur.)

- La préparation du rapport d’examen de la santé financière

- Les tests servant à l’évaluation interne des risques et de la solvabilité (ORSA – Ligne directrice E-19 du Bureau du Surintendant des institutions financières).

#### 4.1.4.4 Le département d’actuariat ‑ assurance individuelle

- L’établissement de normes de sélection et de tarification.

- La tarification des produits d’assurance individuelle (manuels de taux).

- Le calcul des participations à verser aux titulaires de polices.

- L’émission des contrats et des avenants.

#### 4.1.4.5 Le département d’actuariat ‑ assurance collective

- L’établissement de normes de sélection et de tarification.

- La tarification théorique des produits d’assurance collective (manuels de taux).

- La préparation des soumissions demandées par les courtiers et actuaires-conseils.

- La tarification des plus gros groupes selon leur propre expérience.

## 4.2 Les outils utilisés par une société d’assurance de personnes

### 4.2.1 Outils économiques

Les principaux outils économiques utilisés par les assureurs comprennent :

- Le rendement des portefeuilles de placement de l’assureur

- Certains produits sont liés à un portefeuille qui leur est dévolu (par exemple, tel ou tel groupe de polices similaires, comme les polices « vie-entière », ou encore les fonds distincts, etc.).

- Le calcul des provisions actuarielles des produits en vigueur est basé sur le rendement des placements affectés à ces produits, en vertu de la « Méthode canadienne axée sur le bilan ».

- Les analyses économiques et projections des taux de rendement futurs

- Quand on conçoit un nouveau produit ou qu’on retarifie un produit existant, on n’a pas encore de fonds accumulés pour les produits en question. On doit donc faire une hypothèse quant au rendement qu’on fera dans l’avenir sur les primes lorsqu’elles auront été perçues.

### 4.2.2 Outils actuariels

- Des tables de mortalité, qui varient selon leur usage :
  - Assurance vie individuelle avec taux variant selon la sélection des risques, fumeurs et non-fumeurs séparément
  - Assurance vie individuelle sans sélection de risques
  - Rentes individuelles
  - Rentes collectives
  - Assurance vie collective

- Des tables de morbidité, utilisées en assurance invalidité :
    - Table de taux d’incidence de l’invalidité selon l’âge, le sexe, le délai de carence et, dans le cas de l’assurance individuelle, la classe tarifaire (4A, 3A, 2A A et B)
  - Taux de terminaison d’invalidité selon l’âge, le sexe et la durée écoulée depuis le début de l’invalidité

- Des tables de taux d’assurance maladie et invalidité, qui varient selon leur usage et selon la provenance des données ayant servi à les bâtir :
  - Coût de l’assurance médicaments et maladie individuelle, selon âge, sexe et classe de risque
  - Coût de l’assurance invalidité individuelle selon âge, sexe, délai de carence, durée maximale des prestations et classe de risque
  - Coût de l’assurance des soins dentaires individuelle selon âge, sexe et classe de risque
  - Coût de l’assurance médicaments et maladie collective, selon âge et sexe
  - Coût de l’assurance invalidité collective selon âge, sexe, délai de carence et durée maximale des prestations
  - Coût de l’assurance des soins dentaires collective selon âge et sexe

## 4.3 Les différents produits individuels offerts par une société d’assurance de personnes

Cette section constitue un bref survol. Les produits individuels les plus intéressants sont vus en détails dans un module ultérieur.

### 4.3.1 Produits d’assurance

Les principaux produits individuels se répartissent entre l’assurance vie, l’assurance maladie, l’assurance invalidité et l’assurance des maladies graves

#### 4.3.1.1 Produits d’assurance vie

Les produits d’assurance vie comprennent :

- L’assurance vie entière à primes nivelées, qui offre habituellement un montant fixe d’assurance vie, provisionné par une prime constante d’une année à l’autre, et qui accumule un fonds qui sera versé au titulaire de la police s’il la fait racheter par l’assureur, advenant le cas où il n’en voudrait plus.

    - C’est le produit d’assurance le plus traditionnel qui soit.

- L’assurance vie temporaire, qui offre un montant d’assurance fixe mais pour une durée limitée, sans accumulation de fonds.
    
    - C’est aussi un produit d’assurance traditionnel.

- L’assurance temporaire à 100 ans, un hybride entre les deux assurances précédentes, qui offre un montant fixe d’assurance vie, provisionné par une prime constante d’une année à l’autre, mais plus basse que celle de l’assurance-vie entière et qu’on ne peut pas faire racheter par l’assureur si on n’en veut plus.

- L’assurance vie universelle, un mélange d’assurance temporaire, d’assurance vie entière et de fonds d’investissement.

#### 4.3.1.2 Produits d’assurance maladie

Les produits d’assurance maladie comprennent :

- L’assurance maladie qui rembourse le coût des soins non couverts par l’État (hospitalisation en chambre privée ou semi-privée, professionnels de la santé comme les physiothérapeutes et les chiropraticiens, les médicaments sur ordonnance (sauf au Québec, où il est illégal de couvrir les médicaments de la liste de la RAMQ au moyen d’une assurance individuelle) et parfois les soins dentaires.

- L’assurance des soins de longue durée, un produit difficile à vendre car les gens ont tendance à ne pas en percevoir le besoin tant que ce besoin ne s’’est pas manifesté, auquel cas il est trop tard pour acheter le produit….

#### 4.3.1.3 Produits d’assurance invalidité

- **Objectif** : Verse une rente (hebdomadaire ou mensuelle) si l'assuré devient invalide et ne peut plus travailler.
- **Délai de carence** : Prestations versées après un certain nombre de jours suivant le début de l'invalidité.
- **Durée des prestations** :
  - Tant que l'assuré est invalide.
  - Limitée à 2 ou 5 ans dans la plupart des cas.
  - Exception : Certains contrats couvrent jusqu'à 65 ans pour les très bons risques.



#### 4.3.1.4 Produits d’assurance contre les maladies graves

- **Définition** :
  - Contrairement à l’assurance maladie traditionnelle, elle ne rembourse pas les soins mais verse un **montant fixe** si une maladie couverte survient et que l’assuré survit au moins 30 jours.

- **Caractéristiques** :
  - Produit relativement récent.
  - Conçu pour offrir une protection financière individuelle en cas de maladie coûteuse.

- **Public cible** :
  - Travailleurs autonomes et employés sans régime d’assurance invalidité.
  - Personnes ayant besoin de soins privés ou non couverts.

- **Différence avec l’assurance invalidité** :
  - Se déclenche sur le **diagnostic d’une maladie spécifique**.
  - Évite les problèmes liés aux maladies subjectives (ex. : fatigue chronique, fibromyalgie, maladies nerveuses).

### 4.3.2 Produits de rente
Les produits de rente permettent de transformer un capital en une série de versements périodiques. 
- Utiles pour convertir un montant accumulé dans un REER en revenu régulier à vie.


#### 4.3.2.1 Rentes immédiates et rentes dues
- Paiement débutant immédiatement après le versement de la prime.
- **Rentes dues** : Paiement immédiat.
- **Rentes immédiates** : Paiement un mois après la prime.
- Durée fixe sans considération de la survie du rentier (mathématiques financières).
- Si la notion de survie est impliquée, elles relèvent des mathématiques actuarielles.

#### 4.3.2.2 Rentes différées
- Paiement débutant à une date future prédéfinie.
- Chaque versement de prime achète un montant de rente payable plus tard.
- Plus l’achat est anticipé, plus la rente accumule des intérêts.

### 4.3.3 Produits de placement ou d’accumulation

- **Investissements sans garantie de rendement.**
- **Avantages pour l’assureur** : Aucun risque financier lié à un rendement insuffisant.
- **Avantages pour le client** : Potentiel de rendement supérieur aux contrats de rente.

Ces produits sont souvent appelés « fonds distincts » ou « caisses séparées » et peuvent être assortis de garanties d’assurance qui ont pour effet de rendre les fonds insaisissables en cas de faillite du titulaire de contrat.

### 4.3.4 Produits avec ou sans participation

- **Sociétés mutuelles (avant 2000)** : Les assurés sont propriétaires, mais ne reçoivent ni gain en capital ni dividendes.
- **Produits participants** : Les assurés reçoivent une part des profits de la mutuelle.
- **Sociétés par actions** : Les assurés ne sont pas propriétaires.
  - Pour rester concurrentielles, elles offrent aussi des produits participants.
- **Produits non-participants** : Offerts par les deux types de sociétés avec une tarification plus agressive pour rester compétitifs.

## 4.4 Les sociétés mutuelles et les sociétés par actions

Comme on l'a vu à la section précédente, les sociétés d'assurance existent principalement sous la forme de sociétés mutuelles ou de sociétés par actions. Les autres formes (sociétés fraternelles, sociétés de secours mutuel, coopératives) ne concernent que de minuscules sociétés, souvent très vieilles.

### 4.4.1 Sociétés d'assurance mutuelles

**Dominance des mutuelles jusqu'à la fin des années 1990**
- Les mutuelles représentaient environ 25% des sociétés d'assurance de personnes mais contrôlaient 63% de l'actif total.
- À la fin des années 1990, elles dominaient le marché de l'assurance de personnes.

**Changement après 2000**
- Suite à une vague de démutualisations en 2000, la présence des mutuelles chute drastiquement.
- Elles ne représentent plus que 10% des sociétés et contrôlent seulement 6% de l'actif total des assureurs de personnes.

**Caractéristiques des sociétés mutuelles**
- Appartenant aux titulaires de polices avec participation, les sociétés mutuelles ont des droits de propriété non négociables.
- L'accès à de nouveaux capitaux est principalement limité à l'emprunt, souvent par émission d'obligations.

**Le Conseil d'administration des mutuelles**
- Élu par les mutualistes, le Conseil privilégie souvent la popularité des administrateurs sur leur compétence.
- Cela reflète une gouvernance centrée sur les valeurs mutualistes.

**Sources de croissance pour les mutuelles**
- Contrairement aux sociétés par actions, la croissance des mutuelles provient essentiellement de deux sources :
  - Les profits générés par la société.
  - Les emprunts contractés pour financer leur développement.

### 4.4.2 Sociétés d'assurance par actions

**Sociétés d'assurance par actions**

- Les sociétés par actions appartiennent à leurs actionnaires.
- Les actions sont transigées en Bourse, permettant le contrôle par les investisseurs.
- Le Conseil d'administration est nommé par les actionnaires.
- Accès à de nouveaux capitaux par émission d'actions.

**Démutualisation et rémunération des dirigeants**
- La démutualisation permet d'émettre de nouvelles actions et facilite les fusions et acquisitions.
- Les dirigeants peuvent recevoir une rémunération concurrentielle avec des options d'achat d'actions.
- Exemple de rémunération des dirigeants comparée à d'autres institutions financières.


## 4.5 La distribution de l’assurance

Les assureurs distribuent des produits via plusieurs canaux différents, qui dépendent du type de produit (assurance individuelle, collective ou rentes). Cette section traite principalement de l’assurance individuelle.

### 4.5.1 L’assurance collective

L’assurance collective est vendue à des groupes et comprend des protections variées, qui couvrent non seulement des adhérents (habituellement des employés) mais aussi leurs personnes à charge. Elle est distribuée via 3 principaux canaux :

- Des courtiers d’assurance spécialisés en assurance collective

- Des actuaires conseils qui font alors essentiellement un travail de courtage, en plus du design des régimes d’assurance

- Des distributeurs directs, qui vendent des produits d’assurance afférents à des biens ou services qu’ils transigent

  - Par exemple, des banques offrent de l’assurance vie et de l’assurance invalidité hypothécaire ainsi que de l’assurance sur d’autres types de prêts
  - Des concessionnaires automobiles distribuent de l’assurance sur des prêts automobiles financés par les sociétés de crédit des grands manufacturiers automobiles, tels que Financière GM, Crédit Ford, etc.

### 4.5.2 Les rentes

Les contrats de rente sont habituellement vendus directement par des représentants de l’assureur. Souvent, c’est le rentier futur qui contacte l’assureur pour se faire proposer divers projets de rente.

### 4.5.3 L’assurance individuelle

À cause du nombre élevé de produits différents et de la complexité de plusieurs produits, il n’est pas étonnant que les assureurs utilisent plusieurs canaux différents pour vendre leurs produits individuels.
On compte 5 principales familles d’intermédiaires qui distribuent l’assurance individuelle :

- Le représentant rattaché

- Le représentant autonome

- L’agent général

- Le courtier d’assurances

- Le distributeur

Toutes ces personnes détiennent un permis.

#### 4.5.3.1 Introduction

Le représentant en assurance de personnes est la personne physique qui offre directement au public, à un cabinet, à un représentant autonome ou à une société autonome des produits d’assurance individuelle de personnes ou des rentes individuelles d’un ou de plusieurs assureurs. Les représentants portent habituellement le titre de « Conseiller en sécurité financière ». 

L’Autorité des marchés financiers décerne le permis de Conseiller en sécurité financière au Québec (DEC plus examen de l’AMF) qui est un permis de vendre de l’assurance au Québec.

#### 4.5.3.2 Le représentant rattaché

- Le représentant rattaché est un employé de la société d’assurance qu’il représente.
- Il a une entente de service avec la société qui précise les services que cette dernière doit lui rendre.
- Il a une entente d’exclusivité avec la société qu’il représente, par laquelle il s’engage à ne vendre que les produits de cette société.
- La société doit lui procurer une assurance responsabilité professionnelle.
- Il gère les dossiers physiques de ses clients.
- Ses frais de bureau et son espace de travail peuvent être payés par l’assureur.
- Il est payé à commission mais peut aussi gagner un salaire de base.


#### 4.5.3.3 Le représentant autonome

- Le représentant autonome est un travailleur autonome et n’a pas de contrat d’exclusivité avec aucune compagnie d’assurance.
- Il négocie des ententes avec chacune des sociétés qu’il désire représenter et assume l’ensemble des responsabilités relatives à son travail.
- Il doit posséder une assurance responsabilité professionnelle et en assumer les frais.
- Il est responsable de ses frais de bureau et du coût de son espace de travail.
- Il est payé à commission.


#### 4.5.3.4 L’agent général

- L’agence générale est un regroupement de représentants sous la supervision d’un agent général, un contracteur indépendant de la société d’assurance.
- L'agent général est parrainé par un assureur et peut avoir l'exclusivité de la vente de certains produits sur un territoire délimité.
- Il agit comme un grossiste, simplifiant les transactions pour l'assureur en évitant de traiter avec de nombreux petits représentants ou courtiers.
- L'agent général peut régler les réclamations au nom de l’assureur et représente l’assureur.
- Les frais de bureau et le coût de l'espace de travail sont à la charge de l'agent général.
- Il reçoit une surcommission de la part de l’assureur, basée sur son volume d’affaires.

#### 4.5.3.5 Le courtier d’assurances

- Le courtier d’assurances est un contracteur indépendant de la société d’assurance qui a son propre cabinet et emploie plusieurs représentants.
- Il offre et vend les produits de différents assureurs.
- Il peut concevoir des produits, les faire accepter et tarifer par l’assureur et en être le vendeur exclusif.
- Les frais de bureau et le coût de l’espace de travail sont à la charge du cabinet de courtage.
- Le cabinet reçoit une surcommission de la part de l’assureur, basée sur son volume d’affaires.

##### Contrastes avec l'agent général :
- Contrairement à l’agent général, le courtier n’a pas le pouvoir de déterminer le tarif de prime de l’assurance en appliquant les règles de sélection et de tarification de l’assureur.
- Le représentant travaillant pour une agence générale est un employé du cabinet de courtage et non de l’assureur.
- Les commissions sont versées au cabinet qui les partage avec le représentant qui a réalisé la vente.


#### 4.5.3.6 Le distributeur

- Le distributeur, dans le cadre d'activités non liées à l'assurance, propose des produits d'assurance accessoires pour le compte d'un assureur, mais il n'est pas un spécialiste en assurance.
- L'assureur ne peut utiliser un distributeur que s'il fournit un guide de distribution approuvé par l'Autorité des marchés financiers du Québec.
- Le guide de distribution décrit le produit, spécifie la garantie, mentionne les exclusions et signale d'autres assurances similaires sur le marché.
- Si la rémunération du distributeur dépasse 30% de la prime, cela doit être indiqué dans le guide.
- La distribution est principalement limitée à des produits d'assurance simples, mais cela pose toujours des problèmes, en particulier dans les domaines de l'assurance vie, santé et perte d'emploi des débiteurs.

## 4.6 L’encadrement de l’assurance

Les activités d’assurance sont encadrées au niveau provincial pour ce qui est du contenu des produits et de leur commercialisation. Elles sont encadrées au niveau fédéral ou provincial pour tout ce qui touche la solvabilité de l’assureur, selon que sa charte a été émise par un gouvernement provincial ou par le gouvernement fédéral.

### 4.6.1 Le Bureau du Surintendant des institutions financières

Le bureau du Surintendant des institutions financières (BSIF en français, OSFI en anglais) existe en vertu d’une loi du parlement du Canada. Le BSIF est un organisme indépendant du gouvernement du Canada, établi en 1987 dans le but de rehausser la sûreté et la solidité du système financier canadien, dans l’intérêt du public. Il surveille et réglemente les banques et les sociétés d'assurances, les sociétés de fiducie et de prêt, ainsi que les régimes de retraite privés de compétence fédérale.
Il a pour mission de superviser les institutions financières pour s’assurer qu’elles sont en bonne santé financière et qu’elles se conforment aux lois qui les régissent et à ses exigences découlant de l’application de ces lois.
Ses objectifs sont les suivants :

- Superviser les institutions financières pour s’assurer qu’elles sont en bonne santé financière et qu’elles se conforment aux lois qui les régissent et à ses exigences découlant de l’application de ces lois.

- Aviser sans délai la direction et le conseil d’administration de la mauvaise situation financière de l’institution ou de son défaut de se conformer aux lois qui la régissent ou à ses exigences et, le cas échéant, de prendre ou de forcer la direction ou le conseil à prendre des mesures pour corriger la situation sans plus attendre.

- Inciter la direction et le conseil d’administration des institutions financières à se doter de politiques et de procédures pour contrôler et gérer le risque.

- Surveiller et évaluer dans l’ensemble du système ou dans un secteur d’activités en particulier les événements ou les questions qui risquent d’avoir des répercussions négatives sur la situation financière des institutions.

Il élabore les normes de fonds propres à l’intention des institutions financières canadiennes en plus de participer à l’établissement de normes de fonds propres internationales saines. 

Il a émis des lignes directrices touchant :

- Les fonds propres des assureurs

- Les limites et restrictions prudentielles

- La comptabilité et la communication de renseignements

- Les saines pratiques commerciales et financières en général

- Les normes de pratiques commerciales et financières à l’intention des sociétés d’assurance de personnes

Une visite sur le site du BSIF vous permettra de constater la quantité et l’étendue des normes auxquelles les institutions financières doivent se conformer.

Le BSIF administre les lois suivantes, qui lui confèrent certains pouvoirs :

- Loi sur les banques,

- Loi sur les sociétés de fiducie et de prêt,

- Loi sur les associations coopératives de crédit,

- Loi sur les sociétés d’assurances,

- Loi de 1985 sur les normes de prestation de pension,

- Tous les règlements d’application de ces lois.

En pratique, le BSIF réglemente plus de 400 institutions financières (y compris leurs succursales et leurs filiales étrangères) et quelque 1 200 régimes de retraite au Canada appartenant aux catégories suivantes :

- Banques canadiennes

- Banques étrangères

- Succursales de banques étrangères à service complet

- Succursales de banques étrangères n’offrant que des services de prêt

- Sociétés de fiducie

- Sociétés de prêt

- Associations coopératives de crédit

- Associations coopératives de détail

- Sociétés d’assurance-vie :

  - Sociétés constituées ou prorogées sous la Loi sur les sociétés d’assurances (dites à charte fédérale)
  - Sociétés étrangères ayant obtenu une ordonnance portant garantie des risques au Canada
  - Sociétés de secours mutuels

- Sociétés d’assurances multirisques :

  - Couvrent les risques de toutes les branches d’assurance, à l’exception de l’assurance-vie
  - Sociétés constituées ou prorogées sous la Loi sur les sociétés d’assurances (dites à charte fédérale)
  - Sociétés étrangères ayant obtenu une ordonnance portant garantie des risques au Canada

- Régimes de retraite privés fédéraux (organismes de compétence fédérale, tels que les télécommunications et les transports interprovinciaux)

### 4.6.2 L’Autorité des marchés financiers du Québec

L’Autorité des marchés financiers du Québec (AMF) existe en vertu d’une loi de l’Assemblée nationale du Québec. L’Autorité est l’organisme mandaté par le gouvernement du Québec pour encadrer le secteur financier québécois et prêter assistance aux consommateurs de produits et services financiers. Elle se distingue par un encadrement intégré des domaines de l’assurance, des valeurs mobilières, des instruments dérivés, des institutions de dépôt – à l’exception des banques – et de la distribution de produits et services financiers.

L’Autorité a pour mission de :

- Prêter assistance aux consommateurs de produits et utilisateurs de services financiers en cas de différends.

- Veiller à ce que les institutions financières et autres intervenants du secteur financier respectent les normes de solvabilité qui leur sont applicables et se conforment aux obligations que la loi leur impose en vue de protéger les intérêts des consommateurs.

- Assurer l’encadrement des activités de distribution de produits et services financiers en administrant en outre les règles d’admissibilité et d’exercice de ces activités et en prenant toute mesure prévue à la loi à ces fins.

- Assurer l’encadrement des activités de bourse et de compensation et l’encadrement des marchés de valeurs mobilières ainsi que l’encadrement des marchés financiers en général.

- Assurer l’encadrement des marchés de produits dérivés.

- Voir à la mise en place de programmes de protection et d’indemnisation des consommateurs de produits et utilisateurs de services financiers et administrer les fonds d’indemnisation prévus à la loi.

Les activités d’encadrement de l’Autorité se rapportent principalement aux domaines suivants :

- Les assurances et institutions de dépôt.

- La distribution de produits et services financiers.

- Les valeurs mobilières.

- L’assistance à la clientèle et l’indemnisation.

L’Autorité décerne le droit de pratique pour offrir chacun des produits et services financiers suivants au Québec :

- L’assurance individuelle de personnes (assurance vie et assurance contre la maladie et les accidents).

- L’assurance collective de personnes.

- L’Assurance de dommages (IARD).

- L’expertise en règlement de sinistres.

L’Autorité fait appel à des organismes d’autoréglementation (OAR) à qui sont délégués certains pouvoirs d’encadrement ou de réglementation de la conduite de leurs membres, notamment en ce qui a trait à l’élaboration de règles, le contrôle de la conformité des membres avec les règles, la discipline des membres de :

- La Chambre de l’assurance de dommages.

- La Chambre de la sécurité financière.

- L’Organisme canadien de réglementation du commerce des valeurs mobilières (OCRCVM).

## 4.7 Assuris: Protection des assurés en cas de faillite

- **Organisation à but non lucratif** protégeant les assurés canadiens en cas de faillite d’une compagnie d’assurance vie.
- **Mission** : Minimiser la perte des prestations et transférer rapidement les polices à une compagnie solvable.
- **Obligation d’adhésion** : Toute compagnie d’assurance vie opérant au Canada doit être membre d’Assuris.
- **Création** : Fondée en 1990, reconnue par le ministère fédéral des Finances.
- **Cadre légal** : Mentionnée dans la Loi sur les compagnies d’assurance du Canada et dans le Règlement d’application de la Loi sur les assurances du Québec.
- **Garantie en cas de faillite** : Assuris protège les sommes assurées dans certaines limites.

# Fin

- Pour l’assurance vie temporaire (incluant l’assurance vie collective) autre que la Temporaire à 100 ans :

  - Si le capital au décès ≤ 200 000 \\$ : plein montant
  - Si le capital au décès > 200 000 \\$ : Max (200 000 \\$ ou 85% du capital au décès)

- Pour l’assurance vie universelle :

  - Si le capital au décès ≤ 200 000 \\$ : plein montant
  - Si le capital au décès > 200 000 \\$ : Max (200 000 \\$ ou 85% du capital au décès)
  - Si la somme accumulée dans le compte ≤ 60 000 \\$ : plein montant
  - Si la somme accumulée dans le compte > 60 000 \\$ : Max (60 000 \\$ ou 85% de la somme)

- Pour l’assurance vie entière :

  - Si le capital au décès ≤ 200 000 \\$ : plein montant
  - Si le capital au décès > 200 000 \\$ : Max (200 000 \\$ ou 85% du capital au décès)
  - Si la valeur de rachat ≤ 60 000 \\$ : pleine valeur de rachat
  - Si la valeur de rachat > 60 000 \\$ : Max (60 000 \\$ ou 85% de la valeur de rachat)

- Pour l’assurance maladie individuelle ou collective, incluant l’assurance individuelle contre les maladies graves :

  - Si les frais remboursables ≤ 60 000 \\$ : plein remboursement
  - Si les frais remboursables > 60 000 \\$ : Max (60 000 \\$ ou 85% des frais remboursables)

- Pour l’assurance invalidité individuelle ou collective, ainsi que de l’assurance individuelle des soins de longue durée :

  - Si les prestations mensuelles ≤ 2 000 \\$ : plein remboursement
  - Si les prestations mensuelles > 2 000 \\$ : Max (2 000 \\$ ou 85% des prestations mensuelles)

- Pour les rentes en cours de paiement :

  - Si les prestations mensuelles ≤ 2 000 \\$ : plein remboursement
  - Si les prestations mensuelles > 2 000 \\$ : Max (2 000 \\$ ou 85% des prestations mensuelles)

- Pour les produits individuels de fonds distincts, la valeur des fonds n’est pas influencée par la faillite de l’assureur mais les valeurs garanties le sont :

  - Si la valeur garantie ≤ 60 000 \\$ : pleine valeur garantie
  - Si la valeur garantie > 60 000 \\$ : Max (60 000 \\$ ou 85% de la valeur garantie)

- Pour les rentes en capitalisation, individuelles ou collectives (encore en phase d’accumulation) :

  - Si la valeur accumulée ≤ 100 000 \\$ : pleine valeur accumulée
  - Si la valeur accumulée > 100 000 \\$ : 100 000 \\$


Les REER, FERR et CELI sont traités en fonction du type de véhicule dans lequel ils sont placés (fonds distincts, rentes, etc.). Les limites précitées s’appliquent à l’ensemble des prestations d’assurance d’une même catégorie.


Au 30 septembre 2019, Assuris a un fonds de liquidités de 166 M \\$. En cas d’insolvabilité, Assuris se servira de ce fonds pour honorer ses engagements, avant d’imposer une cotisation à ses sociétés membres. Elle a donc la capacité financière de recueillir suffisamment de fonds pour faire face à la faillite de toute société membre.
