# Module 6: Assurance vie

## Objectifs du Module 6
- Connaître les notions de sélection et d’anti-sélection.
- Comprendre la sélection des risques.
- Comprendre le fonctionnement des tables de mortalité.
- Connaître les principales caractéristiques des produits traditionnels d’assurance vie (assurance vie temporaire et permanente).

## 6.1 Les outils de base servant aux calculs actuariels d’assurance vie

Le premier calcul qu’on a à faire en assurance vie est le calcul du coût actuariel de l’assurance. Pour une année d’assurance, ce coût est le suivant :

Coût = (taux de mortalité) x (montant d’assurance) x (facteur d’escompte) où :
- Taux de mortalité = probabilité que l’assuré décède cette année.


- Pour un assuré d'âge $\mathrm{x}$, on dénote cette probabilité par $\mathrm{q}_{\mathrm{x}}$ (à noter qu'en actuariat, $\mathrm{q}_{\mathrm{x}}$ désigne la probabilité de changer d'état, par exemple, passer de vivant à décédé, passif d'actif à invalide, passer d'invalide à actif, passer d'invalide à décédé, etc.; quand plus d'un changement d'état est possible, on qualifie en inscrivant sa nature en exposant, comme $\mathrm{q}_x{ }^{\text {invalidité }}$ ou $\mathrm{q}_x{ }^{\text {mortalité}}$ ou $\mathrm{q}_x{ }^{\text {mortalité d'un invalide}}$; si un seul changement d'état est possible, on sous-entend la nature du $\mathrm{q}_{\mathrm{x}}$ sans l'identifieret on peut alors utiliser la position d'exposant pour dénoter la population à laquelle s'applique ce $\mathrm{q}_{\mathrm{x}}$ ).

- Si on veut préciser que l'assuré est une femme, on peut dénoter cette probabilité par $\mathrm{q}_x{ }^{\text {Femme}}$.
- Si on veut préciser que l'assuré est une femme et qu'elle ne fume pas, on peut dénoter cette probabilité par $\mathrm{q}_x{ }^{\text {Femme Non-fumeur}}$.


- Montant d’assurance = montant payable au décès de l’assuré.
- Facteur d’escompte = facteur représentant l’effet de l’intérêt entre le moment où on établit la prime et le moment où on paye le montant d’assurance.
    - On dénote ce facteur par $v^t$.
    - «t» est le laps de temps en années entre le moment où on établit la prime et le moment où on s'attend à payer le montant d'assurance. Par exemple, si on présume que la personne meurt en moyenne en milieu d'année et qu'on paye immédiatement, $t=1 / 2$ mais si on présume qu'on ne payera qu'à la fin de l'année, $t=1$; finalement, si on pense payer dans 7 ans et demie, $t=7,5$.
    - «v» est appelé «facteur d’actualisation» et est égal à 1 / (1 + i) où «i» est le taux d’intérêt annuel.

## 6.2 Les tables de mortalité – considérations générales

Il existe différentes sortes de tables de mortalité. Essentiellement, il y a 5 principales familles de tables de mortalité :
- Les tables de population, préparées par des agences gouvernementales, qui comprennent tous les décès, y compris ceux des grands malades, et permettent de :
    - Comparer la mortalité d’un pays à l’autre.
    - Comparer la mortalité d’une région à l’autre dans un même pays.
    - Établir des politiques sociales, comme l’âge d’admissibilité à la Pension de sécurité de la vieillesse.
- Les tables d’assurance individuelle, qui permettent de tarifer les produits d’assurance vie individuelle.
- Les tables de rentes individuelles, qui permettent de tarifer les produits de rente individuelle.
- Les tables d’assurance vie collective, qui permettent de tarifer les produits d’assurance vie collective.
- Les tables de régimes de rentes, qui permettent d’estimer les coûts des régimes de retraite.

Les différences entre ces tables de mortalité sont dues aux effets de la sélection et de l’anti-sélection.

## 6.3 L’anti-sélection

  - L'anti-sélection survient lorsque les assurés ont plus d'informations sur leurs risques de santé que l'assureur, menant potentiellement à des primes injustement basses.

  - Exemple : Un individu avec des problèmes de santé non divulgués (ex. mutations géniques BRCA-1 ou BRCA-2 augmentant le risque de cancer du sein) peut bénéficier d'une prime plus basse en raison du manque d'informations détaillées sur sa santé par l'assureur.

  - En assurance-vie, l'absence de sélection médicale peut attirer des individus à plus haut risque de santé ou pratiquant des activités dangereuses, augmentant les coûts pour les assureurs.

  - Pour les produits de rente, les assureurs ne font généralement pas de sélection médicale. Cela tend à attirer des individus en meilleure santé, car ceux en mauvaise santé préfèrent accéder plus rapidement à leurs épargnes en raison d'une espérance de vie plus courte.

  - Les assureurs contrebalancent l'anti-sélection en appliquant des taux de mortalité plus élevés pour l'assurance-vie (sans informations médicales récentes) et des taux plus bas pour les rentes.

  - Paradoxalement, un individu pourrait se voir attribuer des taux de mortalité différents pour une assurance-vie et une rente, soulignant la complexité de l'évaluation des risques dans les produits d'assurance.

## 6.4 La sélection des risques
La sélection des risques permet de classer les risques selon leur probabilité de survenance.

### 6.4.1 Buts de la sélection

- Contrer l’anti-sélection
    - Si on veut réaliser une mise en commun juste et équitable, on doit regrouper des risques qui sont homogènes, donc des risques qui ont à peu près la même probabilité de se réaliser.
- Maintenir l’équité entre les différents assurés
    - En classant les assurés dans des regroupements homogènes, on s’assure que des risques semblables seront traités de façon cohérente et équitable.
- Faciliter la mise en commun des risques
    - En classant les assurés dans des regroupements homogènes, (âge-sexe-état de santé) on peut mettre en commun des risques semblables.
- Garantir que les primes soient proportionnelles au coût actuariel
    - En classant les assurés dans des regroupements homogènes quant au risque de mortalité, on peut leur appliquer un taux de mortalité qui représente leur risque.
    - En appliquant la formule Prime = (taux de mortalité) x (montant d’assurance) x (facteur d’escompte) + dépenses, on s’assure que le coût actuariel et les primes aient une relation proportionnelle.

## 6.4.2 Fonctionnement de la sélection
  - Les assureurs sélectionnent parmi les candidats ceux présentant un niveau de risque acceptable, les classant dans des catégories spécifiques pour fixer les primes en fonction de la mortalité moyenne attendue de ces catégories (basée sur l'âge, le genre, le statut fumeur/non-fumeur, etc.).
  - La sélection implique généralement :
    - Un questionnaire médical pour évaluer les risques de santé, incluant l'historique familial, l'usage de tabac et de drogues, les loisirs et l'emploi pour identifier les activités à risque, la détention d'autres polices d'assurance, le casier judiciaire, ainsi que les revenus et la fortune du demandeur.
    - Des examens médicaux comme des analyses de sang et d'urine, la mesure de la tension artérielle, etc.
    - Un examen médical complet.
    - Un rapport du conseiller en sécurité financière.
    - Une consultation au Medical Information Bureau (MIB) pour vérifier les informations médicales.
  - Lors d'une demande d'assurance, une autorisation est signée permettant à l'assureur d'accéder aux dossiers médicaux et de partager ces informations avec le MIB, que d'autres assureurs peuvent consulter ultérieurement pour s'assurer de l'exactitude des informations fournies.
  - La rigueur de la sélection augmente avec le montant de la couverture d'assurance demandée.



## 6.4.3 Décisions possibles de l’assureur

Après avoir analysé les éléments décrits en 6.4.2, l’assureur prend une des décisions suivantes :

- Accepter le demandeur comme étant un risque standard :
    - La classe standard doit être assez large pour contenir un fort pourcentage des assurés car on veut obtenir une mortalité stable et prévisible, réduire les coûts administratifs, encourager le conseiller en sécurité financière, et avoir une bonne image publique.

- Classer le demandeur parmi les risques tarés :
    - Le risque de mortalité supplémentaire doit être significatif pour qu’il soit classé taré.
    - Les risques tarés se font imputer une mortalité plus élevée en fonction de leur degré de tare.

- Réduire le montant d’assurance demandé :
    - Que le risque soit standard ou surprimé, l’assureur peut décider d’émettre l’assurance pour un montant inférieur au montant demandé.

- Différer ou reporter l’acceptation du risque :
    - Si l’analyse du risque n’est pas concluante, l’assureur peut exiger des informations additionnelles avant d’émettre la police.

- Refuser le risque :
    - Si aucune surprime n’est assez élevée pour permettre l’acceptation d’un risque taré, l’assureur va simplement refuser d’émettre l’assurance.


## 6.5 Les tables de mortalité utilisées en assurance vie individuelle

Les tables de mortalité utilisées en assurance individuelle existent en trois versions :



- **Table sélecte**
    - La table sélecte est une table à 4 dimensions, qui représente l’évolution de la mortalité en tenant compte de la durée écoulée depuis la sélection. Elle contient des taux de mortalité variant selon :
        - L’âge à l’émission de la police (= âge à la sélection du risque)
        - La durée écoulée depuis l’émission de la police
        - Le sexe de l’assuré (hommes et femmes seulement – sexe à la naissance)
        - Le statut de tabagisme (fumeur vs non-fumeur)

- **Table ultime**
    - La table ultime est une table à 3 dimensions, qui représente l’évolution de la mortalité lorsque la sélection est suffisamment ancienne pour n’avoir plus aucun impact sur la mortalité. Elle contient des taux de mortalité variant selon :
        - L’âge atteint
        - Le sexe de l’assuré (hommes et femmes seulement – sexe à la naissance)
        - Le statut de tabagisme (fumeur vs non-fumeur)

- **Table agrégée**
    - Il s’agit de la combinaison d’une table sélecte et d’une table ultime. C’est donc, comme la table ultime, une table à 3 dimensions, qui contient des taux de mortalité variant selon :
        - L’âge atteint
        - Le sexe de l’assuré (hommes et femmes seulement – sexe à la naissance)
        - Le statut de tabagisme (fumeur vs non-fumeur)

Quelques remarques sur les tables précitées :

- En actuariat, on dénote par le symbole " $\mathbf{q}_x$ " la probabilité de changer d'état pour une personne d'âge «x», comme par exemple :
    - passer de vivant à décédé,
    - passer de travailleur actif à travailleur retraité,
    - passer de travailleur actif à travailleur invalide,
    -  passer de travailleur invalide à travailleur actif (rétablissement),
    -  et passer de travailleur invalide à travailleur décédé.

Dans le cas des tables de mortalité d'assurance individuelle, le symbole $\mathrm{g}_{\mathrm{x}}$ représente la probabilité de décès à l'âge « $\mathrm{X}$ ».  



De même, le symbole $\mathbf{q}_{x^H} \mathbf{r}$ représente la probabilité de décès d'un homme à l'âge " $X$ " alors que le symbole $\mathbf{q}_x{ }^F$ représente la probabilité de décès d'une femme à l'âge «X».

Quand on a une table de mortalité sélecte, on dénote par $q_{[x]+N}$ la probabilité de décès à l'âge « $x+N$ » dans le cas d'une personne qui a subi une sélection médicale à l'âge «x».  

De même, le symbole $\mathrm{q}[\mathrm{x}]+\mathrm{N}^{\mathrm{H}}$ représente la probabilité de décès à l'âge « $x+N$ » d'un homme qui a subi une sélection médicale à l'âge «X», alors que le symbole $\mathrm{q}_{[x+N]}{ }^F$ représente la probabilité de décès à l'âge « $x+N$ " d'une femme qui a subi une sélection médicale à l'âge «X».

- Habituellement, les assureurs ne font pas de sélection médicale au-delà de 80 ans car le risque qu’un problème de santé ne soit pas détecté est trop grand.
- De façon générale, les tables de mortalité d'assurance individuelle se terminent à l'âge de 120 ans. On appelle l'âge limite de la table «âge Oméga» (âge $\omega$ ou âge $\Omega$ ), parce que la dernière lettre de l'alphabet grec est la lettre Oméga.

## 6.6 Assurance vie individuelle

Il existe deux grandes familles d’assurance vie :

- Assurance vie temporaire  
  L’ACCAP (Association canadienne des compagnies d'assurances de personnes) définit l’assurance vie temporaire comme une assurance vie selon laquelle il y a le paiement du capital assuré si le décès de l’assuré survient avant l’expiration d’une période déterminée.

- Assurance vie permanente  
  L’ACCAP définit l’assurance vie permanente comme une assurance vie qui procure une protection à vie de l’assuré, moyennant le paiement de primes tel que prévu dans la police, et qui comporte un élément d’épargne permettant d’accumuler une valeur de rachat.


En pratique, bien que les deux définitions soient mutuellement exclusives, il existe des produits hybrides qui combinent des caractéristiques des deux familles, tels que :

- **Assurance vie temporaire à 100 ans** : Produit d'assurance permanente sans valeur de rachat, à prime nivelée, où les fonds des assurés renonçant à leur couverture financent les prestations des autres, réduisant ainsi les primes par rapport à une assurance vie permanente classique.

- **Assurance vie universelle** : Assurance permanente à prime variable combinant assurance vie temporaire et fonds de placement, visant à maximiser les avantages fiscaux des prestations de décès non imposables et des intérêts accumulés.

- **Assurance conjointe** : Option permettant d'assurer deux personnes sous une même police, avec prestation de décès payable au premier ou second décès, selon le contrat.


### L’assurance vie temporaire en général

- **But de l'assurance vie temporaire** : Offre une protection à court terme à coût réduit, idéale pour des périodes de besoin élevé d'assurance, comme l'éducation des enfants, avec une finition avant l'augmentation significative des taux de mortalité.

- **Usages courants** : 
  - Complète les assurances permanentes durant les périodes financièrement exigeantes.
  - Base pour une stratégie d'investissement, en investissant la différence d'économie par rapport à une assurance permanente dans des placements à rendement potentiellement plus élevé.

- **Caractéristiques** : 
  - Coût minimal avec des options de durée variées (1 à 100 ans).
  - Prime nivelée pendant la durée choisie, renouvelable et transformable en police permanente sans preuve d’assurabilité supplémentaire.

- **Renouvellement et transformation** : 
  - Possibilité de renouveler sans nouvelles preuves d'assurabilité, malgré l'augmentation de la prime avec l'âge.
  - Option de transformation en assurance permanente, garantissant une prime fixe à vie.

- **Options supplémentaires** : 
  - Inclut parfois la possibilité d'ajouter de l'assurance sans examen médical lors d'événements de vie clés.
  - Aide à protéger l'assurabilité future et à répondre à des besoins d'assurance croissants.

- **Résiliation** : 
  - Possible par simple cessation de paiement des primes, généralement sans valeur de rachat, reflétant une structure tarifaire anticipant de telles résiliations.

- **Avantages** : 
  - Solution économique pour une protection temporaire.
  - Particulièrement adaptée aux jeunes familles et individus à budget limité nécessitant une couverture significative.
  - Idéale pour sécuriser des prêts ou des hypothèques.


## Types d'Assurance Temporaire et Leur Fonctionnement

Comprendre l'effet de la mortalité et de l'intérêt est crucial en assurance. La mortalité masculine, souvent plus élevée que celle des femmes, est généralement utilisée pour les illustrations, augmentant de manière significative après 40 ans due aux maladies liées au vieillissement.

**Assurance Temporaire de 1 An Renouvelable (T-1)** 

Ce produit est peu courant mais essentiel comme base pour d'autres types d'assurance. Les coûts administratifs ont un impact majeur aux jeunes âges où la mortalité est très faible. Les primes nettes, calculées directement à partir des taux de mortalité et sans frais, bien que ces derniers influencent le niveau des primes.

**Assurance Temporaire de 10 Ans Renouvelable (T-10)**

La T-10 avec des primes nivelées crée une provision mathématique pour compenser l'augmentation de mortalité dans la seconde moitié de la période. Ce produit est généralement renouvelable et transformable, mais la provision mathématique n'appartient jamais au titulaire de la police.

**Assurance Temporaire de 20 Ans (T-20)**

Similaire à la T-10, la T-20 utilise une provision mathématique pour couvrir le coût croissant de la mortalité sur 20 ans. Ce produit n'est habituellement ni renouvelable ni transformable, avec une réserve maximale plus élevée que la T-10.

**Assurance Temporaire jusqu'à 65 Ans (T-65)**

Une T-65, avec des primes nivelées jusqu'à 65 ans, montre une accumulation puis utilisation de la provision mathématique pour l'excédent de mortalité. Non renouvelable et non transformable, la réserve atteint son sommet vers l'âge de 54 ans et diminue rapidement par la suite.

**Comparaison des Réserves Actuarielles**

La longueur du terme et la courbe de mortalité influencent l'asymétrie et le sommet de la réserve actuarielle, qui reste modeste par rapport à la prestation de décès. Plus le terme est long, plus la réserve monte haut, mais dans tous les cas, les réserves restent très modestes par rapport au montant assuré.

## L'Assurance Permanente

L'assurance permanente est définie comme une protection à vie, avec paiement de primes selon les termes de la police, et inclut un élément d'épargne permettant d'accumuler une valeur de rachat.

### Assurance Vie-Entière

- **Définition** : Type d'assurance permanente qui garantit le paiement du capital au décès, à n'importe quel moment.
- **Primes** : Réparties sur la durée du contrat ou une période définie, plus élevées pour des périodes de paiement plus courtes.
- **Réserve de T-100 vs Vie-Entière** : La réserve de T-100 reste inférieure à celle de la vie-entière jusqu'à 100 ans, où les deux valeurs convergent.

### Caractéristiques de l'Assurance Vie-Entière

- Couverture à vie, avec des primes et un montant d'assurance au décès garantis.
- La valeur de rachat accumule l'excédent des primes sur le coût de la mortalité, augmenté des revenus de placement, et égale le montant d'assurance à 100 ans.
- Possibilité de primes nivelées sur toute la vie ou une période plus courte, et option de prime unique pour une couverture immédiatement libérée.

### Rachat et Valeur de Rachat

- Le titulaire peut faire racheter sa police par l'assureur pour recevoir la valeur de rachat accumulée, avec des options pour maintenir la police en cas de non-paiement des primes.

### Options de Non-Déchéance et Avance sur Contrat

- Plusieurs options si le titulaire cesse de payer les primes, comme l'avance de la prime par l'assureur, l'utilisation de la valeur de rachat pour le paiement des primes, ou la conversion en une assurance permanente avec montant réduit sans paiement de primes supplémentaires.
- Possibilité d'emprunter sur la valeur de rachat avec remboursement à tout moment.

### Participations

- Certains contrats d'assurance vie-entière prévoient des participations basées sur les profits de l'assureur, avec plusieurs manières d'utiliser ces participations, telles que le versement au comptant, l'application en réduction de la prime, ou l'achat d'assurance additionnelle.

### Assurance Vie à Capital Variable et Assurance Vie Universelle

- **Développement** : Face à la concurrence des sociétés de placement, les assureurs ont créé l'assurance vie universelle, combinant des primes minimales pour la couverture de mortalité avec des options de placement avantageuses fiscalement.
- **Principes de l'Assurance Vie Universelle** : Combine l'assurance vie avec un fonds d'accumulation, offrant une transparence accrue et la possibilité de varier les primes, la structure des coûts d'assurance, et le choix des investissements.
- **Souplesse** : Permet des ajustements des primes, du montant d'assurance, et des options de placement selon les préférences du titulaire, offrant une solution moderne et concurrentielle aux besoins d'assurance et d'investissement.

## Les Avenants

Un avenant est une modification ou un ajout à une police d'assurance qui change la couverture et les prestations. Voici quelques types courants d'avenants :

### Exonération des Primes
- En cas d'invalidité totale, l'assureur renonce au paiement des primes après 90 jours de carence et rembourse les primes payées depuis le début de l'invalidité.

### Rente d'Invalidité
- Offre un revenu récurrent en cas d'invalidité du payeur de primes, nécessitant l'avenant d'exonération des primes.

### Droit d'Acheter de l'Assurance Supplémentaire
- Permet d'augmenter la couverture sans nouvelles preuves de bonne santé lors d'événements de vie spécifiques.

### Assurance Décès ou Infirmité à la Suite d'un Accident
- Fournit un montant forfaitaire en cas de décès accidentel ou de perte de certaines fonctions à cause d'un accident, selon un barème prédéfini.

#### Couverture Familiale
- Propose une assurance pour les enfants et parfois le conjoint, généralement avec un capital assuré modeste, principalement pour couvrir les frais liés au décès.

#### Assurance Maladies Graves et Assurance des Soins de Longue Durée
- Offrent des prestations pour des situations médicales graves ou des besoins de soins à long terme, ajoutant une protection spécifique au-delà de l'assurance vie standard.


## 6.8 Tarification de l'assurance vie

La prime d'assurance est basée sur plusieurs éléments, dont :
- Le coût pur de l'assurance (prime pure), calculé comme la probabilité de décès multipliée par le capital assuré, basée sur des tables de mortalité.
- Les frais imputés au contrat.
- L'hypothèse de taux de rendement net estimé sur les provisions techniques (réserves actuarielles).
- L'hypothèse de taux de déchéance.
- Les taxes provinciales sur les primes.


### 6.8.1 La prime pure
- La prime pure est le coût de l'assurance pour l'assureur, dépendant de l'âge, du sexe, de la consommation de tabac, des antécédents médicaux, de la consommation d'alcool et de drogues, de la pratique de sports dangereux, et du dossier de conduite.

Après analyse de la proposition, le tarificateur peut :
- Refuser d'assurer.
- Accepter au tarif standard ou à un tarif privilégié.
- Accepter avec une surprime si le profil de l'assuré présente un risque supérieur à la normale, souvent avec une entente de réassurance.

### 6.8.2 Les frais

Les frais inclus par l'assureur dans la prime, souvent appelés « frais de chargement », comprennent :

- **Frais de sélection des assurés** : Coûts associés à l'évaluation du risque présenté par chaque assuré.
- **Frais de distribution** : Commissions versées aux conseillers en sécurité financière pour couvrir leurs dépenses opérationnelles et leur rémunération. Ces commissions peuvent être plus élevées pour les courtiers ou agents généraux indépendants.
- **Surcommissions** : Payées à l'agent général ou au directeur du centre de services financiers pour couvrir les frais de fonctionnement et la rémunération du personnel de gestion.
- **Bonis de production** : Incitatifs financiers pour les conseillers en sécurité financiers et la direction.
- **Frais d'administration** : Coûts associés à la gestion quotidienne des polices d'assurance.
- **Frais liés au paiement des prestations** : Coûts engendrés lors du traitement des réclamations.
- **Marge de profit de l'assureur** : Bénéfice intégré à la tarification pour assurer la rentabilité de l'entreprise.

### 6.8.3 Le taux de rendement

Le taux de rendement crédité sur les réserves actuarielles reflète le rendement des placements de l'assureur, net de l'impôt de 15% sur les revenus de placement. Plus l'échéance de la prestation de décès est éloignée, plus le rendement estimé est élevé, ce qui réduit la réserve actuarielle et, par conséquent, la prime.

### 6.8.4 Les taux de déchéance

Les taux de déchéance influencent directement la tarification de l'assurance. Des taux élevés signifient moins de prestations versées et des primes plus basses. Les taux varient en fonction de l'âge, du sexe, du statut tabagique des assurés, et du produit d'assurance. La majorité des contrats sont annulés avant le décès de l'assuré, influençant ainsi la tarification et la rentabilité des produits d'assurance.

### 6.8.5 Les taxes provinciales

Au Québec, deux taxes s'appliquent aux primes d'assurance :
- **Impôt sur le capital des sociétés d'assurance de personnes** : 3% des primes, augmenté de 2% à 3% en 2014.
- **Taxe compensatoire** : 0.48% des primes, initialement destinée à financer la surveillance des assureurs par les autorités, mais conservée comme source de revenus gouvernementaux après l'harmonisation de la TVQ avec la TPS.
