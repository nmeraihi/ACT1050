## Module 5 : Aspect juridique

### Objectifs du cours #5

- Distinguer entre les régimes soumis à la loi fédérale sur les normes de prestations de pension et ceux soumis à la loi provinciale.
- Connaître et comprendre les principales dispositions juridiques pertinentes au fonctionnement d’un régime de retraite.
- Identifier les principales obligations qui en découlent.
- Reconnaître le rôle et l’importance de la jurisprudence.

## 5.1 Survol

- Le **Droit** se divise en **droit civil** et **droit criminel**, les régimes de retraite étant régis par le **droit civil**.
- Au Canada, il y a deux systèmes juridiques en droit civil : le **Code civil du Québec**, spécifique au Québec, et la **Common Law**, applicable dans les autres provinces.


- Les **lois régissant les régimes de retraite** incluent :
  - **Lois codifiées** telles que le **Code civil du Québec** ou la **Common Law**
  - **Loi sur les régimes complémentaires de retraite** au Québec
  - **Loi sur les normes de prestations de pension** pour les entités sous juridiction fédérale
- La **jurisprudence** joue également un rôle important dans l'interprétation et l'application de ces lois.
- Les gouvernements, en tant que garants de l'ordre public, assurent que les régimes de retraite profitent uniquement aux participants et respectent leurs engagements.


## 5.2 Fonctionnement du système juridique

- Au-delà des principes du **droit civil ou de la Common Law**, les **lois étatiques** et la **jurisprudence** (décisions judiciaires) jouent un rôle clé.
- Dans les systèmes de **droit civil**, contrairement à la **Common Law**, la jurisprudence n'est pas une source principale du droit, mais elle reste importante pour **interpréter les lois** et le code civil.
- Un jugement significatif peut conduire le gouvernement à **formuler une nouvelle loi** pour clarifier ou annuler une interprétation jurisprudentielle.

## 5.3 Les entreprises sous juridiction fédérale

- Au Canada, la régulation des relations entre individus et entre ceux-ci et les entreprises est principalement une compétence provinciale, incluant des domaines tels que le travail, la santé et la sécurité, les régimes de retraite complémentaires et les assurances.
- Certaines lois spécifiques, comme la Charte de la langue française, sont exclusives à des provinces comme le Québec.



- La régulation des relations individuelles et commerciales est majoritairement une compétence **provinciale** au Canada, y compris pour les **régimes de retraite**.
- L'**assurance emploi** est passée d'une compétence provinciale à fédérale dans les années 1940.
- Le gouvernement fédéral contrôle les secteurs interprovinciaux, les **banques** et les **télécommunications**.
- Les régimes de retraite des entreprises **fédérales** sont régis par la **Loi sur les normes de prestations de pension fédérale**.


## 5.4 Dispositions pertinentes du Code civil du Québec

Le Code civil comporte très peu de dispositions intéressant les régimes de retraite. Toutefois, il établit ou définit certains concepts :
- La notion de contrat et comment on doit interpréter un contrat (articles 1377 et suivants);
- La notion de contrat d’adhésion (contrat dont les stipulations essentielles ont été imposées par l’une des parties ou rédigées par elle, pour son compte ou suivant ses instructions, et ne pouvaient être librement discutées), comme par exemple les contrats de rente émis par les assureurs; dans tous les cas, le contrat d’adhésion s’interprète en faveur de l’adhérent ou du consommateur (articles 1379 et 1432);


- Rente viagère : la durée de son service est limitée au temps de la vie d’une ou de plusieurs personnes (article 2371);
- Le capital accumulé pour le service de la rente est insaisissable, tant que ce capital demeure affecté au service d’une rente mais seulement pour la partie du capital qui serait nécessaire pour servir, pendant la durée prévue au contrat, une rente qui satisferait les besoins d’aliments (nourriture, logement, vêtements, chauffage) du rentier (article 2378);
- Celui qui a la responsabilité de verser la rente a la faculté de se faire remplacer par un assureur autorisé en lui versant la valeur de la rente qu’il doit (article 2384).


## 5.5 Principales exigences statutaires régissant les régimes de retraite

- Les régimes de retraite au Canada sont encadrés par des lois provinciales et une loi fédérale, établissant des exigences claires pour leur fonctionnement.
- Le Québec dispose de sa propre législation sur les régimes complémentaires de retraite, tandis que les autres provinces et le territoire fédéral suivent leurs propres réglementations ou la loi fédérale sur les normes de prestations de pension.
- Les provinces, à l'exception du Québec, sont souvent référencées par leurs sigles postaux dans le contexte des réglementations des régimes de retraite (ex: BC pour la Colombie-Britannique, ON pour l'Ontario, etc.).
- Ces lois visent à assurer une certaine uniformité et protection dans la gestion des régimes de retraite, bien que des variations spécifiques à chaque juridiction soient à noter.


### 5.5.1 Admissibilité

#### Loi sur les régimes complémentaires de retraite (Québec)
Les employés à temps plein ou à temps partiel qui, dans l’année civile précédente, ont gagné au moins 35% du MGA ou ont travaillé au moins 700 heures, doivent être admissibles au régime. L’adhésion peut cependant être facultative pour les employés à temps partiel.

#### Loi des autres provinces
Les employés à temps plein qui comptent 2 années de service continu, ainsi que les employés à temps partiel qui comptent 2 années de service continu et qui ont, dans chacune des deux années civiles précédentes, gagné au moins 35% du MGA ou travaillé au moins 700 heures, doivent être admissibles au régime

#### Loi fédérale sur les normes de prestations de pension
Comme pour les autres provinces, mais sans faire référence à 700 heures de travail dans chacune des deux années précédant l’adhésion.

### Commentaire
La loi du Québec est plus généreuse. De façon pratique, les employeurs ont tendance à faire adhérer leurs employés dès qu’ils deviennent employés permanents.


### 5.5.2 Acquisition et immobilisation

#### Loi sur les régimes complémentaires de retraite (Québec)

- **Les prestations de retraite sont immédiatement acquises** par l'employé dès leur constitution.
- Ces prestations **appartiennent entièrement à l'employé** dès le début de sa participation au régime.
- Elles sont **immobilisées**, interdisant leur encaissement avant la retraite.
- Leur unique objectif est de **fournir un revenu de retraite** à l'employé.


#### Loi des autres provinces
- **Acquisition immédiate des prestations de retraite** dans les grandes provinces (BC, AB, MB, ON, NS) et au Québec.
- Dans les autres provinces, **l'acquisition requiert deux années de participation**.
- **L'immobilisation des prestations** survient dès l'acquisition, à l'exception de l'AB et BC où elle est conditionnée à un salaire dépassant **20% du MGA**.


#### Loi fédérale sur les normes de prestations de pension
L’immobilisation se fait après deux années de participation continue au régime.

### Commentaire
- **Avant la réforme des pensions**, les prestations de retraite n'étaient acquises qu'après **45 ans d'âge et 10 années de service** ou de participation au régime, ce qui était considéré comme abusif.
- Les **lois actuelles ont modifié** cette règle pour corriger le problème d'acquisition tardive des droits.
- L'**ancien système présentait une grande injustice** : la participation à un régime de retraite limitait le taux de cotisation au REER de l'employé.
- Si l'employé quittait son emploi avant de remplir la condition de 45/10, **il perdait son espace de cotisation au REER** sans récupérer ses droits.

La condition de 45/10, mentionnée dans le contexte des régimes de retraite avant certaines réformes, se référait à la nécessité pour un employé d'atteindre à la fois l'âge de 45 ans et 10 années de service ou de participation au régime de retraite pour que ses prestations soient acquises et donc garanties. Cette règle exigeait donc qu'un employé ait un certain âge et une durée minimale de participation au régime pour bénéficier pleinement de ses avantages de retraite, ce qui a été jugé abusif et a été modifié par des réformes législatives ultérieures pour permettre une acquisition plus équitable des droits de retraite.

## 5.5.3 Cotisation patronale minimale

### Loi sur les régimes complémentaires de retraite (Québec)
- **Fin de la participation active au régime** : La **cotisation patronale** doit couvrir **au moins 50% de la valeur actualisée** de la rente accumulée **depuis le 1er janvier 1990**.
- **Dispositions particulières** pour les participants contribuant au **financement d'un déficit actuariel** via des **cotisations d'équilibre**.


### Loi des autres provinces
- Dans **toutes les provinces**, y compris au Québec, la **cotisation patronale** doit être **au moins 50% de la valeur actualisée** de la rente, à partir d'une **date spécifique à chaque province**.
- **New Brunswick (NB)** autorise une **cotisation patronale inférieure** si cela est stipulé dans le régime.
- **Île-du-Prince-Édouard (PE)** accepte un **pourcentage inférieur à 50%** pour les rentes indexées à un taux **au moins égal à 75% de l'IPC – 1%**.


### Loi fédérale sur les normes de prestations de pension
La règle est la même qu’au Québec et dans les autres provinces, sauf dans le cas où la rente est indexée à un taux égal ou supérieur à 75% x IPC – 1%.

### Commentaire
- Cette mesure corrige une faiblesse des **régimes à prestations déterminées**, où les cotisations des employés, basées sur un pourcentage uniforme du salaire, pouvaient couvrir plus que le coût réel des prestations pour les employés plus jeunes.
- Avant cette réforme, un employé partant à 35 ans après 10 ans de participation se voyait souvent rembourser seulement la valeur de ses cotisations, qui dépassait la valeur de la rente accumulée.
- Désormais, la mesure garantit que l'employeur contribue **au moins 50%** aux prestations accordées par le régime pour **chaque employé**, assurant une répartition plus équitable du financement des prestations.


## 5.5.4 Cessation d’emploi : traitement des cotisations des employés, lorsqu’elles dépassent 50% du coût du régime

- Lorsqu'un employeur doit couvrir **au moins 50% du coût du régime**, il peut y avoir des **cotisations excédentaires** de la part des employés.
- Cette situation est courante quand un employé jeune quitte l'entreprise ayant un régime à prestations déterminées, car le taux de cotisation uniforme pour tous est souvent supérieur à ce qui est requis pour couvrir la moitié de la rente d'un jeune employé.
- Pour calculer la cotisation excédentaire, on évalue d'abord la **valeur présente (VP) de la rente différée**, on la divise par deux, puis on soustrait ce montant des cotisations accumulées de l'employé, avec intérêt. L'équation est la suivante :

  Cot. excédentaires = $\sum$ Cotisations de l’employé avec intérêt – $\frac{1}{2}$ x VP (rente)


### Loi sur les régimes complémentaires de retraite (Québec)
- Les **cotisations excédentaires** peuvent être utilisées de plusieurs manières :
  - Elles peuvent **augmenter la rente existante** ou **constituer une rente additionnelle**.
  - Alternativement, elles peuvent être **transférées à un autre régime de retraite**, à un **CRI (Compte de retraite immobilisé)**, ou dans le cas d'un employé éligible à la retraite, à un **FRV (Fonds de revenu viager)**, qui est l'équivalent d'un **FERR (Fonds enregistré de revenu de retraite)**.


### Loi des autres provinces
Comme le Québec, la plupart des provinces permettent de transférer les cotisations excédentaires à un CRI ou à un FRV. Les provinces de l’Ouest permettent aussi de s’en servir pour augmenter la rente, comme au Québec. Contrairement au Québec, la plupart des provinces permettent aussi le remboursement des cotisations excédentaires.

### Loi fédérale sur les normes de prestations de pension
La règle est la même qu’au Québec.

## 5.5.5 Cessation d’emploi : Remboursement dans des situations spéciales
Il peut être permis de rembourser à l’employé la valeur des prestations gagnées si elles sont minimes ou s’il est probable que l’employé ne survivra pas jusqu’à sa retraite.

### Loi sur les régimes complémentaires de retraite (Québec)
- Un régime de retraite doit autoriser le **remboursement de la valeur des prestations** si celle-ci est **inférieure à 20% du MGA** (Maximum des Gains Admissibles).
- Il peut également prévoir le remboursement en cas d'**invalidité du participant**, susceptible de diminuer significativement son espérance de vie.
- Un participant qui **quitte son emploi et réside hors du Canada pendant au moins deux ans** a également droit au **remboursement de la valeur de ses prestations**.


### Loi des autres provinces
- La **majorité des provinces**, y compris le Québec, permettent le **remboursement de la valeur des prestations** si celle-ci est **inférieure à 20% du MGA** ou si la rente est **inférieure à 4% du MGA**, bien que cette règle ne soit pas universellement obligatoire.
- **New Brunswick (NB)** et **Île-du-Prince-Édouard (PEI)** appliquent des règles légèrement différentes, tandis que **Terre-Neuve-et-Labrador (NL)** utilise un seuil de **10% du MGA** pour le remboursement.
- Concernant les participants atteints d'une **invalidité réduisant l'espérance de vie**, la plupart des provinces envisagent le remboursement de la valeur des prestations, avec des variations entre l'obligation et la permission selon la juridiction.


### Loi fédérale sur les normes de prestations de pension
Les dispositions sont les mêmes que celles de la loi du Québec.

### Commentaire
Si on supposait que le facteur de 9 utilisé dans le calcul du FE était exact (on sait qu’il ne l’est pas vraiment), une valeur de 20% du MGA correspondrait à une rente égale à 20% / 9 = 2,22% du MGA. On parlerait alors d’une rente d’environ 1 369 \\$ / année ou environ 100 \\$ / mois. Dans le cas des provinces qui établissent la limite à 4% du MGA, il s’agit d’une rente de l’ordre de 2 464 \$ par année ou environ 205 \\$ / mois.

## 5.5.6 Droit de transfert à la cessation d’emploi

### Loi sur les régimes complémentaires de retraite (Québec)
- Régime à prestations déterminées : le transfert est permis jusqu’à 10 ans avant la date normale de retraite.
- Régime à cotisations déterminées : le transfert est permis à tout âge.

Dans tous les cas, le transfert doit se faire à un autre régime de retraite, à un CRI, à un FRV ou à un assureur (achat d’une rente viagère différée).

### Loi des autres provinces
- Dans **la plupart des provinces**, le **transfert** des valeurs de régime de retraite est **autorisé jusqu'à 10 ans avant la date normale de retraite**, sans distinction notable entre les régimes à **prestations déterminées** et ceux à **cotisations déterminées**.
- **Alberta (AB)** et **Colombie-Britannique (BC)**, similaires au Québec, **autorisent le transfert à tout âge** pour les régimes à cotisations déterminées.
- Toutes les provinces offrent la possibilité d'**acheter une rente viagère différée** auprès d'un assureur et autorisent le **transfert à un Compte de retraite immobilisé (CRI)** ou à un **Fonds de revenu viager (FRV)**.


### Loi fédérale sur les normes de prestations de pension
- Selon la **loi fédérale sur les normes de prestations de pension**, un régime doit autoriser le **transfert des fonds** au moins **jusqu'à 10 ans avant la date normale de retraite**, avec la possibilité d'étendre ce délai à moins de 10 ans.
- Le transfert peut se faire **vers un autre régime de retraite**, **un Compte de retraite immobilisé (CRI)**, **un Fonds de revenu viager (FRV)**, ou **pour l'achat d'une rente viagère différée auprès d'un assureur**, similairement à ce qui est prévu au Québec.


### Commentaire
- L'interdiction de transférer la valeur des prestations d'un régime à **prestations déterminées** (et parfois à **cotisations déterminées**) dans les **10 ans précédant la retraite** trouve son origine à une époque de **taux d'intérêt plus élevés**.
- Cette règle a été conçue lorsque les investisseurs institutionnels, surnommés « **zinzins** », bénéficiaient de **rendements supérieurs** à ceux des particuliers.
- En conséquence, la **valeur présente des prestations et des cotisations** connaissait une augmentation significative durant les **10 dernières années avant la retraite**.


## 5.5.7 Date normale de retraite

### Loi sur les régimes complémentaires de retraite (Québec)
La date normale de retraite ne peut pas dépasser le 1er jour du mois suivant celui au cours duquel le participant atteint l’âge de 65 ans.

### Loi des autres provinces
Le Manitoba et Terre-Neuve et Labrador ont une exigence semblable à celle du Québec. Parmi les autres provinces, on relève deux approches : les trois provinces les plus à l’ouest laissent le régime de retraite déterminer l’âge normal de retraite. L’Ontario et les Maritimes (sauf NL) permettent de fixer la date normale de retraite au plus tard 1 an après le 65ème anniversaire de naissance.

### Loi fédérale sur les normes de prestations de pension
L’âge normal de retraite est le plus jeune âge auquel le participant a le plein droit de recevoir sa pleine rente.

### Commentaire
Le Québec est la province où l’âge normal de retraite maximum qu’un régime peut utiliser est le moins élevé. Ceci est cohérent avec le fait que les Québécois ont tendance à prendre leur retraite en étant plus jeunes que les autres Canadiens.

## 5.5.8 Anticipation de la retraite, avec rente réduite

### Loi sur les régimes complémentaires de retraite (Québec)
Au Québec, un participant peut prendre sa retraite dans les 10 ans précédant la date normale de retraite. La retraite progressive, où un participant touche une rente réduite tout en travaillant à temps partiel, est permise.

### Loi des autres provinces
Toutes les provinces permettent au participant de prendre sa retraite dans les 10 ans précédant la date normale de retraite, comme au Québec.


### Loi fédérale sur les normes de prestations de pension
La loi permet au participant de prendre sa retraite dans les 10 ans précédant la date normale de retraite, comme au Québec.

### Commentaire
Cette règle universelle au Canada, combinée au fait que plusieurs régimes de retraite ont 65 ans comme âge normal de retraite, explique pourquoi l’âge de 55 ans (soit 65 ans – 10 ans) est vu comme étant l’âge où la retraite anticipée devient possible.


## 5.5.9 Report de la retraite, avec rente majorée

### Loi sur les régimes complémentaires de retraite (Québec)

- Au Québec, pour un régime à **prestations déterminées**, la rente doit être **revalorisée actuariellement** pour refléter la **date réelle de retraite**.
- Si la retraite est reportée après la date normale, les cotisations versées durant cette période doivent engendrer une **rente additionnelle**.
- Dans un régime à **cotisations déterminées**, les cotisations **continuent d'accumuler** sans ajustement spécifique lié au report de la retraite.


### Loi des autres provinces
La participation au régime peut généralement continuer, sauf si le participant touche la rente. (BC et AB permettent de toucher la rente tout en continuant de participer au régime).

### Loi fédérale sur les normes de prestations de pension
La participation au régime doit continuer, sauf si le participant touche la rente.

### Commentaire
La loi du Québec est la plus explicite, bien que la loi des autres provinces sous-entend que le report de la retraite permet de recevoir une rente accrue.

## 5.5.10 Définition de conjoint

### Loi sur les régimes complémentaires de retraite (Québec)
- Au Québec, pour être reconnu comme **conjoint** d'un participant d'un régime de retraite, il faut remplir l'un des critères suivants :
  - **Être marié** avec le participant.
  - **Être uni civilement** avec le participant, une forme d'union propre au Québec qui peut être dissoute par un notaire.
  - **Vivre en union libre** avec le participant depuis **au moins 3 ans**, ou **un an** si un enfant est né ou va naître de cette union.


### Loi des autres provinces
La loi des autres provinces ressemble beaucoup à celle du Québec : le conjoint doit être un conjoint marié ou conjoint de fait. Dans ce cas, la période minimale de vie commune varie de 1 à 3 ans, selon la province.

### Loi fédérale sur les normes de prestations de pension
La loi fédérale ressemble beaucoup à celle du Québec : le conjoint doit être un conjoint marié ou conjoint de fait. Dans ce cas, la période minimale de vie commune est de 1 an.

### Commentaire
Seul le Québec mentionne l’union civile dans sa loi.

## 5.5.11 Décès avant la retraite

Les lois des provinces à l'est du Manitoba (sauf le NB) font une distinction entre les droits gagnés avant la réforme des pensions (1987, 1988 ou 1989, selon la province) et ceux gagnés depuis. La prestation de décès est plus généreuse relativement aux droits gagnés depuis la réforme.

### Loi sur les régimes complémentaires de retraite (Québec)
- Si un participant **décède avant la retraite**, la valeur des prestations acquises **après 1989** est **versée au conjoint survivant**, ou, à défaut, au **bénéficiaire désigné** ou à la **succession**.
- La valeur de la rente **n'est pas immobilisée** dans le régime à ce stade.


### Loi des autres provinces
- Dans toutes les provinces, y compris le Québec, en cas de décès du participant, la valeur des prestations acquises est **versée au conjoint survivant**, ou, en l'absence de celui-ci, au **bénéficiaire désigné** ou à la **succession**.
- En **Nouvelle-Écosse (NS)** et à l'**Île-du-Prince-Édouard (PE)**, si aucun conjoint survivant n'est présent, on rembourse la **valeur des cotisations du participant avec intérêts**.
- Au **New Brunswick (NB)**, le remboursement des cotisations avec intérêts n'advient que si les prestations **ne sont pas acquises**.
- En **Colombie-Britannique (BC)**, **Alberta (AB)**, et **Manitoba (MB)**, si un conjoint survivant existe, la valeur de la rente doit être **immobilisée** dans le régime.


### Loi fédérale sur les normes de prestations de pension
- Selon la **loi fédérale sur les normes de prestations de pension**, la valeur des prestations acquises est **versée au bénéficiaire désigné** ou, à défaut, à la **succession** en l'absence de conjoint survivant.
- Si un **conjoint survit**, il a droit à la valeur des prestations acquises, évaluée comme si le participant avait **cessé son emploi à la date de décès**.
- Le régime peut également offrir l'option d'une **rente immédiate** au conjoint survivant.


## 5.5.12 Décès après la retraite

Les lois de toutes les provinces, incluant le Québec, ainsi que la loi fédérale sur les normes de prestations de pension, exigent qu'au moins 60% de la rente continue à être versée au conjoint survivant.

## 5.5.13 Coordination avec les prestations du RRQ et la PSV (régime à prestations déterminées)

Dans les régimes à prestations déterminées, il existe deux façons de tenir compte des prestations des régimes d’État :
- Soustraction d'une partie de la rente d’État de la rente du régime privé.
- Modulation de la formule de rente pour créditer un taux plus élevé sur le salaire excédant la MGA que sur le salaire jusqu'à concurrence du MGA.

### Loi sur les régimes complémentaires de retraite (Québec)
On peut soustraire 1/35 de la PSV par année de service avant 1990. Pour la rente du RRQ/RPC, on peut toujours la soustraire à raison de 1/35 par année de service.

### Loi des autres provinces
Il est interdit de réduire la rente de retraite d’un montant basé sur la PSV, pour le service après la réforme des pensions. On peut toujours soustraire la rente du RRQ/RPC à raison de 1/35 par année de service.

### Loi fédérale sur les normes de prestations de pension
Il est permis de soustraire non seulement la prestation du RRQ/RPC mais aussi celle de la PSV.

## 5.5.14 Discrimination selon le sexe

La discrimination selon le sexe peut être obligatoire ou interdite :
- Obligatoire : utilisation de tables de mortalité différentes pour hommes et femmes.
- Interdite : utilisation de tables de mortalité unisexes.

### Loi sur les régimes complémentaires de retraite (Québec)
Au Québec, la discrimination est obligatoire.

### Loi des autres provinces
La discrimination est généralement interdite, sauf à NL, où elle est permise sans restriction, et en Ontario et dans les Maritimes pour les rentes gagnées avant la réforme des pensions.

### Loi fédérale sur les normes de prestations de pension
La discrimination est permise pour les années de service avant 1987 et interdite pour les années de service après 1986.

### Commentaire
- La tendance actuelle est à l'**interdiction de la discrimination selon le sexe**, bien que le Québec fasse exception à cette règle.
- L'interdiction se justifie par le fait que, malgré une **espérance de vie moyenne plus élevée chez les femmes**, cela ne présume pas de la longévité d'une femme par rapport à un homme en particulier.
- L'objectif est de garantir **l'égalité des sexes** en matière de **rente** ou de **valeur de transfert**.


## 5.5.15 En cas de dissolution du mariage ou de l’union conjugale

Lorsqu’une union est dissoute, les droits découlant du régime de retraite sont partageables entre les deux conjoints, selon des modalités particulières.

### Loi sur les régimes complémentaires de retraite (Québec)
- En cas de **séparation**, **divorce**, **nullité du mariage**, ou **dissolution** d'union civile au Québec, les droits de retraite accumulés par le participant peuvent être **partagés** avec son conjoint.
- Le partage est effectué sur **demande écrite** adressée au comité de retraite et suit les dispositions du **Code civil**, d'un **jugement** ou d'une **déclaration commune notariée**.
- Seuls les **conjoints mariés** ou **unis civilement** ont droit à une part des droits accumulés.
- La pratique courante est un partage **fifty-fifty des droits accumulés durant l'union**.


### Loi des autres provinces
- Dans les **autres provinces**, le partage des droits de retraite en cas de séparation conjugale est régi par les lois sur le **droit de la famille** et les **biens familiaux**.
- La législation provinciale spécifie que le participant ne peut céder **plus de 50%** de ses droits de retraite à son ex-conjoint.
- Certaines provinces exigent que le partage des droits d'une rente d'un régime à **prestations déterminées** n'intervienne qu'à la **date normale de retraite** ou lorsque le versement de la rente **débute**.


### Loi fédérale sur les normes de prestations de pension
Le partage des droits se fait selon la loi provinciale ou selon une entente de cession. Toutefois, on précise que la somme des valeurs résultant du partage ne peut pas excéder la valeur des droits s’il n’y avait pas eu dissolution de l’union.

## 5.5.16 Taux d’intérêt minimum sur les cotisations salariales

Cet élément est important dans les régimes à cotisations déterminées car il détermine directement la valeur des fonds accumulés au nom de l’employé.

### Loi sur les régimes complémentaires de retraite (Québec)
- Au Québec, pour un régime de retraite **non assuré**, le **taux d'intérêt crédité** correspond au **taux de rendement net** de la caisse de retraite ou du compte du participant, si les cotisations sont investies selon ses instructions.
- Pour un régime **assuré**, le taux crédité est celui du **rendement net des placements généraux de l'assureur** ou du taux des **dépôts à terme personnels de 5 ans**.


Dans le contexte des régimes de retraite, un régime "assuré" fait référence à un régime de retraite dont les prestations ou les actifs sont garantis par une compagnie d'assurance. Cela signifie que la gestion des fonds, la couverture des risques liés à l'investissement, et parfois le paiement des rentes sont sous la responsabilité de l'assureur. 

### Loi des autres provinces
- Dans les **provinces autres que le Québec**, pour les **régimes assurés**, le **taux d'intérêt crédité** doit être au moins équivalent à celui des **dépôts bancaires à terme de 5 ans**.
- Pour les **régimes non assurés à cotisations déterminées**, le taux crédité correspond au **rendement de la caisse de retraite**.
- Dans un **régime à prestations déterminées non assuré**, le taux d'intérêt crédité doit être au moins égal au **taux de rendement des dépôts bancaires**.


### Loi fédérale sur les normes de prestations de pension
Pour un régime à cotisations déterminées, on crédite le taux de rendement de la caisse de retraite. Pour un régime à prestations déterminées, on crédite le rendement de la caisse de retraite ou celui des dépôts à terme de 5 ans.

## 5.5.17 Fréquence minimale des évaluations actuarielles

Les évaluations actuarielles ne s’appliquent qu’aux régimes à prestations déterminées.

### Loi sur les régimes complémentaires de retraite (Québec)
Au Québec, l’évaluation actuarielle doit être faite à chaque année.

### Loi des autres provinces
Dans les autres provinces, les évaluations actuarielles sont habituellement requises tous les 3 ans. Cependant, dans certaines provinces (ON, NB, NS), une évaluation annuelle est requise si le ratio de solvabilité est inférieur à 85% ou 90 %.

### Loi fédérale sur les normes de prestations de pension
Les évaluations actuarielles sont requises tous les 3 ans. Cependant, une évaluation annuelle est requise si le ratio de solvabilité est inférieur à 120 %.

## 5.5.18 Insaisissabilité

De façon générale, dans toutes les juridictions, les cotisations à un régime de retraite et les prestations qui en découlent, ainsi que les valeurs de transfert, sont insaisissables par les créanciers de l’employeur ou du participant.

## 5.6 Jurisprudence

La jurisprudence en matière de régime de retraite est très importante car elle clarifie des éléments qui ne sont pas toujours précisés dans les textes de loi.

### 5.6.1 Dayco (Canada) Ltd c. CAW Canada – Jugement de 1993 (Cour Suprême)
Ce jugement établit le principe selon lequel les droits d’un participant sont gelés au moment de la prise de sa retraite. Toute modification ayant pour effet de réduire les droits des retraités est illégale et sans valeur.

### 5.6.2 Wronko c. Western Inventory Service Ltd – Jugement de 1994 (Cour d’Appel de l’Ontario)
Ce jugement établit le principe selon lequel l’employeur ne peut pas modifier unilatéralement les conditions d’emploi d’un employé. Si ce dernier refuse une telle modification et que l’employeur le laisse continuer à travailler, alors son contrat d’emploi et les conditions qui s’y rattachent demeurent inchangés. Par conséquent, un employeur ne peut pas modifier un régime de retraite ou d’autres avantages sociaux sans l’accord des employés.

### 5.6.3 Nolan c. Kerry Canada Inc. – Jugement de 1994 (Cour Suprême)
Ce jugement confirme qu’en cas de surplus dans un régime de retraite, l’employeur peut prendre un congé de cotisations, à moins que le règlement du régime ne l’interdise expressément. De plus, lorsqu’un régime de retraite comporte un volet à prestations déterminées et un volet à cotisations déterminées, l’employeur peut utiliser le surplus du volet à prestations déterminées pour acquitter sa cotisation au volet à cotisations déterminées.

### 5.6.4 Lola c. Éric – Jugement de 2013 (Cour Suprême)
Ce jugement confirme que le partage des intérêts financiers d’un couple s’applique seulement aux conjoints mariés ou unis civilement, mais pas aux conjoints de fait. Ce faisant, la Cour Suprême reconnaît le principe selon lequel l’union libre n’est pas régie par la Loi au Québec, sauf lorsqu’elle est expressément mentionnée dans une loi.

### 5.6.5 Association des retraités d’Hydro-Québec c. Hydro-Québec – Jugement de 2005 (Cour d’appel du Québec)
Ce jugement établit que les retraités n’ont pas le droit d’être consultés sur l’utilisation d’un surplus dans un régime de retraite en cours d’existence. L’employeur peut suspendre ses cotisations au régime en cas de surplus, si le régime ne l’interdit pas, et qu’il n’a pas d’obligation d’équité envers les retraités. La Cour confirme aussi que les droits des participants sont cristallisés au moment de la retraite.

### 5.6.6 Monsanto c. Superintendent of Financial Services (Ontario) – Jugement de 2004 (Cour Suprême)
Ce jugement établit qu’en cas de terminaison partielle d’un régime de retraite, tout surplus attribuable aux participants affectés par la terminaison partielle doit être distribué à la date de la terminaison partielle.

### 5.6.7 Sun Indalex Finance, LLC v. United Steelworkers – Jugement de 2013 (Cour Suprême)
Ce jugement établit qu’en cas de faillite de l’employeur, la créance du régime de retraite en situation de déficit actuariel n’est pas prioritaire par rapport à celle du prêteur qui finance les opérations de l’entreprise pendant sa liquidation, car la loi fédérale sur les arrangements avec les créanciers des compagnies a la priorité sur une loi provinciale telle que le Pension Benefits Act de l’Ontario.
