# Module 5 : Contexte juridique

## Objectifs du Module 5

- Connaître la base du droit civil au Canada et les fondements des systèmes juridique québécois et canadien
- Comprendre les principales notions qui gouvernent les contrats au Québec et qui s’appliquent aux contrats d’assurance
- Connaître les principales dispositions des contrats d’assurance individuelle de personnes
- Connaître certaines dispositions particulières aux contrats d’assurance individuelle
- Connaître les principaux défis liés à la gestion des bénéficiaires des contrats d’assurance

## 5.1 Les bases du Droit au Canada et au Québec

- Le domaine du Droit est divisé en deux grands volets : le droit civil et le droit criminel, avec l'assurance relevant du droit civil.

- Au Canada, il existe deux systèmes juridiques dans le droit civil : le Code civil du Québec au Québec et la Common Law dans les autres provinces.

- La coexistence de ces deux systèmes a une origine historique. Avant la conquête britannique, la Nouvelle-France utilisait le droit civil français, la "Coutume de Paris", tandis que le Royaume-Uni et ses colonies suivaient un droit "coutumier" basé sur le droit normand.

- Après 11 ans d'essai infructueux d'implanter le droit coutumier au Québec, le Parlement britannique a rétabli le droit civil français pour la "Province of Quebec" avec l'Acte de Québec de 1774.

- L'évolution du Canada a permis de conserver un droit civil codifié au Québec et d'adopter la Common Law dans les provinces anglophones, conforme aux traditions des colons de l'Empire Britannique.

- Au Québec, le passage d'un droit civil français médiéval à un système moderne a été complexe, aboutissant à la création d'un Code civil en 1857, inspiré du Code Napoléon de 1804, et adopté en 1866 sous le nom de "Code civil du Bas-Canada", remplacé en 1994 par le Code civil du Québec.

- Dans les autres provinces anglophones, la Common Law a été appliquée dès leur constitution en province ou territoire.

## 5.2 Fondement du système juridique

- Les systèmes juridiques sont fondés soit sur un droit codifié, comme le code civil, soit sur un droit coutumier, qui se base sur des jugements pertinents formant la jurisprudence.
- Au Canada, le droit civil est régi par le Code civil du Québec pour la province de Québec et par la Common Law dans les autres provinces.
- Pour plus d'informations, consulter [Justice Canada](https://www.justice.gc.ca/fra/sjc-csj/just/03.html).


### 5.2.1 Code civil

- Le code civil se présente comme un recueil exhaustif de règles, articulées autour de grands principes généraux, destinés à trancher les litiges.

- Dans les juridictions de droit civil, à l'instar du Québec, les tribunaux consultent en premier lieu le Code civil, enrichissant leur interprétation par la jurisprudence existante pour garantir la cohérence des décisions.

- Ce cadre juridique est dynamique, évoluant par des révisions périodiques pour refléter les changements sociétaux et législatifs, illustré par la refonte majeure du Code civil du Québec en 1994 et ses mises à jour subséquentes, comme l'introduction de l'union civile.

- Le Code civil, en synergie avec la Charte des droits et libertés de la personne, orchestre la régulation des individus, de leurs interactions et de leur rapport aux biens, posant les fondations du droit commun au Québec.

- Il agit comme la pierre angulaire sur laquelle se construisent les autres textes législatifs et réglementaires, qui peuvent y apporter des précisions ou introduire des exceptions spécifiques.

### 5.2.2 Common Law

- La Common Law repose sur un droit non codifié, émergeant d'un ensemble de règles établies à partir de précédents judiciaires, orientant les décisions des juges dans des affaires similaires.

- Contrairement au droit civil, la Common Law ne se trouve pas dans un "code" ou un ensemble législatif formalisé, mais elle est compilée dans des textes de doctrine et vit à travers les principes dérivés de la jurisprudence.

- Cette nature principalement jurisprudentielle de la Common Law lui confère une flexibilité lui permettant de s'ajuster aux évolutions sociétales, autorisant les juges à forger de nouvelles doctrines ou à modifier les existantes selon le besoin.

- Les racines de la Common Law remontent au droit coutumier normand, établi suite à la conquête de l'Angleterre par Guillaume de Normandie en 1066, et sa formalisation en tant que système distinct s'est produite au XIIème siècle.

- Les jugements de la Common Law, potentiellement remontant à 1189, constituent une tradition juridique suivie au Royaume-Uni, aux États-Unis (à l'exception de la Louisiane), et dans les pays du Commonwealth, y compris le Canada, reflétant sa portée et son influence historiques étendues.

## 5.2.3 Fonctionnement du système juridique

- Les fondements juridiques des systèmes de droit diffèrent : le Code civil constitue la base du droit dans les juridictions de droit civil, tandis que la Common Law remplit ce rôle dans les systèmes de Common Law.

- Par-dessus cette base, s'ajoutent les lois promulguées par les parlements, enrichissant et précisant le cadre juridique initial.

- La jurisprudence, bien qu'elle n'ait pas la même prépondérance dans un système de Code civil que dans celui de la Common Law, joue un rôle crucial dans les deux, en éclairant l'interprétation des lois et du Code civil.

- Dans un système de Code civil, un jugement significatif peut inciter le gouvernement à réviser la législation pour clarifier ou réfuter une interprétation donnée par les tribunaux.

## 5.2.4 Exemple de fonctionnement du système juridique

Quand on conçoit un produit d’assurance, le cadre juridique est important :

- En Common Law, si on a une idée, qu’elle a du sens et qu’elle n’est pas contraire aux principes généraux découlant de la Common Law, on peut concevoir le produit et le commercialiser.

- Sous le Code civil, si on a une idée et qu’elle a du sens, on doit d’abord vérifier s’il y a une ou plusieurs restrictions ou si le produit visé est nul ou prohibé. Si ce n’est pas le cas, on peut concevoir le produit et le commercialiser.

## 5.3 Nature des contrats selon le Code civil du Québec

- Un contrat est un accord de volontés créant des obligations pour une ou plusieurs personnes envers d'autres.
- La bonne foi doit diriger les actions des parties impliquées, de l'établissement de l'obligation à son accomplissement ou sa résolution.


### 5.3.1 Contrat de gré à gré et contrat d’adhésion

- Un contrat d'adhésion se caractérise par des conditions imposées ou rédigées par une partie, sans possibilité de négociation libre pour l'autre.
- Conformément à l'article 1432 du Code civil, en cas de doute ou de litige, le contrat d'adhésion est interprété en faveur de la partie qui n'a pas rédigé le contrat.
- Tout contrat qui permet une négociation libre des termes par les deux parties est qualifié de contrat de gré à gré, par opposition au contrat d'adhésion.


### 5.3.2 Contrat unilatéral ou bilatéral

- Un contrat unilatéral est un engagement où une seule partie a des obligations envers l'autre, qui elle, n'a pas d'obligation.
- Un contrat bilatéral implique des obligations mutuelles entre les parties, chaque obligation étant conditionnée par celle de l'autre partie.

### 5.3.3 Contrat à titre onéreux ou à titre gratuit

- Un contrat à titre onéreux implique un avantage pour chaque partie en échange de son obligation.
- Un contrat à titre gratuit se caractérise par l'obligation d'une partie envers l'autre, sans recevoir d'avantage en retour.

### 5.3.4 Contrat commutatif ou aléatoire

- Un contrat est considéré comme commutatif lorsque, au moment de sa conclusion, les obligations des parties et les avantages qu'elles en retirent sont certains et déterminés.
- Un contrat est qualifié d'aléatoire lorsque l'ampleur des obligations ou des avantages demeure incertaine.


### 5.3.5 Contrat à exécution instantanée ou à exécution successive

- Un contrat est considéré comme à exécution instantanée lorsque les obligations des parties sont exécutées en une seule fois.
- Un contrat est qualifié d'à exécution successive lorsque les obligations sont exécutées en plusieurs étapes ou de manière continue.


### 5.3.6 Notion de nullité

#### Qu'est-ce que la nullité d'un contrat ?

- La nullité d'un contrat signifie que ce contrat est considéré comme s'il n'avait jamais existé.
- Lorsqu'un contrat est nul, chaque partie doit restituer à l'autre ce qu'elle a reçu en vertu de ce contrat.

#### Types de Nullité :

1. **Nullité Absolue** :
   - Elle concerne des contrats dont la condition de formation s'impose pour la protection de l'intérêt général.
   - Toute personne ayant un intérêt au contrat peut soulever la nullité.
   - Le tribunal peut également la soulever d'office.

2. **Nullité Relative** :
   - Elle s'applique aux contrats où la condition de formation est nécessaire pour protéger des intérêts particuliers.
   - Seules les personnes en faveur de qui elle est établie ou leur cocontractant (s'ils sont de bonne foi et subissent un préjudice sérieux) peuvent invoquer la nullité relative.
   - Le tribunal ne peut pas la soulever d'office.
   - Un contrat frappé de nullité relative peut être confirmé si toutes les parties sont d'accord.

#### Application de la Nullité :

- La nullité est particulièrement importante dans le contexte de l'**assurance funéraire**.

#### Assurance Funéraire :
- L'assurance funéraire est un contrat où une personne s'engage à fournir des services ou à payer des frais funéraires lors du décès d'une autre personne, en échange d'une prime.
- Ce type de contrat est sujet à la **nullité relative**.

#### Demande de Nullité :
- La nullité d'un contrat d'assurance funéraire ne peut être demandée que par ceux qui ont payé la prime ou par l’Autorité des marchés financiers agissant en leur nom.

#### Notions à retenir :
- La nullité signifie l'annulation complète d'un contrat.
- Elle peut être absolue (protégeant l'intérêt général) ou relative (protégeant des intérêts particuliers).
- L'assurance funéraire est un exemple de contrat sujet à la nullité relative.
- Seules certaines parties peuvent demander la nullité, et le tribunal ne peut pas la soulever d'office.
- Un contrat frappé de nullité relative peut être confirmé si toutes les parties sont d'accord.

## 5.4 Le contrat d’assurance individuelle de personne au Québec

Le contrat d’assurance individuelle de personne a des caractéristiques et des contraintes qui lui sont particulières.

### 5.4.1 Nature du contrat

C'est un contrat d'adhésion entre l’assureur, d’une part, et le preneur (appelé aussi titulaire ou détenteur).

- Seul l’assureur rédige le contrat.
- Le preneur peut accepter ou refuser le contrat, mais ne peut pas le modifier.
- Aucune négociation n’est possible.
- Tout ce qui est rédigé de façon imprécise, portant à confusion où à interprétation ambiguë, sera interprété en faveur du preneur par les tribunaux.

#### C’est un contrat aléatoire.

- La promesse de l’assureur est conditionnelle à ce qu’il se produise un évènement dont la réalisation n’est pas certaine :
  - Le risque doit se réaliser, ce qui est incertain.
  - Le contrat doit être en vigueur lorsque le risque se réalise.

#### C’est un contrat unilatéral.

- Seul l’assureur est tenu de respecter le contrat jusqu’à son échéance.
- Le titulaire peut cesser de payer la prime et mettre fin au contrat.
- L’assureur ne peut pas forcer le titulaire à payer ses primes ou à maintenir le contrat en vigueur.

### 5.4.2 Conditions de validité du contrat

Les conditions de validité du contrat sont régies par le Code civil, notamment dans la section qui traite des contrats en général (articles 1371 à 1456) et dans la section qui traite spécifiquement des assurances de personnes (articles 2389 à 2479.1).

#### 5.4.2.1 Conditions générales

Comme tous les contrats, le contrat d’assurance doit être basé sur la bonne foi. Le fait de cacher volontairement quelque chose à l’assureur est suffisant pour que le contrat soit nul. De plus, le contrat doit être écrit.

- Étant donné la période de temps qui peut s’écouler entre le moment de l’entrée en vigueur du contrat et le moment où l’assureur devra honorer son contrat, le contrat doit être appuyé sur un document écrit.

- Il est possible (même quasi certain dans le cas de l’assurance vie) qu’une des parties ne soit plus vivante lorsque le paiement devra être honoré.

#### 5.4.2.2 Conditions particulières à la validité d’un contrat d’assurance

Plusieurs conditions doivent être remplies pour que le contrat soit valide. Notamment :

- Il faut un accord entre les deux parties contractantes (assureur et preneur) :
  - L’assureur doit accepter le risque.
  - Le preneur doit accepter les conditions stipulées au contrat.

- Il faut avoir la capacité légale de contracter :
  - La personne doit être apte à contracter.
  - Un contractant mineur pose des problèmes :
    - Un mineur peut contracter seul pour satisfaire ses besoins ordinaires et usuels (article 157 du Code civil).
    - Mais l’article 163 du Code civil donne au mineur le droit d’annuler un contrat qu’il a conclu seul, s’il en subit un préjudice. Par conséquent, un mineur qui contracterait une assurance vie sur un prêt automobile, dont la prime serait démesurée, pourrait prétendre avoir subi un préjudice…

- Le consentement doit être donné légalement :
  - Il faut la manifestation expresse ou tacite de la volonté d’une personne d’accepter l’offre de contracter que lui fait une autre personne.
  - L’offre de contracter doit comporter tous les éléments essentiels du contrat envisagé.
  - Le contrat doit être conclu sans pression, menace ni dol (tromperie qui peut résulter du silence ou d’une réticence).

- Le contrat doit avoir un objet :
  - C’est-à-dire une raison d’être, donc un but.
  - L’objet ne doit pas être prohibé ni contraire à l’ordre public.
  - L’assureur assure la vie, l’intégrité physique ou la santé d’une personne.

- Le contrat doit avoir une cause :
  - La raison qui détermine chacune des parties à conclure le contrat.
  - Pour l’assureur, c’est la prime qu’il reçoit ou recevra.
  - Pour le preneur, c’est le capital décès ou les prestations qui seront versées du vivant de l’assuré.

### 5.4.3 Les intervenants au contrat

Bien que le contrat soit conclu seulement entre deux parties (l’assureur et le preneur), plusieurs autres parties peuvent y avoir des droits.

- Le payeur (c’est habituellement le preneur, mais ce peut être une autre personne comme l’assuré ou le bénéficiaire).
- L’assuré (pas toujours le preneur).
- Le bénéficiaire (habituellement différent du preneur, surtout si le preneur fait assurer sa propre vie).

##### Exemple :

Grand-papa Généreux a décidé de se procurer auprès de la Compagnie d’assurance ABC une police d’assurance sur la vie de sa petite-fille, Maude, à la naissance de celle-ci. Dix ans plus tard, il décide de transférer la propriété du contrat à son gendre, Guy, le père de Maude. La bénéficiaire de la police d’assurance est Catherine, la mère de Maude.

Après le transfert,
- Compagnie d’assurance : Compagnie d’assurance ABC,
- Titulaire de la police : Guy
- Preneur original : Grand-papa Généreux
- Assurée : Maude
- Bénéficiaire : Catherine

### 5.4.4 La possibilité de rachat du contrat

- Certaines polices d’assurance vie comportent un mécanisme d’accumulation d’argent pour stabiliser la prime avec l'âge de l'assuré.

- Au début de l'assurance, le preneur paie une prime plus élevée que nécessaire.

- L'excédent de prime devient une provision ou "réserve" accumulée avec intérêt, généralement au taux de rendement moyen de l'assureur.

- Lorsque le titulaire souhaite résilier le contrat, cette réserve doit être utilisée.

- Si l'assureur remet tout ou une partie de cette réserve au titulaire, on l'appelle "valeur de rachat".

- Il est important de noter que c'est l'assureur qui paie la valeur de rachat pour se libérer de son obligation envers l'assuré et le titulaire.

## 5.5 Les deux grandes catégories d’assurance individuelle de personnes

Les contrats d’assurance individuelle de personnes comportent deux catégories : 

 - L’assurance vie

 - L’assurance contre les accidents et la maladie.

### 5.5.1 L’assurance sur la vie

L’assurance sur la vie garantit le paiement de la somme convenue, au décès de l'assuré, si le contrat est toujours en vigueur.

### 5.5.2 L’assurance contre les accidents et la maladie

L’assurance contre les accidents et la maladie garantit le paiement d’une somme convenue, qui peut être :

- un remboursement total ou partiel de certains frais (frais médicaux ou dentaires ou soins de longue durée ou soins reçus en cas d’urgence lors d’un voyage à l’étranger) dans le cas de :
  - L’assurance maladie
  - L’assurance soins dentaires
  - L’assurance voyage

- Un montant périodique relié ou non à une perte de revenu causée par une invalidité, dans le cas de l’assurance invalidité.

- Un montant forfaitaire si l’assuré développe une certaine maladie (par exemple, le cancer) et y survit pendant un certain délai (habituellement 30 jours), dans le cas de l’assurance contre les maladies.

### 5.5.3 Les clauses accessoires

- Les clauses d’assurance contre la maladie ou les accidents accessoires à un contrat sur la vie sont régies par les dispositions du contrat d’assurance-vie.

- De même, les clauses d’assurance sur la vie accessoires à un contrat d’assurance contre la maladie ou les accidents sont régies par les dispositions du contrat d’assurance contre la maladie ou les accidents.

- En d'autres termes, les clauses accessoires sont régies par les dispositions du contrat principal.

## 5.6 Prise d’effet du contrat

Cette notion est importante car il y a une différence fondamentale entre la loi du Québec et celle du reste du Canada.

### 5.6.1 Prise d’effet du contrat d’assurance sur la vie

Au Québec, le contrat d’assurance-vie prend effet au moment de l’acceptation de la proposition par l’assureur pour autant que :

- Cette dernière ait été acceptée sans modification.

- La première prime ait été versée.

- Aucun changement ne soit intervenu dans le caractère assurable du risque depuis la signature de la proposition.

Dans les autres provinces, le contrat d’assurance-vie prend effet au moment de la livraison de la police au preneur.

### 5.6.2 Prise d’effet du contrat d’assurance contre la maladie ou les accidents

- Au Québec, le contrat d’assurance contre la maladie ou les accidents prend effet au moment de la livraison de la police au preneur par l’assureur, même si la livraison n’est pas faite par un représentant de l’assureur.

- Dans les autres provinces, le contrat d’assurance contre la maladie ou les accidents prend effet au moment de la livraison de la police au preneur, Donc pas de différence significative avec le Québec dans ce cas-ci.

## 5.7 Contenu du contrat (police d’assurance)

La police est le document qui constate l'existence du contrat d'assurance. La proposition d’assurance doit faire partie de la police si l’assureur veut pouvoir l’évoquer en cas de fraude.

Le code civil exige que le contrat contienne certaines informations. Essentiellement, il s’agit d’exigences dictées par le bon sens et sans lesquelles la police ne pourrait pas exister ou serait bancale. Ainsi, l’assureur doit indiquer dans la police d’assurance de personnes :

- Le nom des parties au contrat (assureur et preneur).
- Le nom de l’assuré (peut être différent du preneur; par exemple son enfant).
- Nom des personnes à qui les sommes assurées sont payables ou, si ces personnes sont indéterminées, le moyen de les identifier (c’est-à-dire le nom du ou des bénéficiaires).
- L’objet de l’assurance.
- Le montant de l’assurance (ou l’équivalent si l’assurance rembourse des frais médicaux ou dentaires).
- La nature des risques couverts (décès, invalidité, soins, etc.).
- Le moment à partir duquel les risques sont couverts et la durée de la couverture.
- Montant ou le taux des primes et les dates auxquelles celles-ci viennent à échéance.
- Délais de paiement des primes.
- Droits de participation aux bénéfices.
- Méthode et tableau devant servir à établir la valeur de rachat (s’il y a lieu).
- Droits à la valeur de rachat (s’il y a lieu).
- Droits aux avances sur police.
- Conditions de remise en vigueur.
- Droits de transformations de l’assurance.
- Modalités des sommes dues (comment et quand elles sont payables).
- Période durant laquelle les prestations sont payables.

De plus, dans le cas d’un contrat d’assurance contre la maladie et les accidents, l’assureur doit indiquer expressément et en caractères apparents la nature de la garantie. Lorsqu’il s’agit d’assurance invalidité, l’assureur doit aussi stipuler :

- Les conditions de paiement des indemnités.

- La nature et les caractéristiques de l’invalidité assurée, sinon l’invalidité sera considérée comme l’inaptitude de la personne assurée à exercer les tâches de son emploi habituel (par opposition à l’incapacité de faire tout genre de travail).

## 5.8 Intérêt d’assurance

- Selon le **Code civil du Québec**, un contrat d'assurance individuelle nécessite que le preneur ait un **intérêt d'assurance** dans la vie ou la santé de la personne assurée au moment de la conclusion du contrat.
- Le **consentement écrit** de la personne concernée est requis si le preneur n'a pas d'intérêt d'assurance direct.
- Un contrat d'assurance est **nul** en cas de cession sans que le cessionnaire n'ait un **intérêt d'assurance** à ce moment-là.
- La notion de **risque** est centrale : sans préjudice potentiel pour le preneur en cas de réalisation des événements couverts, il n'y a pas de base pour une assurance.


Une personne a un intérêt d’assurance dans la vie et la santé de :
- Sa propre personne.
- Son conjoint.
- Ses descendants.
- Les descendants de son conjoint.
- Les personnes qui contribuent à son soutien ou son éducation.
- Ses préposés ou son personnel ou les personnes dont la vie et la santé présentent pour elle un intérêt moral ou pécuniaire.

## 5.9 Obligations du preneur et de l’assuré

- Lors de la demande de preuves d'assurabilité, le preneur et l'assuré doivent déclarer à l'**assureur** les **faits importants**, c'est-à-dire toutes les circonstances connues influençant significativement une compagnie d'assurance dans:
  - L'**établissement de la prime**.
  - L'**appréciation du risque**.
  - La **décision d'accepter le risque**.

- L'obligation est considérée comme correctement exécutée si:
  - La déclaration correspond à celle d'un **assuré prévoyant**.
  - Elle est effectuée **sans réticences importantes**.
  - Les circonstances sont, en substance, **conformes à la déclaration**.

- **Sauf pour l'âge**, les **fausses déclarations** et **réticences** à révéler ces circonstances par le preneur et l'assuré entraînent, sur demande de l'assureur, la **nullité du contrat**, même pour les sinistres non liés au risque dénaturé.

## 5.10 Dispositions générales dans les contrats d’assurance

Les dispositions suivantes sont permises ou exigées par le Code civil et se retrouvent normalement dans les contrats d’assurance individuelle de personnes.

### 5.10.1 Règle d’incontestabilité

Cette règle encadre le droit de l’assureur de demander l’annulation du contrat.

- S’il n’y a pas fraude, la fausse déclaration ou la réticence sur le risque ne peut fonder la nullité ou la réduction de l’assurance qui a été en vigueur pendant 2 ans (sauf dans le cas de l’assurance invalidité, si l’invalidité a débuté durant les 2 premières années de l’assurance).

- Les déclarations inexactes concernant l’usage du tabac ou de drogues sont généralement considérées comme étant des fraudes par les compagnies d’assurance et permettent l’annulation au-delà des 2 premières années.

### 5.10.2 Affection déclarée dans la proposition d’assurance contre la maladie ou les accidents

- L'**assureur ne peut pas**, sauf en cas de **fraude**, exclure ou réduire la garantie en raison d'une affection non déclarée, **sauf si** cette affection est spécifiquement mentionnée dans une clause.

- Une **clause générale** ne peut pas être utilisée pour exclure ou limiter la garantie d'assurance pour une affection non déclarée, **à moins que**:
  - L'affection ne se manifeste **durant les deux premières années** de l'assurance.
  - Il y ait eu **fraude**.

### 5.10.3 Fausse déclaration sur l’âge de l’assuré

- Une déclaration d'**âge erroné** lors de la demande d'assurance entraîne la nullité de l'assurance **seulement si** :
  - L'âge déclaré est **en dehors des limites** fixées par les tarifs de l'assureur.
  - L'assureur demande la nullité du contrat **tant que l'assuré est vivant** et **dans les 60 jours** suivant la découverte de l'erreur.

- Dans les autres cas, le montant de l'assurance est **ajusté** selon le rapport de la prime perçue et de celle qui aurait dû être perçue.

- L'**âge véritable** est crucial pour la fin d'un contrat d'assurance vie si l'assurance prend fin à un âge spécifié et la fausse déclaration est découverte **avant le décès** de l'assuré.

- En assurance contre la maladie ou les accidents, l'âge véritable détermine le début ou la fin du contrat si cela est stipulé. L'assureur peut alors ajuster la prime pour qu'elle corresponde à l'âge réel de l'assuré.

### 5.10.4 Suicide de l’assuré

- Selon l'**article 2441 du Code civil**, l'assureur peut refuser de payer une prestation de décès en cas de **suicide** si :
  - Le suicide a lieu **dans les deux premières années** d'assurance.
  - Une **clause d'exclusion concernant le suicide** est incluse dans le contrat.

- Tout **montant additionnel d'assurance** dû à une modification du contrat est également soumis à une **clause d'exclusion de 2 ans** à compter de la prise d'effet de l'augmentation.
- Cela signifie que certaines réclamations liées à cette augmentation pourraient ne pas être couvertes durant cette période.


## 5.11 Dispositions touchant le bénéficiaire du contrat d’assurance vie

- Le **titulaire du contrat** nomme habituellement un **bénéficiaire**. 

- Si **aucun bénéficiaire n'est nommé**, la **succession** de l'assuré est considérée comme bénéficiaire.

- Deux principaux types de bénéficiaires :
  - La **succession**.
  - Une **personne** (physique ou morale).

### 5.11.1 Mode de désignation du bénéficiaire

- Le **bénéficiaire** peut être désigné de trois manières :
  - Directement **dans la police** d'assurance.
  - Dans un **écrit distinct**.
  - Dans un **testament**.

- Habituellement, la désignation se fait via un **formulaire de désignation de bénéficiaire**, annexé à la police.

### 5.11.2 Succession

- La **succession** devient bénéficiaire si le titulaire du contrat :
  - La désigne explicitement en utilisant les termes « **succession** », « **héritiers légaux** », « **ayants droit** », ou « **légataires** ».
  - Ne nomme **aucun bénéficiaire** spécifique (par exemple, si le bénéficiaire nommé est décédé).


- Désigner la succession comme bénéficiaire est **simple**, mais cela signifie que les sommes versées **peuvent être réclamées par les créanciers** de la succession.

Supposons que vous ayez une assurance-vie de 50 000 \\$ et que vous deviez 20 000 \\$ en dettes diverses au moment de votre décès. Si votre succession est désignée comme bénéficiaire, les créanciers peuvent réclamer jusqu'à 20 000 \\$ de l'assurance pour régler vos dettes, laissant seulement 30 000\\$ à être distribués selon votre testament ou les lois de succession.

### 5.11.3 Personne physique ou morale

La désignation d’une personne physique ou morale à titre de bénéficiaire comporte des avantages et des inconvénients.
**Avantages :**

**Avantages :**
- Le capital assuré est versé plus rapidement.
- Le capital assuré n’est pas accessible aux créanciers de la personne décédée.
- La valeur du contrat d’assurance pourrait être insaisissable par les créanciers du preneur.

**Inconvénients :**
- Risque d’oublier de modifier la désignation de bénéficiaire lorsque le titulaire change d’idée.
- Risque d’être coincé avec un bénéficiaire irrévocable sans son consentement.
- Si le bénéficiaire est nommé par testament, la rédaction du testament doit être très précise.

### 5.11.4 Décès du bénéficiaire

- Si le **bénéficiaire prédécède l'assuré**, le **titulaire** (s'il n'est pas lui-même l'assuré) devient le nouveau bénéficiaire.

Si le titulaire et l'assuré sont des personnes différentes : Dans ce cas, si le bénéficiaire désigné décède en premier, le titulaire du contrat (la personne qui possède et contrôle la police) devient automatiquement le nouveau bénéficiaire. Cela signifie que les bénéfices de l'assurance lui seront versés à la mort de l'assuré.

- Si le **titulaire est l'assuré** et décède avant le bénéficiaire, la **prestation de décès** sera versée à **sa succession**.

Si le titulaire est aussi l'assuré : Si cette personne décède (et qu'elle est également le titulaire du contrat), la prestation de décès de l'assurance est versée à sa propre succession. Autrement dit, l'argent de l'assurance entre dans la masse des biens du défunt et sera distribué selon son testament ou les lois de succession, si aucun testament n'existe.

- Si **l'assuré et le bénéficiaire décèdent en même temps** ou dans des circonstances ne permettant pas d'établir l'ordre des décès, l'assuré est réputé avoir **survécu au bénéficiaire** aux fins de l'assurance.

Si l'assuré et le bénéficiaire décèdent en même temps ou dans des circonstances ambiguës : Dans de tels cas, la loi présume que l'assuré a survécu au bénéficiaire. Cela signifie que la prestation de décès ne sera pas versée comme si le bénéficiaire avait survécu à l'assuré, mais ira plutôt à la succession de l'assuré ou au titulaire du contrat s'il est différent de l'assuré.

### 5.11.5 Subrogation

Pour éviter les problèmes qui découlent du **décès simultané** de l’assuré et du bénéficiaire, on peut nommer un **bénéficiaire subrogé** (bénéficiaire de remplacement) ou un **titulaire subrogé** (titulaire de remplacement).

### 5.11.6 Révocabilité du bénéficiaire

- **Le bénéficiaire peut être révocable ou irrévocable, avec des règles spécifiques au Québec :**
  - **Conjoint marié** : Irrévocable, sauf en cas de stipulation écrite contraire.
  - **Conjoint en union civile** : Irrévocable, sauf en cas de stipulation écrite contraire.
  - **Conjoint de fait** : Révocable.
  - **Autre personne** : Révocable.


- **Le divorce ou la dissolution de l'union civile annule toute désignation du conjoint comme bénéficiaire.**


S’il y a révocation du bénéficiaire, la révocation doit se faire par écrit. Cependant, toute désignation demeure révocable tant que l’assureur ne l’a pas reçue.

## 5.11 Dispositions touchant le bénéficiaire du contrat d’assurance vie

Le titulaire du contrat nomme habituellement un bénéficiaire. Si aucun bénéficiaire n’est nommé, la succession de l’assuré est considérée comme le bénéficiaire de l’assurance. Il existe deux grands types de bénéficiaires : la succession et une personne (physique ou morale).

### 5.11.1 Mode de désignation du bénéficiaire

Le bénéficiaire peut être désigné dans la police, un écrit distinct, ou un testament. Habituellement, le bénéficiaire est désigné sur un formulaire annexé à la police.

### 5.11.2 Succession

La succession est bénéficiaire quand désignée explicitement ou par défaut (si aucun bénéficiaire n'est nommé ou si le bénéficiaire nommé est décédé). Les sommes versées à la succession sont accessibles aux créanciers de cette dernière.

### 5.11.3 Personne physique ou morale

**Avantages de désigner une personne physique ou morale comme bénéficiaire :**
- Le capital assuré est versé plus rapidement.
- Le capital assuré n’est pas accessible aux créanciers de la personne décédée.

**Inconvénients :**
- Risque d’oublier de modifier la désignation.
- Risque avec un bénéficiaire irrévocable sans son consentement.
- Si désigné par testament, la rédaction doit être précise.

### 5.11.4 Décès du bénéficiaire

Si le bénéficiaire décède avant l’assuré, le titulaire devient le nouveau bénéficiaire. Si l’assuré et le bénéficiaire décèdent simultanément, l’assuré est réputé avoir survécu au bénéficiaire pour l'assurance.

### 5.11.5 Subrogation

Nommer un bénéficiaire subrogé peut éviter les problèmes liés au décès simultané de l’assuré et du bénéficiaire.

### 5.11.6 Révocabilité du bénéficiaire

La désignation du bénéficiaire peut être révocable ou irrévocable, avec des conséquences sur la saisissabilité des prestations de décès.

| Distinctions               | Police          | Écrit distinct  | Testament      |
|----------------------------|-----------------|-----------------|----------------|
| Conjoint marié             | irrévocable¹    | irrévocable¹    | révocable      |
| Conjoint en union civile   | irrévocable¹    | irrévocable¹    | révocable      |
| Conjoint de fait           | révocable¹      | révocable¹      | révocable      |
| Autre personne             | révocable¹      | révocable¹      | révocable      |

¹: Sauf stipulation écrite contraire
La désignation de bénéficiaire contenue dans un testament n’est pas opposable à une désignation plus récente ou même antérieure, sauf mention explicite de la police concernée dans le testament.
Au Québec, le propriétaire de la police n’a pas besoin de l’autorisation d’un bénéficiaire irrévocable pour accéder à la valeur de rachat, sauf si la police le stipule.

## Annexe au Module 5 : Lexique de certains termes d’assurance

- **ACCAP (CLHIA)** : Association canadienne des compagnies d’assurance de personnes.
- **Assurance-vie** : Assurance prévoyant le paiement d’une somme au décès de l’assuré.
- **Assurance vie à prime nivelée** : Assurance dont la prime reste la même pendant toute la durée du contrat.
- **Avance sur contrat** : Avance consentie par l’assureur, garantie par la valeur de rachat d’une police.
- **Bénéficiaire** : Personne destinée à recevoir les sommes dues au décès de l’assuré.
- **Capital assuré** : Somme indiquée à la police, versée à l’échéance ou au décès de l’assuré.
- **Participation** : Somme tirée des bénéfices de l’assureur et versée annuellement au souscripteur.
- **Proposition d’assurance** : Déclaration remplie par une personne désirant souscrire un contrat d’assurance.
- **Provisions mathématiques** : Fonds constitués par les assureurs pour respecter leurs engagements.
- **Souscripteur** : Personne qui a souscrit un contrat d’assurance, souvent appelée « titulaire de police ».
- **Valeur de rachat** : Somme payable si le titulaire résilie sa police avant l’échéance.
