# 1.1 Définition de l’actuaire et de l’actuariat

En tant que futurs actuaires, vous serez fréquemment confrontés à la question : « Qu’est-ce qu’un actuaire ? ». Il est important de pouvoir y répondre de manière concise et claire, en particulier quand la question est posée par le grand public.

> _Essentiellement, un actuaire est un spécialiste de l’analyse des conséquences financières d’événements incertains et futurs._

Autrement dit, notre tâche consiste habituellement à estimer les coûts
ou les gains qui découleront d’événements dont on ne sait pas s’ils se
produiront ou quand ils se produiront.

## Quelqes exemples

### Exemple 1

Un actuaire évalue des scénarios financiers tels que le montant nécessaire à épargner pour un revenu mensuel de 100$ à partir de 65 ans jusqu'à la fin de vie, incluant le total à accumuler d'ici cet âge et la cadence d'épargne requise pour atteindre ce montant.

Il doit donc se poser certaines questions comme:  

* Combien il faut avoir accumulé d’argent à 65 ans?  

* À quel rythme devez-vous accumuler de l’argent pour avoir les fonds nécessaires à 65 ans?


Ici, l’incertitude vise les éléments suivants :  

-   Quel rendement ferez-vous sur vos placements ?  

-   Survivrez-vous jusqu’à 65 ans ?  

-   Combien de temps vivrez-vous après 65 ans ?  

### Exemple 2

Si vous voulez un revenu mensuel de 2 000 \$ pour le reste de votre vie, à compter d’un certain âge de de retraite :  

* Combien devez-vous avoir accumulé d’argent si vous désirez prendre votre retraite à 60 ans ?  

* Comment varie cette somme si vous repoussez votre retraite à 65 ans, à 67 ans ou à 70 ans ?  

* À quel rythme devez-vous accumuler de l’argent pour avoir les fonds nécessaires ?

Ici, l’incertitude vise les éléments suivants :  

* Quel rendement ferez-vous sur vos placements ?  

* Survivrez-vous jusqu’à l’âge prévu de votre retraite ?  

* Combien de temps vivrez-vous après le début de votre retraite ?

Les deux questions ci-dessus mènent à une interrogation sociale : notre société a-t-elle les moyens de continuer à faire débuter le paiement de la Pension de sécurité de la vieillesse (une allocation universelle) à 65 ans ? Devrait-on viser plutôt 67 ans ? Ou même 70 ans, considérant la hausse de la dette nationale causée par la crise de la COVID-19 ? La pénurie actuelle de main-d’œuvre change-t-elle la réponse ?

###  Exemple 3

Si vous voulez une assurance vie de 100 000\$ qui vous couvre \>
    jusqu’à votre décès, quel qu’en soit l’âge :

* Quelle prime devriez-vous payer?

Ici, l’incertitude vise les éléments suivants :

* Quel rendement l’assureur réalisera-t-il sur vos primes ?

* Quelle sera votre probabilité de décès à chaque âge ?  

    (NB Celle-ci varie selon votre âge, votre sexe et votre état de
    santé, et selon les progrès futurs de la science médicale)

###  Exemple 4

Si vous voulez une assurance contre le vol :

* Quelle prime devriez-vous payer?


Ici, l’incertitude vise les éléments suivants :

* Quelle est la valeur des biens susceptibles d’être volés ?

* Quelle est la probabilité de vol dans votre quartier …

* … ou dans la zone délimitée par votre code postal ?

> _« Un ou une **actuaire,** du latin « *actuarius* » signifiant un fonctionnaire qui faisait la tenue de livres comptables dans l'antiquité romaine, ***est un professionnel spécialiste de l'application du calcul des probabilités et de la statistique aux questions d'assurances, de prévention, de finance et de prévoyance sociale**.* »_


> _« À ce titre, il ou elle analyse l'impact financier du risque et estime les flux futurs qui y sont associés. L'actuaire utilise des techniques **mathématiques**, issues principalement de la théorie des probabilités et de la statistique, pour décrire et modéliser de façon prédictive certains évènements futurs tels que, par exemple, la durée de la vie humaine, la fréquence des sinistres ou l'ampleur des pertes

>_Un expert en assurance IARD, réassurance, finance, gestion des risques et gestion des risques d’entreprise_

## 1.1.6 En résumé


En résumé, l’actuaire est un expert en mathématiques financières, qui évalue les conséquences financières d’événements incertains, principalement dans les domaines reliés à l’assurance (incluant la réassurance), aux régimes de retraite et à la gestion des risques, ce qui rejoint assez bien la définition donnée en précédemment.

### Risque :


En actuariat, un risque est essentiellement un événement dont la survenance est aléatoire (due au hasard) et cause une perte qu’on peut chiffrer monétairement. On verra plusieurs définitions de «risque» dans un module subséquent mais la présente définition résume assez bien ce dont il s’agit.

### Aversion pour le risque

L’aversion pour les risques est une attitude qui s’exprime par le fait qu’on est prêt à payer plus que le coût « normal » d’un risque pour s’en débarrasser. Par exemple, je peux me débarrasser du risque d’incendie et de perte de ma maison qui vaut 500 000\\$ en achetant une assurance contre l’incendie. Si l’espérance mathématique du risque d’incendie est de 250\\$ (soit 1 chance sur 2 000 de subir une perte de 500 000\\$), j’ai de l’aversion pour le risque si je suis prêt à payer plus que le juste coût de 250\\$ pour acheter l’assurance. L’aversion pour le risque est considérable quand la prime est minuscule comparativement à la perte maximale potentielle.


### Assurance

L’assurance est un mécanisme par lequel une personne transfère un risque à une société d’assurance en contrepartie du paiement d’une prime à cette même société. 

L’assurance se sépare en deux sous-domaines principaux : Les assurances de personnes et Les assurances IARD (Incendie Accidents Risques Divers). 

1. Les assurances de personnes sont celles qui versent une somme d’argent (appelée «prestation») en cas de décès, de maladie, de blessure ou d’incapacité de travailler de la personne assurée. 

2. Les assurances IARD sont celles qui versent une somme d’argent (appelée « prestation ») en cas de dommages matériel subis par les biens de la personne assurée ou en cas d’obligation, par la personne assurée, d’indemniser une autre personne en raison de sa responsabilité civile envers cette autre personne.
rsonne.


### Réassurance


La réassurance est un mécanisme par lequel une société d’assurance transfère un risque à une autre société d’assurance (appelée le réassureur) en contrepartie du paiement d’une prime à cette autre société.


### Régimes de retraite


Un régime de retraite est un mécanisme d’accumulation d’argent au nom d’une personne dans le but de lui produire un revenu quand elle aura cessé de travailler. 

Certains régimes prévoient que les fonds seront transférés à la retraite dans un mécanisme de décaissement au choix de la personne en question alors que d’autres régimes ont leur propre mécanisme de décaissement et vont jusqu’à promettre un revenu bien précis à la retraite, en fonction du salaire gagné et du nombre d’années de participation au régime.

## 1.2 L’univers des services financiers


Par sa formation, l’actuaire est souvent appelé à travailler dans l’univers des services financiers. Cet univers comprend :


- L’assurance (de personnes ou de dommages, individuelle ou collective)

- Les produits de placement

- Les régimes de retraite


C’est un domaine essentiel de la société car on a tous besoin d’épargner sous une forme ou une autre, pour assurer sa propre sécurité financière. Toutefois, c’est un domaine qui n’a pas toujours bonne réputation, car, d’une part, il est mal compris et d’autre part, il attire parfois des individus peu scrupuleux qui n’hésitent pas à bafouer les lois et règlements. Ces individus sont très rares, heureusement.


### 1.2.1 Mauvaise compréhension du secteur financier par le public

La mauvaise compréhension d’une partie du public vient à la fois de la concurrence avec le secteur bancaire et du fait que, dans le secteur financier, les gens qui travaillent auprès de la clientèle sont souvent payés à commission (ils reçoivent donc un pourcentage des sommes confiées par les clients). 

Le client qui achète une assurance, ou des produits de placement, peut trouver exagéré de payer une commission de 2% à l’achat de parts de fonds de placement, en plus de payer des frais de gestion annuels qui peuvent atteindre un autre 2% ou d’apprendre que son représentant reçoit une commission plus élevée que la prime de la première année de sa police d’assurance vie permanente. Surtout lorsqu’il compare avec une banque qui ne prend aucune commission sur les dépôts qu’elle reçoit…

Évidemment, ce client ignore que la Banque fait son profit autrement (elle prête l’argent déposé à un taux d’intérêt plus élevé que celui qu’elle crédite sur les dépôts et elle perçoit des frais de gestion et de transaction sur ses comptes).

### 1.2.2 Rémunération sur base de commissions


Quant aux commissions, il faut comprendre que les représentants en assurance doivent gagner leur vie. Et que si un représentant vend une police permanente qui lui rapporte 1000 \\$ de commissions aujourd’hui, il ne vendra peut-être demain que 3 polices temporaires lui rapportant chacune 50\\$ de commissions. Et il a ses dépenses professionnelles à assumer.

### 1.2.3 Risque de fraude


Ce qui est problématique, c’est le fait que les services financiers, attirant des sommes importantes, offrent des belles occasions de fraude, notamment avec des variantes des stratagèmes de [Ponzi](https://fr.wikipedia.org/wiki/Syst%C3%A8me_de_Ponzi#:~:text=Un%20syst%C3%A8me%20de%20Ponzi%20est,procur%C3%A9s%20par%20les%20nouveaux%20entrants). Une recherche dans Google avec les termes «AMF» et »Ponzi» permet de constater la réapparition régulière de ce type de fraude, dont les exemples les plus patents sont l’affaire [Norbourg](https://fr.wikipedia.org/wiki/Affaire_Norbourg) au Québec (fraude de 155 M\$) et l’affaire [Madoff](https://fr.wikipedia.org/wiki/Bernard_Madoff) (fraude de l’ordre de 65 G \$).

Il est **vital** pour les actuaires de ne jamais s’associer, de près ou de loin, avec des gens susceptibles de commettre de telles fraudes.


## 1.3 Rôle de l’actuaire et domaine de pratique

### 1.3.1 Les principaux domaines de pratique


Plusieurs domaines de pratique différents s’offrent aux actuaires. Nous résumons ci-dessous les principaux :


- Assurance de personnes (individuelle)

- Assurance de personnes (collective)

- Assurance de dommages (Assurance IARD)

- Actuariat corporatif

- Réassurance

- Régimes de retraite

- Organismes publics gérant des programmes sociaux

- Organismes de réglementation

- Gestion des risques d’entreprise

- Enseignement

- Domaines non-traditionnels


Au cours de sa carrière, un(e) actuaire peut travailler dans plusieurs domaines. Toutefois, la spécialisation croissante des domaines de pratique et la nécessité de toujours avoir des connaissances à jour dans un domaine rendent beaucoup plus difficile aujourd’hui de travailler simultanément dans plusieurs domaines.


### 1.3.2 Les différents rôles et fonctions des actuaires


Plusieurs rôles se ressemblent d’un domaine à l’autre, mais même dans ce cas, les actuaires sont susceptibles d’appliquer des techniques différentes ou d’utiliser des outils différents (statistiques, tables actuarielles, etc.) selon leur domaine de pratique.

#### Assurance de personnes (individuelle)


En assurance de personnes, les actuaires accomplissent principalement les tâches suivantes :


- Tarification générale des produits d’assurance vie (temporaire, permanente, universelle, etc.) et d’assurance santé (assurance invalidité prévoyant des paiements périodiques, assurance des maladies graves, assurance des soins médicaux et dentaires, etc.).

- Calcul des provisions que l’assureur doit détenir pour faire face aux obligations futures découlant des contrats d’assurance et de rente qu’il a émis, tant pour les assurés en bonne santé que pour ceux qui sont malades ou invalides (incapables de travailler) ou ceux qui reçoivent des rentes ou auxquels on a promis des rentes.

- Calcul des montants (« participations ») à verser annuellement aux détenteurs de celles, parmi toutes les polices émises, qui comportent une clause de participation en vertu de laquelle une partie des profits générés par l’ensemble de ces polices ou un sous-ensemble de celles-ci est répartie entre les détenteurs de ces polices.

- Analyses d’expérience.

- Développement et tarification des nouveaux produits.

#### Assurance de personnes (collective)


En assurance collective de personnes, les actuaires accomplissent principalement les tâches suivantes :


- Tarification générale des produits d’assurance vie (habituellement temporaire) et d’assurance santé (principalement l’assurance invalidité prévoyant des paiements périodiques et l’assurance des soins médicaux et dentaires, etc.).

- Tarification particulière des produits d’assurance vie (habituellement temporaire) et d’assurance santé (principalement l’assurance invalidité prévoyant des paiements périodiques et l’assurance des soins médicaux et dentaires, etc.) applicable à un groupe en particulier, en tenant compte de l’expérience de ce groupe et de la crédibilité de cette expérience.

- Calcul des ristournes payables aux groupes dont la police comporte une clause prévoyant le paiement d’une telle ristourne en cas de bonne expérience.

- Analyse des demandes d’assurance présentées par des courtiers et autres intermédiaires (incluant des actuaires-conseils) relativement à des groupes, en tenant compte des caractéristiques démographiques et de l’expérience de réclamations de ces groupes.

- Calcul des provisions que l’assureur doit détenir pour faire face aux obligations futures reliées aux assurés frappés d’invalidité et pour lesquels l’assureur s’est engagé à payer des prestations périodiques ou à renoncer au paiement de certaines primes.

- Analyse d’expérience.

- Développement et tarification des nouveaux produits (très rarement car il y a très peu de nouveaux produits en assurance collective).

- Si l’actuaire travaille dans une firme de courtage ou de consultation, il va préparer des appels d’offres et analyser les soumissions produites par les assureurs en vue de remporter le contrat d’assurance du client de l’actuaire.


#### Assurance de dommages (IARD)


En assurance IARD, les actuaires accomplissent principalement les tâches suivantes :


- Tarification générale des produits d’assurance automobile, habitation, agricole, commerciale, etc.

- Identification et analyse de nouvelles variables de tarification (véhicules électriques, variables télématiques telles que le respect des limites de vitesse, par exemple).

- Calcul des provisions que l’assureur doit détenir pour faire face aux obligations futures découlant des contrats d’assurance.

- Développement et tarification des nouveaux produits.

#### Actuariat corporatif


En actuariat corporatif, les actuaires accomplissent principalement les tâches suivantes :


- Calcul de la valeur intrinsèque des portefeuilles d’assurance de la société qui l’emploie.

- Évaluation des provisions que l’assureur doit détenir pour faire face aux obligations futures découlant des contrats d’assurance et de rente qu’il a émis.

- Préparation et production du rapport de l’Actuaire désigné, indiquant notamment le montant des provisions détenues, des provisions minimales requises et le ratio de solvabilité de l’assureur.

- Examen de la santé financière de l’assureur face aux risques spécifiques à sa société d’assurance et rapport aux autorités de surveillance.

- Tests de sensibilité de la solidité financière de l’assureur face à des événements défavorables très rares (avec probabilité de survenance de l’ordre de ½%).

- Établissement des niveaux cible (niveau minimum, niveau réglementaire, coussin de capital et cible interne) de capitalisation de l’assureur.

#### Réassurance


En assurance de personnes, les actuaires accomplissent principalement les tâches suivantes :


- Tarification des produits d’assurance pour lesquels les assureurs pris individuellement n’ont pas un assez gros volume d’expérience (primes et réclamations) pour que celle-ci soit crédible (c’est-à-dire suffisamment stable et fiable).

- Par exemple, en assurance de personnes, plusieurs assureurs canadiens ont des statistiques insuffisantes quant aux assurances suivantes : l’assurance en cas de décès accidentel ou d’infirmité résultant d’un accident; l’assurance des soins médicaux requis d’urgence lors d’un voyage à l’extérieur du Canada; l’assurance en cas d’invalidité donnant lieu à des paiements périodiques.

- De même en assurance IARD, les assureurs canadiens ont des statistiques insuffisantes quant aux assurances suivantes : l’assurance des risques de tremblement de terre; l’assurance responsabilité civile (résidentielle, automobile et assurance des entreprises); l’assurance maritime; l’assurance des entreprises; etc.

- Conception de produits pour les assureurs clients (couramment appelés « cédantes »).

- Actuariat corporatif (comme ci-dessus).

- Conseils aux cédantes.

#### Régimes de retraite


En régimes de retraite, les actuaires accomplissent principalement les tâches suivantes :


- Calcul du passif actuariel (valeur des obligations futures découlant des rentes garanties aux participants).

- Détermination du « coût normal » du régime.

- Détermination de tout surplus ou déficit actuariel (différence entre le passif actuariel et les actifs détenus par le régime).

- Amortissement du surplus ou déficit actuariel sur la période prescrite par la Loi.

- Préparation de rapports actuariels portant sur les éléments ci-dessus.

- Calcul des valeurs associées au partage des droits afférents à la retraite entre les conjoints lors d’un divorce.

- Calcul de la valeur transférable d’une rente de retraite lorsqu’un employé quitte son emploi (et son régime de retraite).


#### Organismes publics gérant des programmes sociaux (Régime de rentes du Québec, régime d’assurance maladie du Québec, etc.)


Dans ce domaine, les actuaires accomplissent principalement les tâches suivantes :


- Analyses démographiques et projection des coûts futurs sur de très longues périodes.

- Calcul du coût du régime à court et à long terme.

- Estimation du coût des modifications proposées au régime.

- Détermination des taux de cotisation devant être payés pour équilibrer le régime compte tenu des paramètres fixés par la Loi.

- Communication avec le ministère responsable du régime.

#### Organismes de réglementation (Autorité des marchés financiers du Québec, Bureau du Surintendant des institutions financières, Commission des services financiers de l’Ontario, etc.)


Dans ce domaine, les actuaires accomplissent principalement les tâches suivantes :


- Analyse des rapports produits par les actuaires désignés des sociétés d’assurance.

- Vérification des méthodes et procédés utilisés par les actuaires pour calculer leurs provisions actuarielles.

- Participation à l’élaboration de normes et de tests de solvabilité.

- Participation à des interventions auprès des sociétés d’assurance susceptibles d’éprouver des difficultés à rencontrer les critères de réglementation.

- Participation à l’élaboration de lignes directrices en matière de gestion des sociétés d’assurance.

#### Gestion des risques d’entreprise


La gestion des risques d’entreprise est un domaine qui devrait normalement relever de la profession actuarielle mais qui a été principalement développé par la profession comptable. Il s’agit d’un domaine très vaste avec beaucoup de potentiel mais les actuaires sont présentement confinés dans la gestion des risques afférents aux sociétés d’assurance. Les actuaires visent principalement à accomplir les tâches suivantes :


- Identification des risques relatifs à un projet ou à une activité.

- Tests de sensibilité des résultats de l’entreprise selon différents scénarios favorables et défavorables.

- Participation à la détermination de la tolérance de l’entreprise face à différents risques.

- Participation à l’établissement de programmes d’identification et de contrôle des risques.

- Soutien à l’élaboration de programmes de gouvernance d’entreprise.

#### Enseignement


En enseignement, les actuaires accomplissent principalement les tâches suivantes :


- Participation à des comités d’orientation mis sur pied par les sociétés d’actuaires.

- Établissement des programmes de formation des grandes sociétés d’actuaires.

- Développement de syllabus détaillés correspondant aux objectifs pédagogiques établis par les sociétés d’actuaires.

- Recherche et rédaction de matériel pédagogique.

- Préparation d’examens corporatifs, correction des examens et établissement des notes de passage.

- Enseignement universitaire.

    - Recherche.
    - Rédaction de manuels pédagogiques et de notes de cours;
    - Enseignement.
    - Test des connaissances acquises par les étudiants.



#### Domaines non traditionnels


De plus en plus, on rencontre des actuaires dans des domaines connexes à l’actuariat, comme par exemple :


- L’administration de régimes de retraite.

- La gestion de la rémunération et des ressources humaines.

- La gestion d’actifs.

- Le courtage d’assurance.

- Le secteur bancaire.

### 1.5 La rémunération des actuaires


#### 1.5.1 Études et enquêtes salariales

Historiquement, l'Institut canadien des actuaires réalisait des enquêtes salariales, abandonnées depuis plus de 20 ans faute de coopération des employeurs.

Les sources alternatives existantes requièrent une interprétation prudente, souvent manquant de définitions claires et de représentativité.

La rémunération varie grandement par région, avec des écarts notables comme des salaires jusqu'à 20% plus élevés à Toronto qu'à Montréal.

Les principales statistiques salariales peuvent être consultées sur les sites suivants :


* [DW Simpson Salary Survey](https://www.dwsimpson.com/salary)  

* [Glassdoor](https://www.glassdoor.ca/)

Il faut aussi faire attention à ces données car les plus hauts salaires correspondent possiblement à des actuaires qui ne font plus d’actuariat (par exemple, le président de telle ou telle société d’assurance).

De plus, dans le cas de DW Simpson, presque toutes les données sont américaines (seulement 3 tableaux portent sur le Canada).

#### 1.5.2 Information plus ciblée

- **Secteur Public en Ontario :** Salaires supérieurs à 100 000 $ accessibles publiquement pour les secteurs public et parapublic sur [Public Sector Salary Disclosure 2020](https://www.ontario.ca/page/public-sector-salary-disclosure).

- **Agence de Revenu du Canada :** Publie l'échelle salariale des actuaires, applicable potentiellement à tous les actuaires de la fonction publique fédérale, consultable sur [Canada Revenue Agency - Actuarial Science Group](https://www.canada.ca/en/revenue-agency/corporate/careers-cra/information-moved/pay-rates/actuarial-science-group.html).

- **Métiers Québec :** Offre des informations sur les salaires des actuaires, mais manque de détails pour une corrélation avec le niveau de responsabilité, disponible sur [Métiers Québec](https://www.metiers-quebec.org/physique/actuaire.htm).


### 1.4.3 Information numérique


En 2017, l’analyse des différentes sources d’information et des discussions avec des employeurs et d’anciens étudiants laissait deviner les niveaux de rémunération suivants pour des finissants en actuariat (les chiffres de 2022 devraient être 15% à 20% plus élevés, compte tenu de la pénurie de maind’oeuvre) :

- **Salaires en administration de régimes de retraite à Montréal et Toronto :**
  - **Montréal :** Salaire initial entre 42 000 \\$ et 45 000 \\$. Augmente à 65 000 \\$ après quelques années.
  - **Toronto :** Salaire environ 10% plus élevé qu'à Montréal.
  - **Bonus :** Possibilité de bonus jusqu'à 3 000 \$.
  - **Examen d'actuariat :** Les professionnels dans ce domaine ne passent généralement pas les examens d'actuariat ni ne bénéficient du programme d’agrément universitaire de l’Institut Canadien des Actuaires.


- **Salaires pour un analyste en actuariat dans une firme d'actuariat-conseil :**
  - **Salaire initial :** Environ 50 000 \$, potentiellement plus élevé (55 000 \$ à 60 000 \$) en raison de possibles confusions avec les administrateurs de régimes de retraite.
  - **Salaire après plusieurs années :** Peut atteindre jusqu'à 90 000 \$.
  - **Bonus :** Variable, pouvant atteindre 5 000 \$ à 10 000 \$ en fonction du niveau de responsabilité et du salaire de base.


- **Salaires en actuariat dans les sociétés d'assurance de personnes :**
  - **Analyste en actuariat :** Salaire initial d'environ 55 000 \\$ pouvant atteindre 85 000 \\$ après plusieurs années. Boni possible de 5 000 \\$ à 10 000 \\$.
  - **Analyste sénior en actuariat :** Salaire variant entre 65 000 \\$ et 100 000 \\$. Boni similaire de 5 000 \\$ à 10 000 \\$, dépendant du niveau de responsabilité et du salaire de base.


- **Salaires en actuariat dans les sociétés d'assurance IARD :**
  - **Analyste en actuariat :** Salaire de départ entre 55 000 \\$ et 60 000 \\$, avec un potentiel d'atteindre 110 000 \\$ après plusieurs années. 
  - **Bonus :** Possibilité de boni de 5 000 \\$ à 10 000 \\$, en fonction du niveau de responsabilité et du salaire de base.

- Un actuaire sénior gagne typiquement de 100 000 \$ à 150 000 \$, plus un boni pouvant atteindre 15% du salaire.

- **Salaires élevés en actuariat :**
  - **Plafond salarial :** Certains salaires atteignent environ 250 000 \\$, avec des bonis jusqu'à 20% de la rémunération.
  - **Associés directeurs et dirigeants :** Dans les firmes d'actuariat-conseil et les postes de haute direction (Président, vice-présidents), les rémunérations peuvent être nettement supérieures, mais ces postes sont souvent exclus des études salariales.
