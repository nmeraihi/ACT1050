## 1.1 Notions de base qualitatives et quantitatives

### 1.1.1 Notion de rente

- Une rente est une série de versements réguliers effectués par une entité à une autre, souvent pour une durée déterminée.
- Les rentes peuvent être de montant fixe (par exemple, 100$ par mois pendant 10 ans) ou varier, comme dans le cas des rentes indexées à l'inflation.
- Les conditions de versement peuvent changer selon des événements spécifiques, comme le décès du rentier ou de son conjoint.
- Bien que les rentes soient généralement payées mensuellement, les exemples théoriques utilisent souvent des paiements annuels pour simplifier les calculs.
- Le capital initial constitue la base d'une rente, chaque versement incluant une part de remboursement du capital et des intérêts générés.


### 1.1.2 Utilité des rentes

- Les rentes offrent un revenu régulier aux personnes ne travaillant plus, notamment à la retraite, assurant ainsi leur sécurité financière.
- Elles permettent de répartir une grande somme d'argent sur une longue période, évitant ainsi les dépenses impulsives.
- Les régimes de retraite contribuent à la responsabilité sociale des employeurs envers leurs employés âgés, améliorant leur réputation.
- Un bon régime de retraite est attractif pour les meilleurs talents, offrant un avantage compétitif en termes d'embauche.
- Les contributions des employeurs aux régimes de retraite sont plus avantageuses fiscalement pour les employés que des augmentations de salaire équivalentes.
- Les régimes de retraite bénéficient de frais de placement réduits par rapport aux investissements individuels, optimisant les rendements pour les employés.


### 1.1.3 Rente immédiate et rente due

Une rente est dite « immédiate » si le versement débute à la fin de la première période qui suit le début de la rente.

- Une rente immédiate débute un mois ou un an après son achat, selon la fréquence de versement choisie.
- Une rente due commence immédiatement le jour de l'achat.
- Les termes "immédiate" et "due" n'ont pas de logique intrinsèque.

### 1.1.4 Formes de rentes

#### Rente « certaine »

- Une rente certaine est garantie pour une durée fixée, indépendamment de la survie du bénéficiaire.


| Durée | Valeur présente du versement |  |
|-------|------------------------------|---|
|       | Rente immédiate              | Rente due |
| 0     | 0,000000                     | 1,000000 |
| 1     | 0,961538                     | 0,961538 |
| 2     | 0,924556                     | 0,924556 |
| ...   | ...                          | ... |
| 10    | 0,675564                     | 0,000000 |
| Total | 8,110896                     | 8,435332 |



Remarques :

- Si la personne fait 4% de rendement annuel sur ses placements, elle n’a besoin que de 0,961538 \\$ aujourd’hui pour payer 1,00 \\$ dans un an et de 0,675564 \\$ pour payer 1,00 \\$ dans 10 ans.
- Les chiffres au tableau précédent constituent une progression géométrique. On peut donc les sommer à l’aide de la formule qui donne la somme d’une telle progression.
- Ainsi, si on appelle 
  - « i » le taux d’intérêt
  - « v » l’inverse de 1 + i, donc v = 1 / (1+i)
  - « d » le taux « d’escompte », soit le taux d’intérêt escompté d’un an, donc d = i / (1+i)
  - « n » le nombre de versements annuels,
  
  Donc i = 4%
  v = 1 / 1,04 = 0,96153846
  d = 0,04 / 1,04 = 3,846154%
  n = 10
  
  On aura pour une rente immédiate :
  Valeur présente = (1-v^n) / i = (1- 0,96153846^10) / 0,04 = 8,1109
  et
  pour une rente due :
  Valeur présente :(1-v^n) / d = (1- 0,96153846^10) / 0,03846154 = 8,4353

- Les valeurs présentes des versements montrent combien doit être investi aujourd'hui pour réaliser un paiement futur, avec un taux d'intérêt de 4% composé annuellement.
- La valeur présente d'une rente est calculée en utilisant la formule de somme d'une progression géométrique.
- La "valeur présente" et la "valeur actualisée" sont des termes interchangeables en actuariat.

#### Rente « certaine » différée


Une rente « certaine » est dite « différée » si on l’achète un certain temps avant le début de sa période de versement. Par exemple, si une personne achète les deux rentes décrites précédemment et qu’elle les achète 10 ans avant le début de leur période de versement, et si le taux d’intérêt annuel est toujours de 4%, elles lui coûteront respectivement

$$
8,110896 \times v^{10}=8,110896 \times 0,96153846^{10}=5,479431 \$ \text{ dans le cas de la rente immédiate et}
$$
$$
8,435332 \times v^{10}=8,435332 \times 0,961538466^{10} =5,698608 \$ \text { dans le cas de la rente due }
$$

#### Rente « viagère »

- Une rente viagère est versée pour toute la durée de vie du bénéficiaire, sans limite prédéfinie.
- Le calcul d'une rente viagère intègre les probabilités de survie et l'intérêt composé.
- L'exemple avec un taux d'intérêt de 4% illustre le calcul pour un bénéficiaire avec une espérance de vie maximale de 10 ans.

Remarques :

- La valeur présente du versement est égale au versement (1,00\$), multiplié par le facteur d’intérêt (comparez avec le tableau précédent), et multiplié par la probabilité de survie.
- Pour un même versement de rente, une rente viagère coûte, toutes autres choses étant égales, toujours moins cher qu’une rente certaine, à cause de l’effet de mortalité.
- Comme la rente est viagère, donc payable toute la vie du rentier, le dernier versement de la rente due n’arrive plus un an avant le dernier versement de la rente immédiate. La valeur de la rente due est alors égale à celle de la rente immédiate, plus 1 versement (le premier versement, qui vaut 1,00 \$ car il n’est pas escompté par l’intérêt et ne subit pas l’effet de la mortalité.
- La survie découle essentiellement de l’hypothèse de mortalité qu’on fait. Plus on suppose que les gens meurent vite, moins on suppose que leur survie soit élevée et moindre sera la valeur initiale de la rente.

#### Rente « viagère » avec période garantie

- Une rente viagère avec période garantie assure des paiements minimums, même en cas de décès prématuré du bénéficiaire.
- L'ajout d'une garantie pour les premiers versements augmente la sécurité pour les survivants du rentier.

#### Rente « viagère » différée

- Tout comme une rente certaine, une rente viagère peut être différée, c'est-à-dire achetée avant le début de la période de versement.
- La valeur de la rente est ajustée pour tenir compte du délai avant le commencement des paiements et de la probabilité de survie jusqu'au début de la rente.


**Observations**

- L'achat anticipé d'une rente viagère différée réduit sa valeur présente, due à la combinaison des facteurs de survie et d'intérêt.
- Il est possible d'ajouter une période garantie à une rente viagère différée, ce qui influence le calcul de sa valeur.
- Les régimes de retraite à prestations déterminées utilisent souvent des rentes viagères différées avec garantie, calculées en fonction des années de service et d'un pourcentage du salaire.


## 1.2 La notion de retraite

### 1.2.1 Notion d’espérance de vie

- L'espérance de vie est une mesure statistique basée sur les probabilités de survie à différents âges, utilisant des tables de mortalité.
- Elle varie selon l'âge, le pays, ou le groupe démographique et sert à évaluer la longévité et à planifier les politiques sociales.
- Au Québec, l'espérance de vie à 65 ans a nettement augmenté depuis 1950, reflétant une amélioration de la longévité.

### 1.2.2 Notion de retraite

- La retraite marque le retrait de la population active, souvent soutenu par des épargnes ou des régimes de pension.
- Les raisons de prendre sa retraite incluent l'âge, les années de service, la santé, ou les économies personnelles.
- La retraite peut être ajustée aux programmes sociaux et varier en fonction de l'espérance de vie, ce qui influence le financement et la durée des prestations.

## 1.3 Les risques reliés à la retraite

Le travail d’actuaire est centré autour de l’existence des risques et de la probabilité que des risques surviennent immédiatement ou dans le futur.

- Le risque principal en retraite est la longévité, soit vivre plus longtemps que son épargne.
- L'espérance de vie augmente, posant un défi pour le financement des retraites.

### 1.3.1 Le risque de longévité

Le principal risque relié à la retraite est le risque de longévité, autrement dit le risque de vivre plus vieux que son épargne.

Voici un exemple : avec un taux d’intérêt de 2% et un capital de 1 000 000 \$ à 65 ans, une personne peut acheter :

- Une rente due d’une durée de 10 ans, au montant de 109 144 \$ / année, ou
- Une rente due d’une durée de 15 ans, au montant de 76 299 \$ / année, ou
- Une rente due d’une durée de 20 ans, au montant de 59 957 \$ / année, ou
- Une rente due d’une durée de 25 ans, au montant de 50 216 \$ / année, ou
- Une rente due d’une durée de 30 ans, au montant de 43 774 \$ / année, ou
- Une rente due d’une durée de 35 ans, au montant de 39 218 \$ / année.

- Les rentes viagères transfèrent le risque de longévité à un assureur, garantissant un revenu à vie.
- Les actuaires jouent un rôle clé dans la gestion du risque de longévité, utilisant des tables de mortalité pour leurs calculs.
- Statistiquement, une part significative de la population peut dépasser l'espérance de vie moyenne, soulignant l'importance de planifier pour une longue retraite.

### 1.3.2 Le risque d’inflation

Le risque d’inflation représente le danger d'une diminution du pouvoir d'achat de l'argent.

- Le risque d'inflation concerne la perte de pouvoir d'achat due à la hausse des prix.
- Le Canada a expérimenté divers taux d'inflation, avec des pics significatifs dans le passé.

- L'inflation peut réduire drastiquement le pouvoir d'achat d'un revenu fixe au fil du temps.
- L'inflation est un phénomène cyclique, où la hausse des prix entraîne une augmentation des salaires, qui à son tour pousse les prix à la hausse.
- La Banque du Canada vise une inflation annuelle entre 1% et 3% pour maintenir la stabilité économique.
- L'indexation des rentes à l'inflation est possible, mais souvent limitée et coûteuse, surtout dans un environnement de faibles taux d'intérêt.

## 1.4 L’aspect sociétal

### 1.4.1 Le vieillissement de la population


- L'introduction de la contraception en 1965 a entraîné une baisse marquée de la natalité au Canada et au Québec, passant de 30 à 10,5 naissances annuelles par 1000 habitants de 1925 à après 2000.
- Le "baby bust", ou déclin de la natalité post-1965, a eu un impact plus significatif sur la démographie que le "baby boom" de 1946-1957.
- Malgré le baby bust, la génération des baby-boomers reste la plus importante en raison de la forte croissance démographique avant 1950 et de la chute de la natalité postérieure.

### 1.4.2 Les preneurs du risque de longévité

Rares sont les individus qui ont accumulé assez d’argent à leur retraite pour assumer eux-mêmes leur risque de longévité. Par ailleurs, la structure actuelle de la population limite la capacité des jeunes générations à assumer le risque de longévité des plus vieilles générations.

#### L’État
- Fournit un revenu de base aux citoyens âgés avec des revenus faibles, assumant le risque de longévité.
- Gère le RRQ et le RPC, qui visent à remplacer environ 25% du revenu gagné par les travailleurs.

#### Les employeurs
- Certains régimes de retraite offerts par les employeurs permettent l'accumulation d'épargne jusqu'à l'achat d'une rente.
- D'autres garantissent une rente basée sur le salaire et les années de service, prenant en charge le risque de longévité.

#### Les assureurs
- Seuls à offrir des rentes viagères, assumant pleinement le risque de longévité.
- Leur produit distinctif est le transfert de risques, comme le risque de mortalité et de longévité.

#### Les individus
- Ceux qui ne convertissent pas leurs épargnes en rente à la retraite assument eux-mêmes leur risque de longévité.
- Garder une partie de l'épargne au lieu de tout convertir en rente peut aider à gérer l'inflation et le risque de longévité.


### 1.4.3 Le rôle de l’État

Le rôle essentiel de l’État est de veiller au bien-être de ses citoyens.

#### Revenu de base à la retraite
- L'État s'assure que les personnes âgées, notamment celles ayant contribué à la richesse collective, disposent d'un revenu de base à la retraite.
- Il gère deux situations principales : les retraités n'ayant pas ou peu travaillé, et les travailleurs retraités.


#### Les personnes n’ayant pas travaillé ou très peu seulement
- Au Canada, l'allocation universelle, la Pension de sécurité de la vieillesse, offre un peu plus de 8 732 \\$ / année (727,67 \\$ / mois) pour les 65 à 74 ans et environ 9 605 \\$ / année (800,44 \\$ / mois) pour les 75 ans et plus
- Pour ceux sans autre revenu, un Supplément de revenu garanti d'environ 13 043 \\$ / année (1 086,88 \\$ / mois) est fourni, surtout pour ceux n'ayant jamais travaillé

#### Revenu de base à la retraite
- La PSV et le SRG remplacent jusqu'à 35% du salaire moyen pour ceux sans revenu de travail, offrant une base pour ceux n'ayant jamais cotisé au RRQ ou RPC.
- La rente du RRQ/RPC représente environ 25% du revenu moyen pré-retraite, avec une bonification potentielle jusqu'à 33,3% à long terme.

#### Règlementation
- L'État établit des lois pour assurer la solvabilité des régimes de retraite et garantir l'avantage fiscal des sommes accumulées.
- Les règlements visent aussi à fournir aux participants un accès à l'information et une participation à la gestion de leur régime de retraite.


### 1.4.4 La place des régimes de retraite parmi les avantages sociaux en contexte de rémunération globale

- Les régimes de retraite sont considérés comme du salaire différé, constituant une part des avantages sociaux offerts par l'employeur.
- Ils s'inscrivent dans la rémunération globale, qui comprend le salaire, les avantages sociaux et d'autres composantes, et peuvent faire l'objet d'arbitrages ou de négociations collectives.

## 1.5 Les outils servant au calcul de la valeur d’une rente

### 1.5.1 Taux d’actualisation
- Le taux d'actualisation permet d'escompter la valeur des versements futurs pour calculer la "valeur présente" de la rente.
- Il reflète le rendement attendu sur les placements soutenant la rente, minoré des frais de placement, d'une marge de profit, et du risque de fluctuation des taux d'intérêt.


### 1.5.2 Tables de mortalité
- Les tables de mortalité évaluent la probabilité de survie du rentier pour les rentes viagères et les rentes conditionnelles à la survie.
- Le choix de la table doit refléter les caractéristiques de la population et prévoir les améliorations de la mortalité, avec des marges pour l'anti-sélection et des méthodes pour projeter les taux de mortalité dans l'avenir.

## 1.6 Hypothèse actuarielles

### 1.6.1 Choix des hypothèses

- Avec la généralisation des logiciels permettant de calculer la valeur actualisée d’une rente, le rôle de l'actuaire est de choisir les hypothèses actuarielles pour le calcul de la valeur actualisée d'une rente, selon les normes de l'Institut canadien des actuaires.
- Il est crucial que l'actuaire sélectionne des hypothèses raisonnables et appropriées, sans se laisser influencer indûment par des pressions externes.

## 1.7 Rôles remplis par les actuaires dans le domaine des régimes de retraite

### 1.7.1 Conception de régime
- Les actuaires conçoivent des régimes de retraite, aidant à estimer la rente future pour les régimes à cotisations déterminées et établissant le coût pour les régimes à prestations déterminées.

### 1.7.2 Évaluation de régime
- Les évaluations périodiques des régimes à prestations déterminées, régies par les normes de l'Institut canadien des actuaires et les lois, sont réalisées par des actuaires qualifiés.

### 1.7.3 Communication
- Les actuaires préparent les relevés annuels pour les adhérents des régimes à prestations déterminées, en assurant l'exactitude et la pertinence des informations transmises.

### 1.7.4 Recherche
- Les actuaires contribuent à la recherche en développant des tables de mortalité et en ajustant les hypothèses aux caractéristiques spécifiques des groupes évalués.
