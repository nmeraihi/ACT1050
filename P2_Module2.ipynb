### Défi du maintien du niveau de vie à la retraite

- Un enjeu clé en matière de retraite est de déterminer le montant nécessaire pour qu'une personne puisse maintenir son niveau de vie à la retraite, en tenant compte de la disparition des dépenses liées à l'emploi.


#### Exemples de dépenses liées à l'emploi

- Frais de transport entre le domicile et le lieu de travail, incluant les coûts de transport en commun, de voiture, d'essence et de stationnement.
- Coût des repas pris à l'extérieur de la maison.
- Achat de vêtements conformes au code vestimentaire exigé par l'employeur, non applicable pour certaines professions (ex : actuaire, réceptionniste).
- Contributions aux programmes sociaux comme le Régime de rentes du Québec ou de Pensions du Canada, Assurance emploi, et assurance parentale.
- Cotisations aux régimes de retraite et aux avantages sociaux offerts par l'employeur.

#### Considérations fiscales

- La plupart de ces dépenses ne bénéficient d'aucun avantage fiscal (ni déduction, ni crédit d'impôt) et sont donc couvertes par des fonds post-imposition.
- La progressivité des taux d'imposition, où les taux augmentent avec le revenu imposable, peut avoir un impact significatif sur le calcul du revenu nécessaire à la retraite pour maintenir le niveau de vie.

## 2.2 Les composantes du revenu à la retraite

Une personne qui cesse de travailler continue d’encourir des dépenses pour son logement, sa nourriture, ses soins, ses vêtements, ses déplacements et ses loisirs. Elle a donc besoin de revenus pour subvenir à ces besoins.


Dans les pays développés, ces revenus proviennent généralement de trois sources :

- Des régimes d’État :
  - Régime de retraite universel
  - Allocations universelles
  - Aide sociale

- Des régimes de retraite organisés par des employeurs ou des associations de travailleurs
- Des épargnes personnelles

### 2.2.1 Programmes d’État (les programmes canadiens sont vus en détails au Module 3)

- **Régimes de retraite universels**

Les programmes d'état des pays occidentaux incluent généralement un régime de retraite universel, accessible à tous les travailleurs. Ces régimes font partie des régimes d'assurance sociale et ont les caractéristiques suivantes :

  - Transfert d'un risque à un organisme gouvernemental ou supervisé par l'État.
  - Établis par une loi.
  - Participation obligatoire des travailleurs.
  - Cotisation par les travailleurs et/ou les employeurs.
  - Auto-financement à long terme par un fonds distinct.

Les détails des régimes varient d’un pays à l'autre mais visent généralement à remplacer une portion des revenus antérieurs à la retraite. Par exemple :

  - **En France**, l’assurance vieillesse vise à remplacer 50% de la moyenne des 25 meilleurs salaires annuels.
  - **Au Royaume-Uni**, la pension d’état est d'environ 185 £ par semaine après 35 années de contributions.
  - **Aux États-Unis**, la Social Security Administration propose une rente basée sur les gains annuels indexés.
  - **Au Canada**, les régimes de Rentes du Québec (RRQ) et de Pensions du Canada (RPC) prévoient une rente d'environ 25% de la moyenne des gains admissibles.

- **Allocations universelles**

Une allocation universelle est :

  - Une allocation monétaire basée sur des critères démographiques.
  - Établie par une loi.
  - Financée par les fonds généraux du gouvernement.

Au Canada, la Pension de sécurité de la vieillesse est une allocation universelle qui, combinée au RRQ/RPC, vise à remplacer environ 40% du revenu d'un travailleur moyen.

- **Aide sociale**

L'aide sociale est destinée aux personnes nécessiteuses et peut nécessiter un test de ressources. Elle est typiquement financée par les fonds généraux du gouvernement. Au Canada, cela inclut le Supplément de revenu garanti et l’Allocation au conjoint, qui sont administrés par le gouvernement fédéral.

Les régimes d'État avec une allocation universelle remplacent une proportion plus élevée des bas revenus que des revenus élevés. Par conséquent, pour obtenir un taux de remplacement du revenu constant, un régime privé devra adapter son taux de remplacement en fonction du niveau de revenu.


### 2.2.2 Régimes de retraite collectifs privés

Dans les pays avec une économie "mixte", l'État vise à fournir un revenu de base à la retraite pour éviter que les citoyens deviennent une charge pour la société. En général, les programmes d’État remplacent environ 40% d'un revenu de base.

Il reste à combler deux écarts :
- La différence entre un taux de remplacement de 40% et un taux jugé plus acceptable.
- Un taux de remplacement sur la portion du revenu qui dépasse celui visé par les programmes d'État.

Les régimes de retraite collectifs privés sont là pour combler ces écarts, soit par l'accumulation de fonds pour acheter une rente (régimes à cotisations déterminées) soit par le versement d'une rente viagère basée sur le salaire et les années de participation (régimes à prestations déterminées).

**Du point de vue de l'État**, un régime de retraite est un mécanisme fiscal permettant à l'employé :
- De différer le paiement de l'impôt sur une partie de ses gains jusqu'à la retraite.
- De bénéficier d'un taux d'imposition réduit sur ces gains différés, si son taux d'imposition à la retraite est inférieur à celui durant sa carrière.
- D'accumuler des revenus de placement à l'abri de l'impôt.

L'État bénéficie aussi car :
- Les sommes accumulées seront imposables lors du retrait, assurant des revenus fiscaux futurs.
- Les retraités avec des revenus de régime privé auront moins besoin d'aide sociale, économisant les dépenses étatiques.
- Ces retraités auront un meilleur pouvoir d'achat et consommeront plus, augmentant les recettes fiscales.

**Du point de vue des employeurs**, les régimes de retraite permettent de :
- Fidéliser les employés.
- Offrir une rémunération globale plus attractive pour le recrutement.
- Projeter une image positive de l'entreprise.

**Du point de vue des employés**, les régimes de retraite offrent :
- Une épargne automatique et obligatoire.
- Un report de l'impôt et une accumulation des fonds à l'abri de l'impôt.
- Une contribution de l'employeur au financement du régime.
- La garantie d'avoir de l'argent à la retraite, avec les fonds mis en fiducie et protégés des créanciers.
- Pour les régimes à prestations déterminées, une rente prévisible et pour les régimes à cotisations déterminées, des options de placement potentiellement plus avantageuses.


### 2.2.3 Épargnes personnelles
- Les épargnes personnelles sont la troisième source de revenu à la retraite, se déclinant principalement en trois formes :
  - Les mécanismes d'accumulation avec report fiscal, tels que les REER au Canada et les IRA aux États-Unis, qui favorisent l'épargne retraite par des avantages fiscaux.
  - Les comptes avec avantages fiscaux comme le CELI au Canada, où les revenus de placement sont exonérés d'impôts.
  - Les placements divers non enregistrés, sans avantages fiscaux spécifiques, incluant des investissements comme l'immobilier et les actions.
- Avant l'âge de 71 ans, il est nécessaire de convertir les REER en produits de décaissement, tels que les FERR (la suite des REER), dont les retraits seront imposés.
- Les revenus de placements non enregistrés sont sujets à l'impôt annuellement, et l'impôt sur les gains en capital est dû au moment de l'encaissement, sauf pour la résidence principale.
- Les personnes contribuant à un régime de retraite collectif et réalisant des épargnes personnelles ont de meilleures chances d'atteindre un taux de remplacement de revenu de 65% à la retraite, surtout si elles partent à la retraite à un âge permettant de bénéficier de la pension publique sans pénalité de retraite anticipée.


### 2.3 Notion de retraite dans un régime de retraite

La retraite est le retrait permanent de la population active pour vivre de ses économies ou de ses revenus de placement. Cependant, cela ne signifie pas nécessairement l'arrêt total du travail. De nombreuses personnes poursuivent des activités dans d'autres domaines après leur retraite principale.

La retraite correspond désormais au moment où l'on commence à retirer des fonds accumulés, souvent sous forme de rente d'un régime de retraite ou de l'épargne constituée.

### 2.3.1 Définition de la retraite

Dans le contexte d'un régime de retraite, la retraite est le commencement du retrait des épargnes accumulées. Elle peut être progressive avec un travail à temps partiel tout en percevant une rente, ou totale sans plus travailler pour l'employeur.

### 2.3.2 Retraite obligatoire ou non

Auparavant, la retraite obligatoire permettait de se séparer d'employés moins productifs. Aujourd'hui, au Canada, on ne peut forcer la retraite, sauf preuve d'incapacité à effectuer le travail. L'abolition de la retraite obligatoire n'a pas eu un grand impact, car peu de travailleurs souhaitent travailler au-delà de l'âge "normal" de retraite.

En 2020, l'âge moyen de la retraite était de 62,3 ans dans le secteur public et de 64,3 ans dans le secteur privé ([Statistique Canada](https://www150.statcan.gc.ca/t1/tbl1/fr/tv.action?pid=1410006001&pickMembers%5B0%5D=4.3)).

### 2.3.3 Âge « normal » de retraite

L'âge "normal" de retraite est celui où un employé peut toucher sa rente sans réduction. Une retraite avant cet âge est dite anticipée et les versements sont réduits. À cet âge correspond la "date normale de retraite". Certains régimes définissent plutôt une combinaison d'âge et d'années de service totalisant un certain nombre (comme 85 ou 90) pour déterminer l'âge de retraite.


### 2.4 Principaux types de régimes de retraite

#### Philosophie et objectifs généraux

En matière de régimes de retraite, il existe deux philosophies principales :

- Les régimes **à prestations déterminées** promettent une rente de retraite basée sur des paramètres tels que le salaire de la carrière, les années de service, et l'âge de la retraite, dans le but de remplacer un pourcentage du revenu antérieur.
  
- Les régimes **à cotisations déterminées** reposent sur l'épargne conjointe de l'employeur et de l'employé pour financer la retraite. Le montant de la rente dépendra de l'épargne accumulée, sans garantie de niveau de remplacement du revenu ni de rendement.

#### 2.4.1 Régime à prestations déterminées

**Objectif et cible**

L'objectif d'un régime à prestations déterminées est de remplacer, avec les régimes d'État, un certain pourcentage du revenu pré-retraite. La formule de la rente prend généralement en compte un pourcentage du revenu pré-retraite, réparti sur le nombre d'années de carrière.

**Types de régimes à prestations déterminées**

Un régime à prestations déterminées s'engage à verser une rente basée sur :

- Un salaire de référence, qui peut être :
  1. Le salaire moyen de la carrière,
  2. Le salaire moyen indexé,
  3. Le salaire moyen avec revalorisation ad hoc,
  4. Le salaire moyen des dernières années de carrière,
  5. Le salaire moyen des meilleures années consécutives.
  
- Le nombre d'années de participation au régime.

Les régimes à prestations forfaitaires sont moins courants et prévoient une rente fixe par année de service.

**Exemples de formules de rente**

1. Rente = 1,6% du salaire annuel par année de participation (régime à salaire de carrière).
2. Rente = 1,6% du salaire annuel indexé par année de participation (régime à salaire de carrière indexé).
3. Rente = 1,6% du salaire annuel avec revalorisation ad hoc (régime à salaire de carrière avec revalorisation).
4. Rente = 1,6% du salaire moyen des cinq dernières années par année de participation (régime à salaire final).
5. Rente = 1,6% du salaire moyen des cinq meilleures années consécutives par année de participation (régime à meilleur salaire).

La rente attendue peut varier significativement selon la formule de rente utilisée, ce qui peut causer des surprises au moment de la retraite.


L'actuaire de l'exemple prend sa retraite avec un salaire final de 135 000 $. En fonction de la formule, la rente peut varier grandement, pouvant être une surprise pour ceux qui ne maîtrisent pas les détails de leur régime de retraite.

### Contraintes

### Contraintes liées au changement d'employeur

- La valeur d'un régime de retraite à salaire final ou à meilleur salaire est affectée par la fréquence de changement d'emploi de l'employé. Les régimes analysés supposent que tous les employeurs offrent des régimes identiques sans possibilité de transfert des droits acquis.
- Le changement fréquent d'emploi peut transformer un régime à salaire final en un régime à salaire carrière, réduisant potentiellement les avantages pour l'employé.
- Lors d'un changement d'employeur, l'employé se voit souvent offrir le choix entre conserver sa rente différée ou opter pour une valeur de transfert. Cette dernière peut être transférée au régime de retraite du nouvel employeur ou à un compte d'épargne individuel enregistré, comme un REER au Canada.


#### Acquisition de droits à la rente

- Au Canada, les employés des provinces de Colombie-Britannique, Alberta, Manitoba, Ontario et Québec acquièrent immédiatement un droit à la rente de retraite en adhérant à un régime de retraite, tandis que dans les autres provinces, ce droit doit être acquis au plus tard après 2 ans de participation (3 ans à l'île-du-Prince-Édouard).



#### Évaluation actuarielle

- Le coût d'un régime de retraite dépend de nombreuses variables indépendantes de la volonté de l'employeur, telles que le taux de rendement de la caisse, l'évolution des salaires, l'inflation, la mortalité, l'âge de départ à la retraite et la cessation d'emploi avant la retraite.
- Le coût du régime ne peut être estimé qu'à travers une évaluation actuarielle, qui prend en compte ces variables dans un calcul complexe basé sur des hypothèses.
- Les prévisions peuvent différer de la réalité, entraînant des écarts entre le coût estimé et le coût réel, résultant en un surplus actuariel (favorable) ou un déficit actuariel (défavorable).
- Un déficit actuariel peut être initial (lié au service avant l'entrée en vigueur du régime) ou courant (découlant des écarts entre les hypothèses et la réalité depuis la dernière évaluation).
- L'évaluation actuarielle est cruciale pour estimer le coût normal du régime, le passif actuariel (valeur actualisée des obligations du régime) et la situation de surplus ou déficit actuariel, ainsi que pour déterminer les paiements nécessaires pour amortir tout déficit selon les exigences légales.



#### Cotisation des employés

À cause de leur générosité, les régimes à prestations déterminées nécessitent souvent une cotisation de la part de l’employé. Dans le passé, il y a eu des abus et certains régimes se trouvaient financés à 100% par les cotisations des employés. Ce n’est heureusement plus le cas et les lois exigent habituellement que l’employeur paye au moins la moitié du coût du régime.

#### Cotisation de l’employeur

L’employeur cotise la différence entre le coût du régime, tel que calculé par un actuaire, et la cotisation de l’employé. Habituellement, la cotisation de l’employeur doit représenter au moins 50% du coût du régime.

#### Risques associés aux régimes de retraite

- L'employé bénéficiant d'une promesse de rente n'assume aucun risque direct, mais peut être exposé à des risques indirects, tels que le risque d'inflation, surtout dans un régime à salaire carrière non indexé.
- L'employeur, en revanche, supporte tous les risques liés au régime de retraite, notamment :
  - Le risque que le rendement des fonds soit inférieur aux prévisions de l'évaluation actuarielle.
  - Le risque de sous-estimation de l'augmentation future des salaires.
  - Le risque d'une longévité accrue des retraités au-delà des estimations.
- Ces risques augmentent la probabilité que le régime de retraite soit plus coûteux que prévu.

#### Propriété du surplus actuariel

- En raison de sa responsabilité de provisionner adéquatement le régime, l'employeur peut se considérer comme propriétaire d'un éventuel surplus actuariel, arguant que ce surplus résulte de cotisations excessivement prudentes de sa part.
- L'employeur peut opter pour un congé de cotisations si le régime affiche un surplus et que le retrait de ce surplus est impossible ou non spécifié par les termes du régime.
- Les employés peuvent également revendiquer la propriété du surplus, en particulier si celui-ci découle d'une performance de placement supérieure aux attentes, ce qui n'est pas directement lié à la prudence de l'employeur.
- La question de la propriété des surplus actuariels a été largement débattue, notamment au Québec où la Loi 57 encadre l'utilisation de ces surplus.


### Avantages et inconvénients d’un régime à prestations déterminées

#### Avantages pour l’employé

- L’employé ne supporte aucun risque direct (comme le risque de rendement futur déficient ou le risque de longévité).
- Le taux de remplacement du salaire est garanti dans un régime à salaire final ou à meilleur salaire.
- Pas de nécessité pour l’employé de déterminer l'épargne requise en dehors du régime de retraite.
- Particulièrement avantageux pour ceux qui passent la majorité de leur carrière avec le même employeur.
- Possibilité d'acheter la reconnaissance de service antérieur à l'entrée en vigueur du régime.


#### Inconvénients pour l’employé

- Risques indirects tels que :
  - Le risque d'inflation, qui réduit la valeur des crédits de rente accumulés dans les premières années.
  - Le risque lié au changement d'emploi, qui peut transformer un régime à salaire final en un régime à salaire de carrière si l'employé quitte plusieurs années avant la retraite.
  - Le risque de sous-capitalisation, où la rente promise n'est garantie que si le régime est suffisamment provisionné.
- Impact fiscal potentiellement défavorable en limitant les contributions à des véhicules d'épargne individuels comme le REER au Canada.
- Peut conduire à une certaine désinvolture chez certains employés quant à la nécessité d'épargner individuellement pour la retraite.

#### Avantages pour l’employeur

- Dans certains pays, la possibilité d'avoir un régime peu coûteux en faisant supporter la majorité des coûts par les employés.
- En période d'inflation élevée, un régime à salaire de carrière peut être économiquement avantageux du fait de l'érosion de la valeur des rentes par l'inflation.
- Le régime peut toujours être présenté de manière attrayante aux employés, renforçant la compétitivité dans l'attraction de talents.
- Possibilité de proposer des programmes de retraite anticipée en cas de réduction des besoins en main-d'œuvre.

#### Inconvénients pour l’employeur

- Coûts potentiellement élevés en l'absence d'inflation et incertitude quant au coût exact du régime.
- Coûts d'administration plus élevés que pour un régime à cotisations déterminées.
- Risques de litiges concernant la propriété des surplus éventuels.
- Moins populaire auprès de certains employés qui préféreraient voir un montant précis leur appartenir plutôt qu'une promesse de rente future.
- L'employeur assume les risques directs, comme le risque de rendement insuffisant des placements et le risque de longévité des employés.


### 2.4.2 Régimes à cotisations déterminées

#### Objectif et cible

L'objectif principal d'un régime à cotisations déterminées est d'accumuler le maximum de fonds pour la retraite, sans garantir un niveau spécifique de remplacement de revenu ou un rendement sur l'épargne. Généralement, les fonds accumulés sont utilisés pour acheter une rente viagère à la retraite. À la cessation d'emploi, les sommes peuvent être transférées dans des comptes spécialisés comme un CRI ou un Fonds de revenu viager.

#### Types de régimes à cotisations déterminées

Les régimes à cotisations déterminées sont généralement moins variés. Ils se caractérisent souvent par des cotisations fixes de l'employeur, basées sur un pourcentage du salaire de l'employé, qui peuvent varier selon des critères non discriminatoires comme l'ancienneté ou le taux de cotisation de l'employé.

#### Exemples de formules de cotisation

- Régime simple : 4% du salaire cotisé par l'employeur, sans cotisation de l'employé.
- Autre régime simple : 9% du salaire cotisé par l'employeur et l'employé chacun.
- Régime plus complexe : L'employé choisit de cotiser entre 0% et 4% du salaire, et l'employeur cotise 5% plus un quart de pour cent pour chaque pour cent cotisé par l'employé.
- Régime imitant un régime à prestations déterminées avec des cotisations de l'employeur augmentant avec l'ancienneté de l'employé, et une cotisation fixe de l'employé de 4%.


#### Placements

Les cotisations sont investies soit dans une caisse de retraite pour les grands régimes, soit réparties entre divers fonds communs selon le choix de l'employé.

#### Contraintes

- Les régimes à cotisations déterminées offrent une flexibilité supérieure comparée aux régimes à prestations déterminées, notamment en facilitant le transfert des fonds accumulés en cas de changement d'emploi vers un nouveau régime d'employeur ou un compte d'épargne individuel.
- L'acquisition des contributions de l'employeur est souvent immédiate, éliminant le besoin d'une période d'acquisition prolongée.
- Ces régimes ne requièrent pas d'évaluation actuarielle, simplifiant leur gestion, et les informations relatives aux cotisations et à l'accumulation des fonds sont régulièrement communiquées aux employés.
- L'employeur est habituellement tenu de correspondre aux cotisations de l'employé, mais n'a pas à gérer de surplus ou de déficit actuariel.


#### Risques

- L'employé porte la totalité des risques, y compris le risque de rendements insuffisants des fonds qui pourraient compromettre l'accumulation d'une rente de retraite adéquate.
- Bien que le risque d'inflation soit généralement atténué par l'ajustement à la hausse des rendements des placements en période d'inflation, l'employé reste exposé à la volatilité du marché et à d'autres facteurs économiques influençant le rendement de ses investissements.


#### Avantages pour l’employé

- Simplicité, facilité de compréhension et de comparaison du régime.
- Connaissance permanente de la valeur des fonds accumulés dans le régime.
- Choix personnel des investissements pour les cotisations.
- Absence de risque d'inflation ou lié au changement d'employeur, sauf dans les cas où la cotisation patronale varie avec l'ancienneté.
- Capitalisation toujours complète à 100% du régime.
- Impact direct sur les droits de cotisation à des véhicules d'épargne individuels, tels que le REER au Canada, réduits des cotisations au régime de retraite.
- Flexibilité dans l'achat d'une rente, possibilité d'opter pour une rente indexée.

#### Inconvénients pour l’employé

- Assumption complète des risques directs, tels que le risque de rendement futur insuffisant ou le risque de longévité.
- Difficulté à estimer le taux de remplacement du salaire que le régime fournira.
- Risque de choix d'investissement inappropriés.
- Perception d'injustice due à la variabilité des rentes à l'achat, influencée par les taux et les tables de mortalité au moment de la retraite.

#### Avantages pour l’employeur

- Simplicité du régime et coût connu à l'avance.
- Faible coût d'administration et absence de nécessité de consulter un actuaire.
- Facilité de compréhension du régime pour les employés.

#### Inconvénients pour l’employeur

- Aucun bénéfice des rendements élevés sur les investissements.
- Moindre popularité du régime auprès de certains syndicats qui favorisent les régimes garantissant une rente.
- Incapacité à utiliser le régime pour encourager la retraite anticipée en cas de réduction de la main-d'œuvre.

### 2.4.3 Régimes mixtes

#### Objectif et cible des régimes mixtes

- Les régimes mixtes visent à intégrer les bénéfices des régimes à prestations déterminées avec ceux à cotisations déterminées, proposant ainsi une option hybride.

#### Approches des régimes mixtes

- **Régime à deux volets** : Combine de manière équilibrée les caractéristiques d'un régime à prestations déterminées (ex. rente basée sur 0,8% du salaire annuel par année de service) et d'un régime à cotisations déterminées (ex. contribution de l'employeur de 2% du salaire).
  
- **Régime à prestations cible** : Opère comme un régime à cotisations déterminées mais vise un taux de remplacement du revenu spécifique grâce à des contributions planifiées. Nécessite des évaluations actuarielles pour ajuster les contributions en fonction du montant de la rente escomptée.

#### Contraintes, avantages et inconvénients

- Les **régimes à deux volets** allient les aspects des régimes à prestations déterminées et à cotisations déterminées, mais les avantages et inconvénients de chaque type ne se compensent pas toujours entièrement.

- Les **régimes à prestations cibles** partagent les caractéristiques des régimes à cotisations déterminées, avec l'avantage supplémentaire pour l'employé d'avoir une meilleure prévisibilité du taux de remplacement du revenu à la retraite, contrairement aux régimes strictement à cotisations déterminées.


### 2.5 Le pouvoir d’achat et sa mesure

#### 2.5.1 Notion de pouvoir d’achat

- Le pouvoir d'achat indique la capacité à acquérir des biens et services avec un certain capital ou revenu, crucial dans l'évaluation du revenu de retraite.
- Les rentes de retraite, hormis celles des régimes gouvernementaux, ne sont souvent pas ajustées à l'inflation, ce qui peut diminuer le pouvoir d'achat réel d'un revenu fixe au fil du temps.



#### 2.5.2 L’inflation et sa mesure

- L'inflation représente la diminution du pouvoir d'achat de la monnaie, observable par une augmentation généralisée et soutenue des prix, distincte de l'augmentation du coût de la vie.
- L'indice des prix à la consommation (IPC) est l'outil principal pour mesurer l'inflation, reflétant les changements des prix que les consommateurs rencontrent, basé sur un panier fixe de biens et services.
- Au Canada, l'IPC est publié par Statistiques Canada et est ajusté régulièrement pour refléter les dépenses actuelles des familles. La Banque du Canada vise une inflation de 2% (dans une fourchette de 1 à 3%), essentielle à sa politique monétaire, avec une révision de cette cible tous les cinq ans en collaboration avec le gouvernement fédéral.

### Impact de l'inflation sur le pouvoir d'achat des rentes de retraite

- Un retraité de 1975 recevant une rente mensuelle de 1000 \\$ non indexée verrait son pouvoir d'achat réduit à 23% de ce qu'il était en 1975 d'ici 2015, à cause de l'inflation.
- Un retraité de 2005 avec la même rente de 1000 \\$ maintiendrait un pouvoir d'achat d'environ 850 \\$, reflétant une inflation moins sévère depuis 1995, autour de 2% par an.

### Relation entre inflation et taux d'intérêt

- L'intérêt est le "loyer" de l'argent; prêter à un taux inférieur à l'inflation entraîne une perte de pouvoir d'achat. Un taux d'intérêt égal à l'inflation permet juste de maintenir le pouvoir d'achat.
- Pour obtenir un rendement réel, le taux d'intérêt doit être supérieur à l'inflation, souvent d'environ 1,5% de plus que le taux d'inflation à moyen terme.
- Les investissements en actions visent à rapporter significativement plus que les taux d'intérêt pour justifier le risque, typiquement 3% à 4% au-dessus du taux d'intérêt à long terme.
- Les actifs d'un régime de retraite sont généralement investis dans un portefeuille équilibré incluant revenus fixes et actions, visant un rendement moyen de 2% au-dessus de l'inflation.
- Pour évaluer la valeur actuelle des rentes dans un régime à prestations déterminées, on utilise un taux d'actualisation reflétant le rendement attendu des actifs du régime.


### 2.5.4 Les pièges de l’approche « prestations déterminées »

Les régimes à prestations déterminées sont théoriquement plus avantageux que ceux à cotisations déterminées, surtout en période de faible inflation, car ils garantissent un taux de remplacement du salaire. Cependant, ils comportent certains risques :

- Un régime basé sur le salaire de carrière peut s'avérer insuffisant en période de forte inflation.
- Un régime basé sur le salaire final est profitable si l'employé reste avec le même employeur pendant toute sa carrière, ou du moins les 20 dernières années, faute de quoi la rente peut être dévaluée par l'inflation.
- La plupart des régimes à prestations déterminées ne proposent pas d'indexation automatique des rentes, ce qui peut réduire considérablement le pouvoir d'achat des retraités en cas d'inflation.

## 2.6 Le rôle de l’actuaire

### 2.6.1 Régime à prestations déterminées

Un régime à prestations déterminées exige le travail d'un actuaire pour réaliser une évaluation actuarielle, afin de définir le coût du régime et l'amortissement d'un possible déficit actuariel. Les lois imposent généralement que ces régimes soient évalués par un actuaire qualifié.

### 2.6.3 Régime à cotisations déterminées

Bien que l'intervention d'un actuaire ne soit pas une obligation légale pour les régimes à cotisations déterminées, il est conseillé de faire appel à leurs services pour offrir des conseils sur le niveau de cotisation nécessaire pour atteindre un certain taux de remplacement du revenu.
